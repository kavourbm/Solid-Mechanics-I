## Tensor and Its Transformation

Solid mechanics deals with and quantifies various **types of object**, including force, displacements, stresses, strains, etc. The objects in solid mechanics is constructed as **tensors** that can be studied systematically. This chapter aims to initialize this concept for in-depth discussions in the later chapters. We start with the most elementary object: numbers. 

### Numbers

In dealing with mechanics problem in a scientific manner, we must have the variables and parameters **quantified precisely**. For which, we use numbers and operate on them in a number of ways in a computer. Therefore, understanding the behavior of numbers subject to various operations is fundamentally important to ensure that we obtain is **consistent, predictable, and reliable**. 

Generally speaking, a number is a single element in a [**field**](https://en.wikipedia.org/wiki/Field_(mathematics)) defined with the **standard arithmetic operations** (addition +, subtraction $-$, multiplication $\times$ (often omitted in expression), and defined-division $\div$ (or / or $âˆ’$ or sometimes :), all satisfy necessary conditions including commutativity, associativity, and distributivity for addition). In all our operations in mechanics, we require all these conditions hold for all numbers, which may standing along or as a part of other objects like vector, matrix, or in general tensor.  

A number can be a real-number or a complex-number. Real numbers form the set $ \mathbb{R}$ that is a collection of numbers of different types including zero, all integers, all rational numbers and all irrational numbers, such as $\sqrt{2}, \pi, e$ and $\frac{1+\sqrt{5}}{2}\approx 1.618$ (the golden ratio). All these numbers can be positive or negative,  filling the entire one-dimensional (1D) real numbers space $ \mathbb{R}$. With the addition and scalar multiplication, it is a real vector space. Figure \ref{image:Real_number_line2} shows a schematic 1D line space where lives real numbers. Space $ \mathbb{R}$ is closed under standard arithmetic operations, collectively noted as $\divideontimes$ (addition or subtraction or multiplication or defined-division). This means that any of these operations on any numbers in $ \mathbb{R}$ produces a new number $d$: 

\begin{equation} \label{eqSop}
d = a \divideontimes b  \; \; \;  \forall a \; \text{and} \;  b \in \mathbb{R}
\end{equation} 

which will still be in $\mathbb{R}$. This enables the most fundamental property: **closure**, meaning that no numbers disappear under any of these operations. Thus, a number becomes a building block for more complicated mathematic structures for tensor operations.

Clearly, $\mathbb{R}$ is a ordered field, because all the real numbers can be ranked based on its value. The square of a real number is always positive. 

In theory $\mathbb{R}$ is boundless (bounded by $\pm \infty$) and seamless. In a computer however, it is bounded and with gaps, due to the limited bits in computers. For example, in Numpy this informations for np.float64 can be found using:

```python
print("Largest float in NumPy = ", np.finfo(np.float64).max)
print("Largest float in NumPy = ", np.finfo(np.float64).min)
print("Number of significant decimal digits = ",
      np.finfo(np.float64).precision)
```

Real numbers include integers by default. Because of its exactness in digital representation in computers (within its bits limit \cite{liu2024NF}), integers is also defined as integer type and is often used for indexing, ordering and ranking. 

All complex-numbers form the set $\mathbb{C}$. It is, in a way, an extension from the real set  $\mathbb{R}$, by introducing the imaginary number: $i=\sqrt{-1}$. Any member in $\mathbb{C}$ is made up with real numbers $a$ and $b$ in the form of $c=a+ib$. Set $\mathbb{C}$ is closed under these defined arithmetic operations, **transcendental operations**, such as exponents with generally a $\mathbb{C}$ base and a $\mathbb{C}$ power, and its inverse operation: types of logarithms. Set $\mathbb{C}$ is known as **algebraically closed**, meaning any polynomial of degree $n$ defined in $\mathbb{C}$ has always $n$ roots in $\mathbb{C}$. Simply put: no root will disappear, ensuring solution existence, when algebraic operations are used.  

More detailed discussion on closure properties of numbers and their representation in a computer can be found in \cite{liu2024NF}. The take away is as follows: 

1. When these defined arithmetic operations are used, we should be able to find a solution, if the problem is well posed. **Solution exists** under the defined operations if the solution exists physically. 

2. When algebraical operations are involved, we may need to switch to complex numbers, and the solution could be multiple. In such a case, we may need to choose a physically meaningful one(s). 

3. The most often encountered failure in numerical computations is the undefined division: **division by zero**. 

4. The 2nd often encountered is **under- or overflows**, because of the physical limits of finite digits in a computer. 

Thus, when failing to obtain a solution, check these related issues.  

### Vectors: collection of numbers

#### Vectors defined in a computer

Generally speaking, a vector is a collection of numbers (of real or complex) in properly arranged structure so that its behavior can be managed, analyzed, and computed for desired purposes. Each of these numbers are often called **element** or **component**. Depending on the field of applications, precise definition may vary. In the framework of mechanics in general, it is built from numbers vertically or horizontally. It becomes a building block for more complicated mathematic objects, such as matrices.  

A vector is also viewed as an one-dimensional (1D) array. It can have two forms: line-up numbers called **row-vector**, or a stack of numbers called **column-vectors**.  In theory of the standard linear algebra, row- and column-vectors behave differently, when used in mathematical derivations and operations. A vector in Python is simply a **1D array** by default. The number of the numbers used in forming the vector is called length or and can be queried using code function len(). Its dimension is 1 and can confirmed using ndim(). Consider an array $\mathbf{v}$ with $n$ numbers, or len($\mathbf{v}$)=$n$. It has has a shape of ($n$,), as shown in the following example. 

```python
v = np.array([5, 2, 8])       # create an 1D array with length 3

print(f'Array:{v}, its dimension:{v.ndim}, length:{len(v)} and',
      f'shape:{v.shape}')
```

It is an **ndarray** class in numpy. Readers may use the help() to find out the details about it, including attributes and methods it has. 

```python
# help(v)  # reader may use this to find out the details  
```

Note that an numpy array behaves differently from the row- or column-vector in the standard linear algebra, where the **transpose** of a row-vector becomes a column-vector and vice versa.  However, the transpose of a numpy array have no effect, as shown in the following example.

```python
print(f'Array v.T:{v.T}, its dimension:{v.T.ndim}, length:{len(v.T)}',
      f'and shape:{v.T.shape}')
```

A row-vector needs, when it must be used, to be specifically defined as a matrix with only one row with a shape of (1, $n$). It is a two dimensional (2D) matrix but has only one row. 

```python
v_row = v[None,:] # adds in a new dimension without elements

print(f'Array v_row:\n{v_row}, its dimension:{v_row.ndim},',
      f' length:{len(v_row)}, and shape:{v_row.shape}')
```

Similarly, if column-vector is needed, one must specifically defined as a matrix with only one column with a shape of ($n$,1). 

```python
v_row = v[:,None] # adds in a new dimension without elements

print(f'Array v_row:\n{v_row}, its dimension:{v_row.ndim}',
      f', length:{len(v_row)}, and shape:{v_row.shape}')
```

With this kind of specific definition, the rules of operations follow largely those defined in the standard linear algebra. Care is needed when using the Numpy 1D arrays, because of some different behavior. A more detailed discussion can be found at Section 2.9.13 in Ref.\cite{liu2022ML}.

#### Basic properties of vectors

Vectors are widely used in sciences, engineering, and recently intensively in machine learning. The most basic two features of a vector is its **magnitude** and **direction**. The former is represented graphically by the length and the latter is by an arrow. When we treat a vector as an object in isolation, these two features are sufficient to define the vector.  

A vector in an n-D space $\mathbb{R}^n$ can have a general form of 

\begin{equation} \label{eqBV1}
 \mathbf{v}=\left[ \begin{array}{c} \mathcal{v}_1 \\ \mathcal{v}_2 \\ \vdots \\ \mathcal{v}_n \end{array} \right] 
\end{equation} 

where $\mathcal{v}_i, i=1,2,...,n$ is the $i$th element of the vector on the $i$th axis of the generalized **Cartesian coordinate** system. The zero vector has zero for all its components. 

**Direction of the vector**

The direction of the vector is defined by all its elements in reference to the coordinate system. Each $\mathcal{v}_i$ lives in a 1D space. This means that a vector in $\mathbb{R}^n$ is defined by $n$ numbers each in $\mathbb{R}^1$, ndim($\mathbf{v}$)=1, and len($\mathbf{v}$)=$n$. 

**Magnitude of a vector**

The magnitude or physical length of the vector must be a positive real number. It is denoted as $\|\mathbf{v}\|_2$ and is computed using 

\begin{equation} \label{eqvabs}
\boxed{ \|\mathbf{v}\|_2\equiv \sqrt{\mathcal{v}_1^2 + \mathcal{v}_2^2 +...+ \mathcal{v}_n^2}=\sqrt{\sum_{i=1}^n \mathcal{v}_i^2} }
\end{equation} 

where $\|\mathbf{v}\|_2$ is also known as L2 norm and it is a physical measure of length of the vector. In general, we can use different types of norms, but L2 as the default. Thus we often drop the subscript 2, and denote the magnitude of a vector using $\|\mathbf{v}\|$.

#### The standard basis vectors

#### Definition

Consider space $\mathbb{R}^n$. We can define a set of $n$ unit vectors that can used as basis vectors. This special set of unit vectors is known as the **standard basis vectors** (SBVs), and it is of importance in vector analysis. It can be expressed as follows. 

\begin{equation} \label{eqBVN1}
 \mathbf{i}_1=\left[ \begin{array}{c} 1 \\ 0 \\ \vdots \\ 0 \end{array} \right]; \; \; 
 \mathbf{i}_2=\left[ \begin{array}{c} 0 \\ 1 \\ \vdots \\ 0 \end{array} \right]; \; \; \dots ; \; \;
 \mathbf{i}_n=\left[ \begin{array}{c} 0 \\ 0 \\ \vdots \\ 1 \end{array} \right]  
\end{equation} 

Clearly, each one of these standard basis vectors has length 1, by the definition of Eq.(\ref{eqvabs}). These SBVs form a set of vectors for the coordinate system, and they are always **linearly independent**. This means that any one of these basis vectors in the set cannot be written as a linear combination of the rest vectors in the same set. This can be easily proven by simple observation. Mathematically, we see that in any of the basis vectors, there is always only one entry of unit 1, and all the rest entries are zero. Take any basis vector, say the $i$th. Its $i$th entry is 1. All the remaining basis vectors have only zero at their $i$ entry. Therefore, it is not possible to find a linear combination of the remaining basis vectors to produce a vector that has nonzero at the $i$th entry. Thus, the $i$th SBV is linearly independent of the rest SBVs. **This means that any vector in the space can be written as a linear combination of the BSVs**.

#### The basic arithmetic operations for vectors

Consider a space $\mathbb{R}^n$ that contains all possible vectors with length $n$ and each of these $n$ elements are in $\mathbb{R}$. We now defined the basic properties and operations for these vectors. 

1) **The zero vector:**
The zero vector $\mathbf{0}$ is a vector with **all** its elements being 0. A nonzero vector shall have at least one nonzero element. 

2) **Scaling:**
For given an arbitrary vector $\mathbf{v}$ $\mathbb{R}^n$, a new vector $\mathbf{w}$ can be created by scaling a vector $\mathbf{v}$ by multiplying an arbitrary scaling factor $\alpha \in \mathbb{R}$: 

\begin{equation} \label{eqVscale}
\mathbf{w}=\alpha \mathbf{v}=\left[ \begin{array}{c} \alpha\mathcal{v}_1 \\ \alpha\mathcal{v}_2 \\ \vdots \\ \alpha\mathcal{v}_n \end{array} \right] \forall \alpha \in \mathbb{R} 
\end{equation} 

where $\mathcal{v}_i, i=1,2,...,n$ is the $i$th element of the vector. The scaling is applied on each of the elements of the vector $\mathbf{v}$. Thus, the $i$th element of the resultant new vector $\mathbf{w}$ becomes $\alpha \mathcal{v}_i  (i = 1, 2, ..., n)$, and each of which is a number in $\mathbb{R}$. Even if  $\alpha=0$, the resulted $\mathbf{w}$ is still in $\mathbb{R}$, because $\mathbb{R}$ has a zero vector by definition. Therefore, the new vector $\mathbf{w}$ is always a member in $\mathbb{R}^n$. Because $\alpha \mathcal{v}_i =  \mathcal{v}_i \alpha$ based on the commutative property of standard arithmetic operation of multiplication for numbers, we thus have $\alpha \mathbf{v}=\mathbf{v}\alpha$, meaning scaling to vectors are also commutative. This allows us moving scalar numbers around in a chain of operations involving vectors. 

3) **Addition:**
Given two arbitrary vectors $\mathbf{v}$ and $\mathbf{u}$ all in $\mathbb{R}^n$, a new vector $\mathbf{w}$ is created by adding them up: 

\begin{equation} \label{eqVadd}
\mathbf{w}=\mathbf{v} + \mathbf{u} = \mathbf{u}+\mathbf{v} = \left[ \begin{array}{c} \mathcal{v}_1+\mathcal{u}_1 \\ \mathcal{v}_2+\mathcal{u}_2 \\ \vdots \\ \mathcal{v}_n+\mathcal{u}_n \end{array} \right] \; \;  \forall \mathbf{v} \; \text{and} \; \mathbf{u} \in \mathbb{R}^n    
\end{equation} 

The addition is done to each of the corresponding elements of these two vectors $\mathbf{v}$ and $\mathbf{u}$. Thus the resultant new vector $\mathbf{w}$ is also in $\mathbb{R}^n$, because each of its elements, $(\mathcal{v}_i+\mathcal{u}_i)$, is a number in $\mathbb{R}$. Because of the commutative property of standard arithmetic operation of addition for scalar numbers, vectors in addition operation are also commutative. 

4) **Linear combination:**
Because of,Such a space is call **vector space**, in which 

With the definition of zero vector and the above two basic operations defined in items 2) and 3), it is clear that the space $\mathbb{R}^n$ formed by the vectors is closed under addition and scaling. any linear combination of any two vectors $\mathbf{u} \in \mathbb{R}^n$ and $\mathbf{v} \in \mathbb{R}^n$ with a arbitrary scaling factor $\alpha \in \mathbb{R}$ and a arbitrary scaling factor $\beta \in \mathbb{R}$, $\mathbf{w}=\alpha \mathbf{u}+\beta \mathbf{v} $, is  in $\mathbb{R}^n$.  Therefore, the $\mathbb{R}^{n}$ of vectors is called a linear **vector space**. Its members are all vectors with $n$ elements are all in $\mathbb{R}$.  By default, all vectors dealt with in this book have all the above basic operations defined, and are vector spaces.  

Note that vector space does mean a space formed just by vectors. The space of real numbers $\mathbb{R}$ is a also a vector space, because it is closed under addition and scaling base on the defined standard arithmetic operations. The essential point is that any linear combination of the members in the space shall still stay in the same space. We will see in later chapters also that matrices form also vector spaces. 

Now, any arbitrary vector $\mathbf{v}$ in $\mathbb{R}^n$ can be expressed as multiples of the SBVs:

\begin{equation} \label{eqBVN2}
\boxed{
\begin{split}
\mathbf{v}=\left[ \begin{array}{c} \mathcal{v}_1 \\ \mathcal{v}_2 \\ \vdots \\ \mathcal{v}  \end{array} \right] 
& = \mathcal{v}_1 \left[ \begin{array}{c} 1 \\ 0 \\ \vdots \\ 0 \end{array} \right]+ 
 \mathcal{v}_2 \left[ \begin{array}{c} 0 \\ 1 \\ \vdots \\ 0 \end{array} \right]+ \dots +
 \mathcal{v}_n \left[ \begin{array}{c} 0 \\ 0 \\ \vdots \\ 1 \end{array} \right] \\
& = \mathcal{v}_1\,\mathbf{i}_1 + \mathcal{v}_2\,\mathbf{i}_2 + \dots + \mathcal{v}_n\,\mathbf{i}_n
 \end{split} }
\end{equation} 

where $\mathcal{v}_i (i = 1, 2, ..., n)$ are scalar numbers that are the $n$ independent elements of $\mathbf{v}$, and $\mathbf{v}$ is a linear combination of all the SBVs. 

Notice that we omitted the standard $\times$ symbol for multiplication of real numbers to avoid confusion, because in vector operations, $\times$ has entirely a different meaning, known as the **cross-product** that to be discussed later in section on [Cross product of vectors](#section_cpv). 

#### Other arithmetic operations for vectors

For vectors, we often perform also the following space-staying operations.   

1) **Subtraction:**

Similarly, a new vector $\mathbf{w}$ can be created by subtracting a vector $\mathbf{u}$ from another vector $\mathbf{v}$ all in $\mathbb{R}^n$. 

\begin{equation} \label{eqVsubt}
\mathbf{w}=\mathbf{v} - \mathbf{u} =-\mathbf{u}+\mathbf{v} = (\mathcal{v}_1-\mathcal{u}_1)\mathbf{i}_1 + (\mathcal{v}_2-\mathcal{u}_2)\mathbf{i}_2 + \dots + (\mathcal{v}_n-\mathcal{u}_n)\mathbf{i}_n \; \;  \forall \mathbf{v} \; \text{and} \; \mathbf{u} \in \mathbb{R}^n     
\end{equation} 

The subtraction is only applied to each of the corresponding elements of these two vectors $\mathbf{v}$ and $\mathbf{u}$. Thus the resultant new vector $\mathbf{w}$ is also in $\mathbb{R}^n$, because each of its elements, $(\mathcal{v}_i-\mathcal{u}_i)$, is in $\mathbb{R}$. Vectors in subtraction operation are also commutative, for the same reason. Note that subtraction is just the negated addition, and can be induced by a scaling operation with $\alpha=-1$ and the defined addition. Thus, nothing new. We list it here just for better clarity.  

2) **Element-wise multiplication (also known as the Hadamard product or Schur product):**

Given two vectors $\mathbf{v}$ and $\mathbf{u}$ both in $\mathbb{R}^n$, we perform:

\begin{equation} \label{eqVmul}
\mathbf{w}=\mathbf{v} * \mathbf{u} =\mathbf{u} * \mathbf{v} =(\mathcal{v}_1 \mathcal{u}_1)\mathbf{i}_1 + 
(\mathcal{v}_2 \mathcal{u}_2)\mathbf{i}_2 + \dots + (\mathcal{v}_n \mathcal{u}_n)\mathbf{i}_n \; \;  \forall \mathbf{v} \; \text{and} \; \mathbf{u} \in \mathbb{R}^n     
\end{equation} 

As the name suggests, the multiplication is performed, respectively, to each of the corresponding elements of these two vectors $\mathbf{v}$ and $\mathbf{u}$. Thus the resultant new vector $\mathbf{w}$ is also in $\mathbb{R}^n$, because each of its elements, $(\mathcal{v}_i \mathcal{u}_i)$, is in $\mathbb{R}$. Note that we use symbol $*$ to denote element-wise multiplication, following the Python convention. Symbols $\odot$ and $\circ$ are also used in the literature for this operation. 

3) **Element-wise division:**

Given two vectors $\mathbf{v}$ and $\mathbf{u}$ both in $\mathbb{R}^n$, we perform:

\begin{equation} \label{eqVdiv}
\mathbf{w}=\mathbf{v} / \mathbf{u} =(\mathcal{v}_1 / \mathcal{u}_1)\mathbf{i}_1 + 
(\mathcal{v}_2 / \mathcal{u}_2)\mathbf{i}_2 + \dots + (\mathcal{v}_n / \mathcal{u}_n)\mathbf{i}_n \; \;  \forall \mathbf{v}  \; \text{and} \; \mathbf{u} \in \mathbb{R}^n \; \text{with} \; u_i \ne 0 
\end{equation}  

As the name suggests, the division is performed, respectively, to each of the corresponding elements of these two vectors $\mathbf{v}$ and $\mathbf{u}$. Thus the resultant new vector $\mathbf{w}$ is also in $\mathbb{R}^n$, because each of its elements, $(\mathcal{v}_i / \mathcal{u}_i)$ when $u_i \ne 0$, is in $\mathbb{R}$. We use symbol $/$ to denote element-wise multiplication, flowing Python's convention. Note that element-wise division can often fail, because all the elements in $\mathbf{u}$ have to be nonzero for the defined operation. 

### Inner-product  or dot-product of vectors
Inner-product or dot-product is essential for all linear algebraic operations, and will be used frequently throughout the entire book. Let us introduce it in detail.    

#### Definition: a scalar produced using two vectors

Consider two nonzero vectors $\mathbf{u}$ and $\mathbf{v}$ both in $\mathbb{R}^n$, as shown in Fig.\ref{fig:u_dot_v}.

<img src="./images/u_dot_v.png" width="450"/>
\begin{figure}
  \includegraphics[width=8cm]{}
  \caption{\label{fig:u_dot_v} Two vectors $\mathbf{u}$ and $\mathbf{v}$ and the angle $\theta$ between them}
\end{figure}

To find the angle $\theta$ between them, we defined an operation using these two vectors, called inner-product or dot-product: 

\begin{equation} \label{eqdefDot}
\boxed{ \mathbf{u} \cdot \mathbf{v} =\|\mathbf{u} \|\ \|\mathbf{v} \|\cos\theta  }
\end{equation} 

where $\mathbf{u} \cdot \mathbf{v}$ is computed using multiplications of their corresponding elements. 

\begin{equation} \label{eqDot0}
\boxed{ \begin{split}
\mathbf{u} \cdot \mathbf{v}= (\mathcal{u}_1\mathcal{v}_1 ) + 
(\mathcal{u}_2\mathcal{v}_2 ) + \dots + (\mathcal{u}_n\mathcal{v}_n )  \; \;  &\forall \mathbf{u} \; \text{and} \; \mathbf{v} \in \mathbb{R}^n 
\end{split} }
\end{equation} 

It is the sum of elements of the vector of element-wise multiplication defined above, which is a scalar number in $\mathbb{R}$. 

Because we already know how to compute the norm of a vector using Eq.(\ref{eqvabs}), and it will not be zero for any nonzero vector, the angle $\theta$ can now be computed using Eqs.(\ref{eqdefDot}) and (\ref{eqDot0}), which gives,

\begin{equation} \label{eqctheta}
\boxed{
\cos\theta = \underbrace{\cos(\mathbf{u}, \mathbf{v})}_{\text{direction cosine}} =  \frac{\mathbf{u} \cdot \mathbf{v}}{\|\mathbf{u} \|\ \|\mathbf{v} \|} =\mathbf{\bar{u}} \cdot \mathbf{\bar{v}}  }
\end{equation} 

This gives an important concept of **direction cosine** of two vectors, which is used very often in mechanics problems, and a key variable to construct **tensors**. The angle $\theta$ is found using, 

\begin{equation} \label{eqtheta}
 \theta = \arccos\left(\frac{\mathbf{u} \cdot \mathbf{v}}{\|\mathbf{u} \|\ \|\mathbf{v} \|}\right) = \arccos\left(\mathbf{\bar{u}} \cdot \mathbf{\bar{v}}\right)
\end{equation}

Care is needed when computing $\theta$ using the foregoing equation, because $\theta$ can be positive and negative. 

It is obvious that the inner-product is commutative: $\mathbf{u} \cdot \mathbf{v}=\mathbf{v} \cdot \mathbf{u}$. This is also called symmetry. If the vectors are complex, it is conjugated symmetry: $\mathbf{u} \cdot \mathbf{v}=\overline{\mathbf{v} \cdot \mathbf{u}}$. Widely used notations for inner-product is given as follows. 

\begin{equation} \label{eqDot}
\begin{split}
\mathbf{u} \cdot \mathbf{v}&= \mathbf{v} \cdot \mathbf{u}  \;\; &\text{dot product notation}\\ 
   &=\langle\mathbf{u}, \mathbf{v}\rangle = \langle\mathbf{v}, \mathbf{u}\rangle  \;\; &\text{inner product notation}   \\
   &= \mathbf{v}^\top \mathbf{u} = \mathbf{u}^\top \mathbf{v} \;\; &\text{matrix product notation} \\
   &= (\mathcal{v}_1 \mathcal{u}_1) + 
(\mathcal{v}_2 \mathcal{u}_2) + \dots + (\mathcal{v}_n \mathcal{u}_n)  \; \;  &\text{for computation}  
\end{split}
\end{equation} 

All these notations are widely used in the literature, and will be used in this book interchangeably. 

To make a better sense of the inner-product defined, let us look at a few special cases. 

#### Inner-product associated norm

First, assume $\mathbf{u}$ and $\mathbf{v}$ are identical. In this case, $\cos\theta = 1$, because the in-between angle is zero, and Eqs.(\ref{eqdefDot}) becomes,

\begin{equation} \label{eqvv}
\begin{split}
\mathbf{v} \cdot \mathbf{v} &=\|\mathbf{v} \|\ \|\mathbf{v} \| =\|\mathbf{v} \|^2 \\
\Longrightarrow \|\mathbf{v} \|&=\sqrt{\mathbf{v} \cdot \mathbf{v}} = (\mathcal{v}_1 \mathcal{v}_1) + 
(\mathcal{v}_2 \mathcal{v}_2) + \dots + (\mathcal{v}_n \mathcal{v}_n)
\end{split}
\end{equation} 

which is exactly Eq.(\ref{eqvabs}). The inner-product has an associated norm that is the L2 norm. Thus, the inner-product are often used to compute the L2 norm of a vector. 

Suppose $\mathbf{u}$ and $\mathbf{v}$ are in the same direction but different magnitudes. In this case, we still have $\cos\theta = 1$, because the in-between angle is zero, and Eqs.(\ref{eqdefDot}) gives,

\begin{equation} \label{equv}
\mathbf{u} \cdot \mathbf{v} =\|\mathbf{u} \|\ \|\mathbf{v} \| \;\;\; \text{when} \;\;\; \mathbf{u} \parallel  \mathbf{v} 
\end{equation} 

These two vectors are said parallel. When $\mathbf{u}$ and $\mathbf{v}$ are in the opposite direction, $\cos\theta = -1$, they are also parallel, but 

\begin{equation} \label{equv2}
\mathbf{u} \cdot \mathbf{v} = -\|\mathbf{u} \|\ \|\mathbf{v} \|  
\end{equation} 

#### Orthogonality
Assume $\mathbf{u}$ is perpendicular (orthogonal) to $\mathbf{v}$. In this case, $\cos\theta = 0$, because their in-between angle is either $ \pi$ or $-\pi$, and Eqs.(\ref{eqdefDot}) leads to,

\begin{equation} \label{equllv}
\mathbf{u} \cdot \mathbf{v} =\|\mathbf{u} \|\ \|\mathbf{v} \|\times 0 = 0 
\end{equation} 

Vectors that satisfy this inner product property are mutual orthogonal vectors. The inner-product is often used as tool to check the orthogonality among vectors.  

Equation (\ref{equllv}) gives also a clear definition for **orthogonality** of two vectors. 

It is obvious that the inner product of the set of SBVs satisfy 

\begin{equation} \label{eqSBVDot}
\langle\mathbf{i}_i, \mathbf{i}_j\rangle = (\mathbf{i}_i \cdot  \mathbf{i}_j)=\mathbf{i}_i^\top \mathbf{i}_j=\mathbf{i}_j^\top \mathbf{i}_i=\left\{ 
\begin{array}{ c l } 1 &   \;\; i=j \\ 0 &\;\; i\ne j
\end{array} \right. 
\end{equation} 

#### Orthogonal projection

Finally, let $\mathbf{u}$ be a unit vector. In this case, $\|\mathbf{u} \|= 1$, we have,

\begin{equation} \label{equv1}
\mathbf{u} \cdot \mathbf{v} =\underbrace{\|\mathbf{u} \|}_{1}\|\mathbf{v} \|\ \cos\theta =\|\mathbf{v} \| \cos\theta 
\end{equation} 

This is the orthogonal projection of $\mathbf{v}$ on $\mathbf{\bar{u}}$! It is the red arrow shown in Fig.\ref{fig:u_dot_v}. The magnitude (length) of the projected vector is

\begin{equation} \label{equvOA}
|\overline{OA}| = \mathbf{v} \cdot \mathbf{\bar{u}} = \|\mathbf{v} \|\ \cos\theta  
\end{equation} 

This can only happen when $\angle OAC$ is the right angle. This intuitively shows that $\mathbf{v}$ is orthogonally projected on $\mathbf{u}$. 

Similarly, one can project $\mathbf{u}$ on $\mathbf{v}$, and the projected vector is the green vector shown in Fig.\ref{fig:u_dot_v}. This time, the magnitude of the projected vector is 

\begin{equation} \label{equvOD}
|\overline{OD}| = \mathbf{u} \cdot \mathbf{\bar{v}} = \|\mathbf{u} \|\ \cos\theta  
\end{equation} 

Meaning that the $\angle ODB$ is the right angle, and $\mathbf{u}$ is orthogonally projected on $\mathbf{v}$.  

Inner-products will be frequently used for **orthogonal projections** in this book. 

#### Derivation of the law of cosines
We studied the law of cosines for arbitrary triangles at maybe high schools. Using inner-product, we can easily derive this law. 

Consider two given vectors $\mathbf{u}$ and $\mathbf{v}$ with an angle $\theta$ in between. Let $\mathbf{w}$ be the difference vector: $\mathbf{w}=\mathbf{u}-\mathbf{v}$. Thus, $\mathbf{u}$, $\mathbf{v}$ and $\mathbf{w}$ forms a triangle. If we want to calculate the length of $\mathbf{w}$, we can use its inner-product and the standard arithmetic operations:

\begin{equation} \label{eqcos0}
\begin{aligned}
\mathbf{w} \cdot \mathbf{w} &=(\mathbf{u} -\mathbf{v} )\cdot (\mathbf{u} -\mathbf{v} ) =\mathbf{u} \cdot \mathbf{u} -\mathbf{u} \cdot \mathbf{v} -\mathbf{v} \cdot \mathbf{u} +\mathbf{v} \cdot \mathbf{v} \\&=\|\mathbf{u}\|^{2}-\mathbf{u} \cdot \mathbf{v} -\mathbf{u} \cdot \mathbf{v} +\|\mathbf{v}\|^{2} =\|\mathbf{u}\|^{2}-2\mathbf{u} \cdot \mathbf{v} +\|\mathbf{v}\|^{2} 
\end{aligned} 
\end{equation}  
which gives the law of cosines:
\begin{equation} \label{eqcos}
\|\mathbf{w}\|^{2}=\|\mathbf{u}\|^{2}+\|\mathbf{v}\|^{2}-2\|\mathbf{u}\| \|\mathbf{v}\| \cos\theta  
\end{equation}  

#### Power, energy and work-down
In science and engineering, an inner-product is often the outcome of a pair of complementary vectors. For example, 
+ the **power** of an object is the inner-product of force-vector and velocity-vector that the object possesses.  
+ The **angular momentum** is the inner-product of moment of inertia and the angular velocity. 
+ The **traction** force-vector on the boundary of a solid is the inner-product of the (Cauchy) stress-vector and the outwards normal vector of the boundary.  
+ The **stress energy** in a stressed solid is the inner-product of stress-vector and strain-vectors. 
+ The **work-done** by to an object is the inner-product of force-vector applied on it and displacement-vector caused by the application \cite{liu2013finite}.

### Outer-product of two vectors
#### Outer-product produces a matrix 

An inner-product of two vectors produces a scalar, as we seen in the previous section. In contrast, outer-product of two vectors produces a matrix. It has applications in linear algebra, particularity in matrix operations. 

Let two vectors $\mathbf{u} \in \mathbb{R}^m$ and $\mathbf{v} \in \mathbb{R}^n$ be

\begin{equation} \label{equm}
\begin{split}
\mathbf{u} &=\begin{bmatrix}u_{1}& u_{2}& \cdots & u_{m}\end{bmatrix}^\top \\
\mathbf{v} &=\begin{bmatrix}\mathcal{v}_{1}&\mathcal{v}_{2}&\cdots &\mathcal{v}_{n}\end{bmatrix}^\top
\end{split}
\end{equation}

The outer-product $\mathbf{u}\otimes \mathbf{v}$ becomes a matrix in $\mathbb{R}^{m\times n}$: 

\begin{equation} \label{eqouter}
\boxed{
\begin{split}
\underset{m\times n}{\mathbf{A}}& =\underset{m\times 1}{\mathbf{u}} \; \otimes \;\underset{n\times 1}{\mathbf{v}} \\
&=\begin{bmatrix}u_{1}\\ u_{2}\\ \vdots \\ u_{m}\end{bmatrix}\begin{bmatrix}\mathcal{v}_{1}&\mathcal{v}_{2}&\cdots &\mathcal{v}_{n}\end{bmatrix} =\begin{bmatrix}u_{1}\mathcal{v}_{1}&u_{1}\mathcal{v}_{2}&\dots &u_{1}\mathcal{v}_{n}\\u_{2}\mathcal{v}_{1}&u_{2}\mathcal{v}_{2}&\dots &u_{2}\mathcal{v}_{n}\\\vdots &\vdots &\ddots &\vdots \\u_{m}\mathcal{v}_{1}&u_{m}\mathcal{v}_{2}&\dots &u_{m}\mathcal{v}_{n}\end{bmatrix}
\end{split}  }
\end{equation}

The $ij$th element in $\mathbf{A}$ is $u_i\mathcal{v}_j$. The outer-product is same as the matrix multiplication of $\mathbf{u} \mathbf{v}^\top$. Thus, the shapes of $\mathbf{u}$ and $\mathbf{v}$ are always compatible for this operation. This type of matrix is known as **rank-one matrix**, because it has essentially one independent row or column. 

Now, if we use the SBVs to write vectors $\mathbf{u}$ and $\mathbf{v}$, as in Eq.(\ref{eqBVN2}), we shall have

\begin{equation} \label{equvBs}
\begin{split}
\mathbf{u}&= \mathcal{u}_1\,\mathbf{i}_1 + \mathcal{u}_2\,\mathbf{i}_2 + \dots + \mathcal{u}_m\,\mathbf{i}_m \\
\mathbf{v}&= \mathcal{v}_1\,\mathbf{i}_1 + \mathcal{v}_2\,\mathbf{i}_2 + \dots + \mathcal{v}_n\,\mathbf{i}_n
\end{split}
\end{equation} 

The outer-product $\mathbf{u}\otimes \mathbf{v}$ can be re-written as: 

\begin{equation} \label{eqouteris}
\underset{m\times n}{\mathbf{A}} =\underset{m\times 1}{\mathbf{u}} \, \otimes \,\underset{n\times 1}{\mathbf{v}} =u_{1}\mathcal{v}_{1} \underbrace{\mathbf{i}_1\otimes \mathbf{i}_1}_{\mathbf{I}_{11}} + u_{1}\mathcal{v}_{2}\underbrace{\mathbf{i}_1\otimes \mathbf{i}_2}_{\mathbf{I}_{12}}  + u_{2}\mathcal{v}_{1}\underbrace{\mathbf{i}_2\otimes \mathbf{i}_1}_{\mathbf{I}_{21}}  + ... + u_{m}\mathcal{v}_{n}\underbrace{\mathbf{i}_m\otimes \mathbf{i}_n}_{\mathbf{I}_{mn}}
\end{equation}

where $\mathbf{I}_{ij}$ is called standard basis matrix (SBMs), given by the outer-products of the SBVs: 

\begin{equation} \label{eqiouterj}
\underset{m\times n}{\mathbf{I}_{ij}} =\underset{m\times n}{\mathbf{i}_i\otimes \mathbf{i}_j}; \; i=1,2,...,m; j=1,2,...,n
\end{equation}

Let us first look at an example. 

```python
u = np.arange(1,4)           # numpy 1D array 
v = np.ones(5) * 2           # numpy 1D array 
print(f"Vector u: {u}; Vector v: {v}")

A, B = np.outer(u, v), np.outer(v, u)
print(f"v âŠ— u: \n{B}; its rank = {lg.matrix_rank(B)}")
print(f"u âŠ— v: \n{A}; its rank = {lg.matrix_rank(A)}")

uc = u.reshape(u.shape[0],1) # change u to a column-vector
vc = v.reshape(v.shape[0],1) # change v to a column-vector
print(f"uc@vc: \n{uc@vc.T}") # numpy matrix multiplication
```

#### Cross product of vectors

An inner-product produces a scalar number, and an outer-product produces a matrix. Is there a vector operation produces a vector? Yes. A cross-product does this. It is also called vector-product of two vectors. An excellent [wiki page](https://en.wikipedia.org/wiki/Cross_product) is available on this topic. This section briefly outlines the major points with demonstrations in Python codes. Our notation follows largely that in the wiki page so that readers can make easy connections. 

##### Definition

Given two linearly independent vectors $\mathbf{a}$ and $\mathbf{b}$ both in $\mathbb{R}^3$, the cross-product, is denoted as $\mathbf{a}\times \mathbf{b}$, and often read "a cross b". The outcome is a vector $\mathbf{c}$ also in  $\mathbb{R}^3$ perpendicular (or orthogonal) to both $\mathbf{a}$ and $\mathbf{b}$. Therefore, it is normal to the plane of $\mathbb{R}^2$ spanned by $\mathbf{a}$ and $\mathbf{b}$ in $\mathbb{R}^3$. Together with $\mathbf{a}$ and $\mathbf{b}$ it spans a space of $\mathbb{R}^3 \in \mathbb{R}^3$, as shown in Fig.\ref{fig:Cross_product_vector}

The direction of the cross-product has only two choices: either positive or negative. This is why the cross-product is called a pseudo vector. The positive direction of the cross-product is defined by the so-called right-hand-rule, as shown in right figure in Fig.\ref{fig:Cross_product_vector}.

##### Formulation 

When these two vectors are not linearly independent meaning they are in-line or collinear. Their cross-product is zero vector. This is also true when any of these two vectors are zero-vector.

The magnitude of the cross-product equals the area of the parallelogram formed by $\mathbf{a}$ and $\mathbf{b}$, as shown in Fig.\ref{fig:Cross_product_vector}

The formula for the cross-product definition is given by

\begin{equation} \label{eqcab}
\boxed { 
\mathbf{c} = \mathbf{a} \times \mathbf{b} =\underbrace{\left\|\mathbf{a} \right\|\left\|\mathbf{b} \right\|\sin(\theta )}_{\|\mathbf{c} \|}\ \mathbf{n} }
\end{equation}

where $0^\circ \le \theta \le 180^\circ$ is the angle measured from $\mathbf{a}$ to $\mathbf{b}$, and $\mathbf{n}$ is the unit vector perpendicular to the plane space spanned by $\mathbf{a}$ and $\mathbf{b}$ following the right-hand-rule. As seen in Eq.(\ref{eqcab}), the positivity of the direction of $\mathbf{c}$ is controlled by the sine function that is an odd function. Therefore, if $\mathbf{b}$ crosses $\mathbf{a}$,  $\theta$ will be negative, and then the direction will reverse. In other word, cross-product is anti-commutative. 

##### Volume of a parallelepipedon 

The volume V of a parallelepiped formed by three vectors $\mathbf{a}$, $\mathbf{b}$ and $\mathbf{c}$ as its edges can be computed using a combination of a cross product and a dot product, called **scalar triplet product**, as shown in Fig. \ref{fig:Vol_3vectors}.

The triplett product can be done in any of the three order, and they all gives a scalar: 

\begin{equation} \label{eqVabc}
\mathbf{a} \cdot (\mathbf{b} \times \mathbf{c} )=\mathbf{b} \cdot (\mathbf{c} \times \mathbf{a} )=\mathbf{c} \cdot (\mathbf{a} \times \mathbf{b}) 
\end{equation}
 
Note that the scalar triplet product can be negative, the volume of the parallelepiped should be the absolute value of the scalar:

\begin{equation} \label{eqVabs}
V=|\mathbf{a} \cdot (\mathbf{b} \times \mathbf{c} )|
\end{equation}
 

##### Differences between the dot-product and cross-product

1) The magnitude of the cross-product is computed using the sine of the angle between two vectors. Thus, the cross-product can be a measure of perpendicularity of two vectors. In contrast, the dot-product is a measure of parallelism, as mentioned earlier.

2) Given two vectors with unit length (unit vectors), if they are perpendicular, the cross-product has a magnitude of 1,  and if the two are parallel, the magnitude is zero. While, the dot-product of two perpendicular unit vectors is zero, and is 1 if they are parallel.

3) The magnitude of the cross-product of two unit vectors gives the sine of the angle between two vectors $Î¸$, which is always positive by definition ($Î¸$ is 0Â°âˆ¼180Â°, and measured counterclockwise). In contrast, the dot-product of two unit vectors gives the cosine of $Î¸$, and hence $Î¸$ can be positive or negative.  

##### Derivation of formulas 

Because the cross-product involves direction change, the computation becomes more complicated. For easy in understanding, we first examine the cross-products of the standard basis vectors (SBVs). Let us consider 3D spaces, and denote these three SBVs as $\mathbf{i}$, $\mathbf{j}$, and $\mathbf{k}$. Based on the right-hand-rule, we have

\begin{equation} \label{eqSBVcross}
\begin{array}{l,l,l}
\color{blue}{\mathbf{i}} \times \color{red}{\mathbf{j}} =\color{green}{\mathbf{k}},  \;\;\;\;\;\;  &\color{blue}{\mathbf{j}} \times \color{red}{\mathbf{i}} = -\color{green}{\mathbf{k}} \\ 
\color{red}{\mathbf{j}} \times \color{green}{\mathbf{k}} =\color{blue}{\mathbf{i}},   \;\;\;\;\;\;  &\color{red}{\mathbf{k}} \times \color{green}{\mathbf{j}} = -\color{blue}{\mathbf{i}} \\ 
\color{green}{\mathbf{k}} \times \color{blue}{\mathbf{i}} =\color{red}{\mathbf{j}},  \;\;\;\;\;\;  &\color{green}{\mathbf{i}} \times \color{blue}{\mathbf{k}} = -\color{red}{\mathbf{j}} \end{array}
\end{equation} 

Note the negative sign resulted from the anti-commutativity. In addition, because of the collinearity, we have  

\begin{equation} \label{eqiicross}
\color{blue}{\mathbf{i}} \times \color{blue}{\mathbf{i}} =\color{red}{\mathbf{j}} \times \color{red}{\mathbf{j}} =\color{green}{\mathbf{k}} \times \color{green}{\mathbf{k}} =\mathbf{0} 
\end{equation} 

Using these SBVs, the cross-product of any two vectors $\mathbf{a}$ and $\mathbf{b}$ can be, respectively, written as

\begin{equation} \label{eqabcross}
\begin{alignedat}{3}
\mathbf{a} &=a_{1}\color{blue}{\mathbf{i}} &&+a_{2}\color{red}{\mathbf{j}} &&+a_{3}\color{green}{\mathbf{k}} \\
\mathbf{b} &=b_{1}\color{blue}{\mathbf{i}} &&+b_{2}\color{red}{\mathbf{j}} &&+b_{3}\color{green}{\mathbf{k}} \end{alignedat}
\end{equation} 

where $a_i$, $a_j$, and $a_k$ are the elements of $\mathbf{a}$, and  $b_i$, $b_j$, and $b_k$ are the elements of $\mathbf{b}$, respectively, on its corresponding SBVs. 

Now, following the standard arithmetic operations and the definition of cross-products, we have

\begin{equation} \label{eqabcross2}
\begin{split} 
\mathbf{c}=\mathbf{a} \times \mathbf{b} = &(a_{1}\color{blue}{\mathbf{i}} +a_{2}\color{red}{\mathbf{j}} +a_{3}\color{green}{\mathbf{k}} )\times (b_{1}\color{blue}{\mathbf{i}} +b_{2}\color{red}{\mathbf{j}} +b_{3}\color{green}{\mathbf{k}} )\\
= &a_{1}b_{1}(\color{blue}{\mathbf{i}} \times \color{blue}{\mathbf{i}} )+a_{1}b_{2}(\color{blue}{\mathbf{i}} \times \color{red}{\mathbf{j}} )+a_{1}b_{3}(\color{blue}{\mathbf{i}} \times \color{green}{\mathbf{k}} )+ \\ &a_{2}b_{1}(\color{red}{\mathbf{j}} \times \color{blue}{\mathbf{i}} )+a_{2}b_{2}(\color{red}{\mathbf{j}} \times \color{red}{\mathbf{j}} )+a_{2}b_{3}(\color{red}{\mathbf{j}} \times \color{green}{\mathbf{k}} )+ \\ &a_{3}b_{1}(\color{green}{\mathbf{k}} \times \color{blue}{\mathbf{i}} )+a_{3}b_{2}(\color{green}{\mathbf{k}} \times \color{red}{\mathbf{j}} )+a_{3}b_{3}(\color{green}{\mathbf{k}} \times \color{green}{\mathbf{k}} ) \end{split}
\end{equation} 

Using Eqs.(\ref{eqSBVcross}) and (\ref{eqiicross}), the foregoing equation becomes, 

\begin{equation} \label{eqabcross3}
\begin{aligned}
\mathbf{c}=\mathbf{a} \times \mathbf{b} ={}&\quad \ a_{1}b_{1}\mathbf{0} +a_{1}b_{2}\color{green}{\mathbf{k}} -a_{1}b_{3}\color{red}{\mathbf{j}} \\&-a_{2}b_{1}\color{green}{\mathbf{k}} +a_{2}b_{2}\mathbf{0} +a_{2}b_{3}\color{blue}{\mathbf{i}} \\&+a_{3}b_{1}\color{red}{\mathbf{j}} \ -a_{3}b_{2}\color{blue}{\mathbf{i}} \ +a_{3}b_{3}\mathbf{0} \\
={}&\underbrace{(a_{2}b_{3}-a_{3}b_{2})}_{c_{1}}\color{blue}{\mathbf{i}} +\underbrace{(a_{3}b_{1}-a_{1}b_{3})}_{c_{2}}\color{red}{\mathbf{j}} +\underbrace{(a_{1}b_{2}-a_{2}b_{1})}_{c_{3}}\color{green}{\mathbf{k}} \\
={}& c_{1}\color{blue}{\mathbf{i}} +c_{2}\color{red}{\mathbf{j}} +c_{3}\color{green}{\mathbf{k}} 
\end{aligned}
\end{equation}

We obverse a nice duality formula for computing the elements of vector $\mathbf{c}$, if we re-denote $c_1=c_{23}, c_2=c_{31}, c_3=c_{12} $:

\begin{equation} \label{eqdual}
c_{ij}=a_{i}b_{j}-a_{j}b_{i}
\end{equation}

When $i,j = 1,2,3$, the above duality formula gives exactly 3 distinct values , which we denote as $c_1, c_2, c_3$ to express vector $\mathbf{c}$. 

Alternatively, $\mathbf{c}$ can be expressed in the determinant of a matrix (that will be discussed in detail in the next chapter) as follows. 

\begin{equation} \label{eqcM}
\boxed{
\begin{split}
\underbrace{\mathbf{a\times b}}_{\mathbf{c}} &=\begin{vmatrix}\mathbf{i} &\mathbf{j} &\mathbf{k} \\a_{1}&a_{2}&a_{3}\\b_{1}&b_{2}&b_{3}\\\end{vmatrix} &=
\underbrace{\begin{vmatrix}a_{2}&a_{3}\\b_{2}&b_{3}\end{vmatrix}}_{c_{1}}\;\mathbf{i} \;\underbrace{-\begin{vmatrix}a_{1}&a_{3}\\b_{1}&b_{3}\end{vmatrix}}_{c_{2}}\;\mathbf{j} + \underbrace{\begin{vmatrix}a_{1}&a_{2}\\b_{1}&b_{2}\end{vmatrix}}_{c_{3}}\;\mathbf{k} }
\end{split}
\end{equation} 

Alternatively, we can use the following matrix-vector formulation in computation. 

\begin{equation} \label{eqcMv}
\boxed{
\begin{split}
\underbrace{\mathbf{a\times b}}_{\mathbf{c}} &=\begin{bmatrix}0 &-a_3 &a_2 \\a_3 &0 &-a_1 \\-a_2 &a_1 &0 \end{bmatrix} \begin{bmatrix}b_1 \\b_2 \\b_3 \end{bmatrix}
\end{split} }
\end{equation} 

The cross-product has applications in various areas, including computational geometry, evaluation of angular momentum and torque, and calculation of Lorentz forces, calculation of moments by forces, just to name a few. It is often use to determine the direction normal to any pair of vectors, which are needed in many operations, such as the coordinate transformation in the finite element methods \cite{liu2013finite} for stress analysis in structures. Interest readers may refer to the [wiki page](https://en.wikipedia.org/wiki/Cross_product) and the references therein, for more detailed formulations, properties, history and applications. 

Note we did not discuss about cross-products in higher dimensions. This is because it would get into the concept of **exterior algebra** or **tensor algebra**. From Eq.(\ref{eqdual}) we see our formula works just right for 3D When the dimension gets higher, 4D, for example, $i$ and $j$ shall vary from $1\sim 4$. In such cases, the duality formula would produce a lot more terms than 4, meaning it be come a tensor for higher dimensions. Discussion on exterior or tensor algebra is out of the scope of this book. We thus stop here. 

#### Computation
We now write some simple code to examine cross-products. 

```python
a = [2., 1, 3]
b = [5., 3, 2]

c1 =   a[1]*b[2]-a[2]*b[1]
c2 = -(a[0]*b[2]-a[2]*b[0])
c3 =   a[0]*b[1]-a[1]*b[0]
print(f"elements of vector c = a x b: c1={c1}, c2={c2}, c3={c3}")

c_ab = np.cross(a, b)
c_ba = np.cross(b, a)
print(f"c = a x b: {c_ab}\nc = b x a: {c_ba}")
print(f"c dot a: {c_ab.dot(a):.4f}, c dot b: {c_ab.dot(b):.4f}")
```

### Coordinate transformation

In solving a mechanics problem, we deal with objects including numbers, vectors, tensors. Such an object is usually represented in reference to an orthogonal coordinate system, such a Cartesian, or a polar, or cylindrical, or spherical. We often perform coordinate transformation, which can result in changes in the representation. Coordinate transformation refers often as a **rotation**, or a **change** from one type of (often orthogonal) coordinate system to another type of (often orthogonal) coordinate system, such as change from the Cartesian to polar coordinate system. Coordinate transformation of any object occurs usually at a point. 

Therefore, when we say coordinate transformation, it does NOT include translation. The change of an object due to the location change is dealt with using the concept of **function**: the change in an object when the location in the structure member changes.  

A scalar number is a tensor of 0th order. The rule of coordinate transformation for a scalar is the simplest: 

> Scalar does not change when the coordinate transformation. It doest not require a index with respect to the coordinates. 

This is simply because a scalar can be transformed by above definition: When a scalar is represented in a coordinate, we have only a single one-dimensional (1D) axis, say $x$, in $\mathbb{R}$, as shown in Fig.\ref{image:Real_number_line2}. The scalar has a single value right on the axis of a given 1D coordinate. In a 1D space, the axis can only be translated along its axis direction, and cannot have rotation (otherwise, the space becomes 2D or 3D). 

The single value can still be viewed as obtained by an orthogonal projection, but the angle of the projection is zero. Any rotation (with zero angle) of the coordinate has no effect at all to any scalar. In other words, the transformation matrix $\mathbf{T}$ (discussed in detail later) is simply 1. 

### Vectors in mechanics

In mechanics, vectors are often encountered important objects. Typical ones are force vector and displacement vector. 

#### Notation for coordinate systems

A vector has two basic attributes: length and direction. To quantify a vector in mechanics, we use an orthogonal coordinate system. A vector can then be equivalently described in a number of scalars in an array. 

Consider a three-dimensional (3D) space $\mathbb{R}^{3}$, in which we define a Cartesian coordinate system $\mathbf{x}$ that is orthogonal: $x$-, $y$-, and $z$-axis are the three mutually orthogonal axes of $\mathbf{x}$. Each of these axes has a unit vector $\mathbf{i},\mathbf{j}$, and $\mathbf{k}$ with unit length, respectively. Note that we use **boldface** to denote a vector, or matrix that is a collection of vectors. 

In indicial rotation, these axes are denoted as $x_1$, $x_2$, and $x_3$, or $x_i \; (i=1,2,3)$, and its unit vectors are denoted as $\mathbf{i}_1$, $\mathbf{i}_2$, and $\mathbf{i}_3$ or $\mathbf{i}_i \; (i=1,2,3)$. The unit vectors are also denoted as $\mathbf{e}_1$, $\mathbf{e}_2$, and $\mathbf{e}_3$ or $\mathbf{e}_i \; (i=1,2,3)$. Indicial notation is more concise and often more convenient in derivations. Hence we often switch to it in this book, when necessary. Readers will be getting used to it, after doing this for some times. 

#### Description of vectors

Consider generally a vector in mechanics denoted as $\mathbf{v}$ in $\mathbb{R}^{3}$, as shown in Fig.\ref{image:Vector3D}. The vector can be a force vector, displacement vector, velocity vector, or  acceleration vector. 

<img src="./images/Vector3D.png" alt=" " width="350"/><br>
\begin{figure}
  \includegraphics[width=5cm]{}
  \caption{\label{image:Vector3D} Any vector $\mathbf{v}$ in $\mathbb{R}^{3}$ can be decomposed into three components: $v_1\mathbf{i}_1$, $v_2\mathbf{i}_2$ and $v_3\mathbf{i}_3$ following also the orthogonal projection rule.}
\end{figure}

We present the $\mathbf{v}$ uniquely under the Cartesian coordinate system. We first decompose it to two vectors: a vertical component $v_3\mathbf{i}_3$ and an in-plane vector $\mathbf{v}_s$ following the orthogonal projection rule. The $\mathbf{v}_s$ is then further decomposed into two vectors: $v_1\mathbf{i}_1$ and $v_2\mathbf{i}_2$ following also the orthogonal projection rule. This process resulted to three components of $\mathbf{v}$: $v_1, v_2$ and $v_3$, respectively, on $x_1$-, $x_2$-, and $x_3$-axis. The vector can now be written uniquely as

\begin{equation} \label{eqv3d}
\mathbf{v} = \begin{bmatrix}v_1\\v_2\\v_3\end{bmatrix}=\begin{bmatrix}v_1 & v_2 & v_3\end{bmatrix}^\top
\end{equation} 

where $^\top$ stands for transpose. A change to any of these three components results in another different vector. 

#### Equivalence in representation, first order tensor

From Fig.\ref{image:Vector3D}, we found the following: 

 1. It is clear that this representation of a vector is **unique** under a fixed coordinate, meaning that any change, in length or in direction, in $\mathbf{v}$ will results in unique changes in $v_1, v_2$ and $v_3$, and vice versa, so long the **orthogonal projection** rule is followed. This mutual unique representation ensures the **equivalent** representation, meaning $\mathbf{v}$ and $\begin{bmatrix}v_1 & v_2 & v_3\end{bmatrix}^\top$ are identical. Such an equivalence is ensured by the orthogonal projection rule in an orthogonal coordinate system. 
\vspace{1pt}
 2. When the coordinate $\mathbf{x}$ is rotated (transformed) to $\mathbf{x}'$, vector $\mathbf{v}$ is now represented as,
 
\begin{equation} \label{eqv3dp}
\mathbf{v} = \begin{bmatrix}v_1'\\v_2'\\v_3'\end{bmatrix} 
\end{equation} 

<span style="display: inline-block; margin-left: 29px;">The values of $v_1'$, $v_2'$, and $v_3'$ are different from $v_1$, $v_2$, and $v_3$. However, these two basic attributes: length and direction of the vector $\mathbf{v}$ remain unchanged. The vector under our transformation rule (orthogonal projection on orthogonal coordinates) is a tensor. Because we need only one index to uniquely locate its components, it is called first order tensor. Because it is in 3D, it is a three dimensional tensor.</span> 

#### Tensors in general 

Tensor is a term we often hear about, but it confuses many, including the author in the past for long time. If you do not have a clear idea about it, you are not alone. To put this concept in a short sentence, it states as follows. 

*A tensor is an object defined in a system, where the change of the values describing the object obeys a certain rule, when a transformation occurs in the system*. 

This is a very general definition. It implies that there are different types of tensor in different fields or systems with different transformation rules. 

A more straightforward question may be: it a square matrix a tensor? The answer is no or yes. If the matrix contains merely stacks of numbers, it is not. If there is a certain rule attach to it, so that the numbers in it change following a system-wide rule when a transformation occurs, it becomes a tensor. A 2nd order tensor can be in a square matrix form, but a square matrix is not automatically a tensor. 

#### Definition of tensor in mechanics of materials

In in mechanics of materials (or solid mechanics, continuum mechanics or elasticity theory) system, all the objects we deal with are tensors of different order. Scalars are of 0th order, forces and displacements are vectors and are 1st order, stress and strain are 2nd, and elasticity constants 4th. These objects are defined in an orthogonal coordinate that can transform (rotate or type change). The rule for operating on these objects and to obtain the values to describe them is the **orthogonal projection**. 

The order of the tensor is identified by the number of the indexes the object carry often as a subscript, in indicial notation. 0th order tensors (scalar numbers) carry no index, 1st order (vector) carries 1, 2nd order (stress) carries 2, and $n$th $n$. These indexes refer to the coordinates. Thus, when the coordinate transforms, the components of the object changes accordingly. 

In the array notation (as the one used in Python), the order of the tensor is identified by the number of the axes the object, which is the shape of the array in numpy. 0th order tensors (scalars) is a scalar variable (float or integer, it has no shape), 1st order has a shape of either ($n$, or ($n$,1), or (1,$n$), where n is the dimension of the array, 2nd ($n$,$n$), and $n$th ($n,n,n,...,n$) with $n$ $n$s. 

Matrix notation (with multi-dimensional matrix defined) is essentially the same as the array notation, although there are some subtleties in computer representation and operation rules, as discussed in Chapters 1 and 2 in \cite{liu2022ML}.  

The order of the tensor is also the times of the operations needed to get the coordinate-transformed values, and hence the mathematical formula of each of them has a different form, but these mathematical rules are all obtained following the **orthogonal projection** rule. 

The first high-level question that may need to be answered is why we would need such a concept of tensor. This is because with it, the mechanics family can have responsible objects, and all the objects follow the same rule to behave, and thus all these objects can be evaluated consistently and correctly, when a transformation takes place. This enable us to do whatever necessary transformations needed to get what we want in proper ways.  

The second high-level question that why we transform the coordinate. This is because we often want to find the values of the tensor objects at one particular direction, or find the critical (maximum, minimum, etc.) values and their corresponding directions. 

In summary, in mechanics of materials, we deal with tensors: scalars, vectors (force vector, displacement vector, etc.), stresses, strains, elasticity, just to name most important ones. All these variable objects need to satisfy the **principle of equilibrium**. They are often described in reference to a coordinate that is usually **orthogonal**, so that we can quantify and operate on them in a meaningful manner in numerical values or symbolic expressions. When the coordinate transforms, all the values of these variables may change. Therefore, there must be a rule that governs such changes, so that the entire system behaves in a manner for consistent and correct quantification.  The objects may have different form of mathematical representations, including scalar (single number), vector (a stack of numbers), matrix (two-dimensional stack of numbers), an elasticity (four-dimensional stack of numbers), but they all are tensors, because they all follow the same rule governing the transformation. It is the **orthogonal projection rule** in mathematical operations, and the **principle of equilibrium** in mechanics law. 

### Transformation rules

#### Definition of coordinate transformation

As mentioned earlier, when we say coordinate transformation, we always refer to **coordinate rotation** or **type change** of the coordinate system. This is because when we study the object changes, we are interested the change at the same location in solids. Therefore, the transformation can only be a rotation or type-change, because the translational position is fixed. We also require that when the coordinate rotates, the entire coordinate system **rotates rigidly** together. 

For changes of objects with the location change, we shall use the usual **function** concept in mathematics, which is a separated issue form coordinate rotation. In this case, the object becomes a variable over space, and we require all the variables being sufficiently smooth functions of the coordinates for differentiations. Functions of variables is another important aspects to be studied in  independently of the coordinate transformation. This is because many governing law of mechanics are the resulted from the function behavior of the variables. 

To put it more clearly: the *tensor concept deals with the coordinate transformation*, and the *function concept deal with the variation of the object along with the coordinate*. 

#### Representation of a tensor in a coordinate 

In mechanics of materials, we use the following assumptions for representation of a tensor in a coordinate system. 

1) **Orthogonal coordinate:** We use orthogonal coordinate systems, meaning all the axes in the coordinate system are, at least locally, mutually perpendicular. 

2) **Right-hand-rule:** We require also these axes follows the conventional right-hand-rule: x-cross-y gives z. This is for consistence and convenience purposes. 

In this book, we will use the Cartesian coordinate as the default, which is also in many Python modules. Other coordinate systems many be used in similar manner. Conversion between coordinate systems can be done based the so-called curvilinear coordinate that are at least locally orthogonal. 

#### Essential rule of coordinate rotation

**Orthogonal projection rule:** When a tensor is represented in coordinates, we follow the orthogonal projection rule to compute the component values of the tensor, in addition to the equilibrium las of mechanics. This means that the component value of the tensor on an axis is the value obtain always by orthogonally projecting the tensor onto the axis. When the coordinate transforms, the new component values of the tensor is obtained also follow the same orthogonal projection rule, in the entire operation process. 

### Transformation Matrix

#### A simple 2D case

As mentioned, tensor concept is about the coordinate transformation. When the coordinate rotates to a new one, the representation of the tensor in the new coordinate changes. To quantify such changes, we must figure out the relationship between these two, the original and the rotated, coordinate systems, and in precise mathematical formulas.

For easy visual, consider first a simple 2D case. Let the original $\mathbf{x}$ with two axes denoted in conventional manner as $x$ and $y$. The rotated new coordinate $\mathbf{x}'$  with axes $x'$, $y'$, as drawn in 
Figure \ref{image:xxp2Dv}. The prime symbol stands for objects in the new coordinate.

<img src="./images/xxp2Dv.png" alt=" " width="500"/><br>
\begin{figure}
  \includegraphics[width=5cm]{}
  \caption{\label{image:xxp2Dv} A point with coordinate values of $x$ and $y$ in coordinate $\mathbf{x}$, and with $x'$, $y'$ in coordinate $\mathbf{x}'$ that is obtained by rotating coordinate $\mathbf{x}$ by $\theta$. All the operations here is based on the orthogonal projection rule.}
\end{figure}

From Fig. \ref{image:xxp2D}, we find the follow relationship via simple geometric examination following the orthogonal projection rule: 

\begin{equation} \label{eqxxp2D}
x'= x\cos\theta+y \sin\theta 
=x\underbrace{\cos(x',x)}_{\cos\theta}+y\underbrace{\cos(x',y)}_{\cos(90Â°-\theta)}  \\
y'= -x\sin\theta+y \cos\theta 
=x  \underbrace{\cos(y',x)}_{\cos(90Â°+\theta)} +y\underbrace{\cos(y',y)}_{\cos\theta} 
\end{equation}

Equations (\ref{eqxxp2D}) can be written in a matrix form of,

\begin{equation} \label{eqxxpMa}
\boxed{
\underbrace{\begin{bmatrix} x'\\ y' \end{bmatrix}}_{\mathbf{x}'}
=\underbrace{\begin{bmatrix}
\cos(x',x) & \cos(x',y) \\
\cos(y',x) & \cos(y',y) \end{bmatrix}}_{\mathbf{T}}
\underbrace{\begin{bmatrix} x\\ y \end{bmatrix}}_{\mathbf{x}} }
\end{equation}

These cosines are called **direction cosines**. Equation (\ref{eqxxpMa}) can be written simply as: 

\begin{equation} \label{eqxxpM} 
\mathbf{x'}=\mathbf{T} \mathbf{x} 
\end{equation}

where $\mathbf{x}$ is a vector of the original coordinate, which uniquely represent the coordinate $\mathbf{x}$, and $\mathbf{x}'$ is a vector of the rotated coordinate. We use **boldface** to denote a vector or matrix. Matrix $\mathbf{T}$ is called transformation matrix. It is clear that matrix $\mathbf{T}$ contains only the angular relationship of these axes of the two original and new coordinates. For 2D cases, it can be written as follows, which is seen more often in the literature. 

\begin{equation} \label{eqxxpMd}
\boxed{
\mathbf{T}= \begin{bmatrix}
\cos\theta & \sin\theta \\
-\sin\theta & \cos\theta \end{bmatrix} }
\end{equation}

It is important to note that although the values of the components in $\mathbf{x}$ and $\mathbf{x}'$ are different, they both represent the same point in 2D space (a position vector), as shown in Fig.\ref{image:xxp2D}.

We emphases that all these expressions obtained following the **orthogonal projection rule** and using orthogonal coordinates, as shown in Fig.\ref{image:xxp2D}. Because of this, $\mathbf{T}$ has the following property:

\begin{equation} \label{eqTTT2D}
\begin{split}
\mathbf{T}\mathbf{T}^\top &=\begin{bmatrix}
\cos\theta & \sin\theta \\
-\sin\theta & \cos\theta \end{bmatrix} 
 \begin{bmatrix} 
\cos\theta & -\sin\theta \\
\sin\theta & \cos\theta \end{bmatrix} \\ &=
 \begin{bmatrix} 
\cos^2\theta + \sin^2\theta & -\cos\theta \sin\theta + \sin\theta \cos\theta   \\
-\sin\theta \cos\theta + \cos\theta \sin\theta &\sin^2\theta + \cos^2\theta  \end{bmatrix} 
 \\ &= \begin{bmatrix} 
1 & 0 \\
0 & 1 \end{bmatrix} = \mathbf{I} 
\end{split}
\end{equation} 

This property of $\mathbf{T}$ is called **unitary** or **orthogonal**. It is invertible, and its inverse transformation can be simply done using the transpose of $\mathbf{T}$:

\begin{equation} \label{eqTTT-1}
\mathbf{T}^{-1}=\mathbf{T}^\top
\end{equation} 

where subscript stands for transpose. It is also easy to confirm that $\mathbf{T}^\top \mathbf{T} =\mathbf{I}$. 

Because of this property of $\mathbf{T}$, the inverse transformation of Eq.(\ref{eqxxpM}) becomes, 

\begin{equation} \label{eqxxpMi} 
\mathbf{x}=\mathbf{T}^\top \mathbf{x'} 
\end{equation}

#### General definition of transformation matrix

With the explicit examination on 2D cases for rotation matrix for coordination transformation, we are now ready to have a general definition. For convenience in description and formulation, we use indicial notion. 

Consider a Cartesian coordinate $\mathbf{x}$ with axes $x_1$, $x_2$, and $x_3$. It is rotates to $\mathbf{x}'$ with axes $x'_1$, $x'_2$, and $x'_3$, the values of the components of an object (obtained also via the orthogonal projection rule) will change. We shall discuss the general mathematic rule of transformation to compute the components of the object, in the new coordinate $\mathbf{x}'$.  To this end, we define rotation matrix for coordinate transformation that bridges the original and new coordinates. 

Naturally, the rotation matrix for coordinate transformation, $\mathbf{T}$, shall contain the angular relationship of these axes of the two coordinates. This must be true regardless the dimensionality. Based on the orthogonal projection rule, $\mathbf{T}$ for 3D cases should be formed by the direction cosines of the angle between the axes of the original coordinate $x_1$, $x_2$, and $x_3$, and the rotated coordinate $x'_1$, $x'_2$, and $x'_3$. 

\begin{equation} \label{eqTMdef}
\boxed{
\mathbf{T}=\begin{bmatrix}
a_{11}=\cos{(x'_1,x_1)} & a_{12}=\cos{(x'_1,x_2)} & a_{13}=\cos{(x'_1,x_3)}\\ a_{21}=\cos{(x'_2,x_1)} & a_{22}=\cos{(x'_2,x_2)} & a_{23}=\cos{(x'_2,x_3)}\\ a_{31}=\cos{(x'_3,x_1)} & a_{32}=\cos{(x'_3,x_2)} & a_{33}=\cos{(x'_3,x_3)} \end{bmatrix} }
\end{equation}

It contains all the direction cosines of these two coordinate systems. This definition is a natural extension from the 2D cases discussed in the previous section. Here, we introduced indicial notation $a_{ij}, i,j=1,2,3$ to denote the entries of $\mathbf{T}$. The right-top $2\times 2$ sub-matrix is the same $\mathbf{T}$ given in Eq.(\ref{eqxxpMa}), which can be re-written as,

\begin{equation} \label{eqa2by2}
\mathbf{T}= \begin{bmatrix}
a_{11}=\cos\theta & a_{12}=\sin\theta \\
a_{21}=-\sin\theta & a_{22}=\cos\theta \end{bmatrix} 
\end{equation}

\end{equation}

We write the following code for future use to create this type of transformation matrices in Sympy for symbolic computations. 

```python
def transf_YPRs(ðœƒ, about = 'z'):
    '''Create a transformation matrix for coordinate transformation\
    Input ðœƒ: rotation angle, in Sympy \
          about: the axis of the rotation is about \
    Return: Sympy matrix of transformation matrix of shape (3,3)
    '''
    c, s = sp.cos(ðœƒ), sp.sin(ðœƒ)
    
    if about == 'z':
        # Yaw: rotates about z by ðœƒ 
        T = sp.Matrix([[ c, s, 0],
                       [-s, c, 0],
                       [ 0, 0, 1]]) 
    elif about == 'y':
        # Pitch: rotates about y by ðœƒ 
        T = sp.Matrix([[ c, 0,-s],
                       [ 0, 1, 0],
                       [ s, 0, c]])  
    elif about == 'x':
        # roll: rotates about x by ðœƒ 
        T = sp.Matrix([[ 1, 0, 0],
                       [ 0, c, s],
                       [ 0,-s, c]])
    else: # randomly generated unitary matrix as transformation matrix:
        print("Rotation axis must be given: z, y, or x")
        
    return T, about
```

#### Property of rotation matrix, a unitary matrix

Because $a_{ij}$ are obtained by the orthogonal projection rule, $\mathbf{T}$ is a tensor. It has a matrix form, and is in general asymmetric. Because the axes of the coordinate systems are kept mutually orthogonal, we shall have the following orthogonal condition: 

\begin{equation} \label{eqOrtha}
a_{ij}a_{kj} =\delta_{ik}
\end{equation} 

where $\delta_{ij}$ is the Kronecker delta defined as 

\begin{equation} \label{eqDelta}
\delta_{ij}= 
\begin{cases}
    1,& \text{if } i = j\\
    0,& \text{if } i\ne j
\end{cases}
\end{equation} 

In Eq.(\ref{eqOrth}), we see that index $j$ is repeated which implies a summation: 

\begin{equation} \label{eqsum}
a_{ij}a_{kj} = a_{i1}a_{k1}+a_{i2}a_{k2}+a_{i3}a_{k3}
\end{equation} 

It is known as the Einstein notation of summation: any repeated index in a term in an expression implying a summation over the range of the repeated index. As the result of such as summation, the index will disappear, and is said *contracted*. Therefore, any letter can be used for a repeated index, so long it is differ from other indexes in the term and the range is the same. For example, $a_{ij}a_{kj}$ is the same as $a_{im}a_{km}$, and they all result in a same 2nd order tensor with indexes $ik$. For these reasons, a repeated index is called a **dummy index**. 

Indexes $i$ and $k$ in Eq.(\ref{eqOrth}) are all single in a term in the equation, and cannot be contracted. They are called **free-indexes**. Clearly, one can use any other symbol for these free-indexes, so long it does not interfere with other indexes and the range is the same. For example, $a_{ij}a_{kj} =\delta_{ik}$ is the same as $a_{rj}a_{sj} =\delta_{rs}$. 

Because the coordinate rotates rigidly, and coordinate  $\mathbf{x}'$ is also orthogonal, we have

\begin{equation} \label{eqOrthI}
a_{ij}a_{ik} =\delta_{jk}
\end{equation} 

Due to Eq.(\ref{eqOrth}), the transformation matrix $\mathbf{T}$ hall satisfy the following orthogonal condition in matrix form:

\begin{equation} \label{eqTTT}
\mathbf{T}\mathbf{T}^\top=\mathbf{I} \;\;\; \text{ or } \;\;\; \mathbf{T}^\top \mathbf{T} =\mathbf{I}
\end{equation} 

where $\mathbf{I}$ is a $3\times 3$ identify matrix for $\mathbb{R}^3$. Here we follow the standard matrix multiplication rule. Eq.(\ref{eqTTT}) means that $\mathbf{T}$ is an **unitary matrix**. We have see this property explicitly for 2D case in the previous section. The unitary property $\mathbf{T}$ implies that 

+ The determinant of $\mathbf{T}$ (with the right-hand rule enforced) is 1. 

+ When it operates on a vector, it will not change the vector length. 

+ When it operates on a square matrix will not change the determinate of the matrix. Therefore, it explains also that why the determinate of a matrix is invariants under coordinate transformation. This can be proven easily: because $\det(\mathbf{TM}) = \det(\mathbf{T}) \det(\mathbf{M})$ and $\mathbf{T}$ has a determinant of 1. The same holds if $\mathbf{T}$ operates from the right. 

When $\mathbf{T}$ acts on a tensor in $\mathbf{x}$, the new components of the same tensor in the rotated coordinate $\mathbf{x}'$ can be obtained. The length of $\mathbf{x}$ and $\mathbf{x}'$ is the same. The transformation follows the orthogonal projection rule, because of the definition of $\mathbf{T}$. 

It may be worth to note that this general definition of tensors for mechanics of materials applies, in theory, to arbitrary dimension. In $n$-dimensions, the range of $i$, and $j$ become $n$, and the dimension for $\mathbf{T}$ will be $n\times n$. The property of $\mathbf{T}$ will still hold, as long as the transformation rules defined in Section 2.2 is followed. 

Let us write some Python codes for creating rotation matrices in 3D, and then examine its properties. 

In indicial notation, Eq.(\ref{eqxxpM}) can be written as 

\begin{equation} \label{eqxxpM2} 
x'_i=a_{ij} x_j
\end{equation}

And Eq.(\ref{eqxxpMi}) can be written as 

\begin{equation} \label{eqxxpMi2} 
x_i=a_{ji} x'_j
\end{equation}

The transpose of a matrix is equivalent to **swapping** the indexes of a second order tensor. 

#### Python code for rotation matrices, in Numpy array

We write a code function that can be used to generate some typical rotation matrices or a random rotation matrix. With such a matrix, one can rotate the coordinate, resulting in description change of objects, such as vector, defined in the coordinate. This function will be used later for multiple times to perform coordinate transformation. 

```python
def transferM(theta, about = 'z'):
    '''Create a transformation matrix for coordinate transformation\
    Input theta: rotation angle in degree \
          about: the axis of the rotation is about \
    Return: numpy array of transformation matrix of shape (3,3)'''
    from scipy.stats import ortho_group
    
    n = 3        # 3-dimensonal problem
    c, s = np.cos(np.deg2rad(theta)), np.sin(np.deg2rad(theta))
    
    if about == 'z':
        # rotates about z by theta 
        T = np.array([[ c, s, 0.],
                      [-s, c, 0.],
                      [0.,0., 1.]]) 
    elif about == 'y':
        # rotates about y by theta 
        T = np.array([[ c, 0.,-s],
                      [ 0, 1.,0.],
                      [ s, 0., c]])  
    elif about == 'x':
        # rotates about x by theta 
        T = np.array([[ 1.,0., 0.],
                      [ 0., c, s],
                      [ 0.,-s, c]])
    else: # randomly generated unitary matrix as transformation matrix:
        T = ortho_group.rvs(dim=n)          # Generate a random matrix
        T[2,:] = np.cross(T[0,:], T[1,:])   # Enforce the righ-hand rule
        
    return T, about
```

Let us examine in detail $\mathbf{T}$ matrices created using transferM(), and check whether the orthogonality and the right-hand rule are observed. 

```python
from scipy.stats import ortho_group

np.random.seed(8)       # set a seed, random number can be re-produced
about, theta = 'random', 30.                 # rotation angle and axis
Tr, _ = transferM(theta, about = about)
print(f'Trransformation tensor {theta:3.2f}Â°, w.r.t. {about}:\n{Tr}')
print(f"Determinant of Tr= {lg.det(Tr):.6f}\n")
print(f"Is Tr orthogonal? \n{Tr.dot(Tr.T)}")
print(f"Is Tr.T orthogonal? \n{(Tr.T).dot(Tr)}\n")

print(f"Check righ-hand rule:")
print(Tr[0,:],np.cross(Tr[1,:], Tr[2,:])) #Check righ-hand rule: 1=2X3
print(Tr[1,:],np.cross(Tr[2,:], Tr[0,:])) #Check righ-hand rule: 2=3X1
print(Tr[2,:],np.cross(Tr[0,:], Tr[1,:])) #Check righ-hand rule: 3=1X2
```

A coordinate transformation matrix is unitary, meaning that its inverse is its transpose. Let us confirm.

When a two-side application of $\mathbf{T}$ on a matrix, the eigenvalues of the resulted new matrix are unchanged, even if the matrix is a random asymmetric matrix. This is because the coordinate transformation is a linear operator that changes the basis for description for a vector or a matrix, and hence will not affect its physic features such as the length of a vector and the determinant (kind of size) of the matrix. The eigenvectors, however, will be altered, because of the change of basis.

### Rotation matrix

#### Derivation for 2D cases

Readers may familiar with the **rotation matrix**, often denoted as $\mathbf{R}$. It is used to rotate a vector in a coordinate system counterclockwise, which results in new components of the vector in the same coordinate system. Assume the vector, before rotation, is denoted as $\mathbf{v} = (x, y)$ in the coordinate system. After it is rotated, it becomes $\mathbf{v}' = (x', y')$, as shown in Fig. \ref{image:vvp2D}. The prime symbol stands for the new coordinate values.

<img src="./images/vvp2D.png" alt=" " width="300"/><br>
\begin{figure}
  \includegraphics[width=5cm]{}
  \caption{\label{image:vvp2D} A vector with coordinate values of $x$ and $y$ in a coordinate system. It is rotated counterclockwise by $\theta$, which results in new vector with coordinate values $x'$, $y'$ in the same coordinate system. All the operations here is based on the orthogonal projection rule.}
\end{figure}. 

Simple analysis based on Fig.\ref{image:vvp2D}, we have

\begin{equation} \label{eqxyp}
x'=r'\cos(\alpha + \theta)=\underbrace{r'\cos\alpha}_{x} \cos\theta - \underbrace{r'\sin\alpha}_{y}\sin\theta \\ =
x\cos\theta-y\sin\theta \\[9pt]
y'=r'\sin(\alpha + \theta)=\underbrace{r'\sin\alpha}_{y} \cos\theta + \underbrace{r'\cos\alpha}_{x}\sin\theta \\ =
x\sin\theta+y\cos\theta 
\end{equation} 

where $\alpha$ is the angle of the vector with respect to the $x$-axis, and $\theta$ is the rotation angle to the vector. The foregoing equation can be written in matrix form:

\begin{equation} \label{eqRMv}
\mathbf{v}'=\underbrace{\begin{bmatrix} x'\\ y' \end{bmatrix}}_{\mathbf{v}'}
=\underbrace{\begin{bmatrix}\cos \theta &-\sin \theta \\\sin \theta &\cos \theta \end{bmatrix}}_{\mathbf{R}} \underbrace{\begin{bmatrix}x\\y\end{bmatrix}}_{\mathbf{v}}=\mathbf{R}\mathbf{v}
\end{equation} 

where $\mathbf{R}$ is the rotation matrix given as 

\begin{equation} \label{eqRM}
\mathbf{R}=\begin{bmatrix} 
\cos\theta &-\sin\theta \\
\sin\theta &\cos\theta  \end{bmatrix}
\end{equation} 

#### Relationship between transformation and rotation matrix

Comparing Eqs.\ref{eqxxpMd} and \ref{eqRM}, it is clear that 

\begin{equation} \label{eqRTT}
\mathbf{R}=\mathbf{T}^\top
\end{equation} 

This is simply because $\mathbf{T}$ rotates the coordinate with the object (vector) stay still, but $\mathbf{R}$ rotates the object (vector) within the same coordinate. Thus, the effects of their actions are mutual reverse. Thus, Eq.(\ref{eqRTT}) is valid for any higher dimensions.

### Function, an object varying over space

Notice that when we talk about coordinate transformations we stand at a point in solid and examine the change of the components of an object In mechanics of materials. 

We now discuss about functions that deal with the variation of a object in space. In a problem of mechanics materials, a variable, such as a component of displacement or stress or strain tensor, can change from location to location. This means that it can vary with the coordinates in a coordinate system. The variation of a variable with respect to coordinates is called **function** of coordinates. It can be defined as an operator $f()$ that takes inputs of coordinates in an n-dimensional space $\mathbb{R}^{n}$ and produces a new number in $\mathbb{R}$ corresponding to the coordinates. It is a mapping from $\mathbb{R}^{n}\to \mathbb{R}$. In the 3D Cartesian coordinate system, for example, the function is often expressed as $f(\mathbf{x})$ or $f(x, y, z)$, where $x, y$ and $z$ are the coordinates as the independent variables of the function $f$. 

In higher dimensions, there are multiple components. Each of which can be a function of coordinates. For a displacement vector in 3D solids, for example, it has 3 scalar component functions. It thus forms a vector function of 3 scalar functions.  For a stress tensor in 3D solids, we have 9 (6 independent) scalar functions.  

Note that the function for an object is a concept for accounting for the variation of the object along the coordinates. It is a separated concept of the coordinate transformation. Therefore, a tensor at a location is still a tensor at any other location in the coordinate system, so long the same rule is followed for coordinate transformation at locations. However, operations to a tensor object with respect to the coordinates can produces a new tensor object of different order. We will discuss about this in great detail, because it happens all the time in mechanics formulation. In our discussion, we assume all the scalar functions involved are sufficiently continuous and smooth and hence differentiable to the required order, and allowing us to take limits when necessary.

### Gradient of a scalar function, a vector

Gradient of a function converts a scalar number (0th first order tensor) to a vector (1st order tensor). 

Consider a scalar function defined in a domain in the coordinate system. A typical such as a function is one of the components of a displacement vector or stress tensor. 

#### Definition

The gradient of an arbitrary scalar **differentiable** function $f(\mathbf{x})$ of the independent variables (that is a vector) becomes a vector of three functions defined as 

\begin{equation} \label{eqGf}
\boxed{
\nabla f(\mathbf{x}):= \underbrace{\begin{bmatrix} \frac {\partial}{âˆ‚x}\\ \frac {âˆ‚}{âˆ‚y} \\ \frac {âˆ‚}{âˆ‚z} \end{bmatrix}}_{\nabla} \; f(\mathbf{x}) = \begin{bmatrix} \frac {âˆ‚f(\mathbf{x})}{âˆ‚x}\\ \frac {âˆ‚f(\mathbf{x})}{âˆ‚y} \\ \frac {âˆ‚f(\mathbf{x})}{âˆ‚z} \end{bmatrix} }
\end{equation} 

where $\nabla$ (pronounced "nabla") stands for **gradient operator**, and has a vector form. The components in resulted vector $\nabla f(\mathbf{x})$ are, respectively, the partial derivatives of the scalar function with respect to these 3 coordinates. The shape of $\nabla f(\mathbf{x})$ is the same as that of $\mathbf{x}$. We see now, $f$ is a scalar number, a $0$th order tensor, but its gradient becomes a vector of 1st order tensor! It can be subjected to the same rule of coordinate transformation.

### Gradient of a vector of function

Gradient of a vector of functions converts a vector (1st order tensor) to a matrix (2nd order tensor). 

Consider a differentiable vector field in 3D Cartesian coordinates defined by a vector of three scalar functions:  

\begin{equation} \label{eqvf}
\mathbf{f(\mathbf{x})}=f_x (\mathbf{x})\mathbf{i} +f_y (\mathbf{x}) \mathbf{j} +f_z (\mathbf{x}) \mathbf{k}=\begin{bmatrix} f_x (\mathbf{x})  \\ f_y (\mathbf{x}) \\ f_z (\mathbf{x}) \end{bmatrix}
\end{equation} 

where $\mathbf{i}, \mathbf{j}$ and $\mathbf{k}$ are, respectively, the unit vector on $x, y$ and $z$ axes, $f_i (\mathbf{x}) \;  (i=x, y, z)$ are the components of the vector function $\mathbf{f}(\mathbf{x})$. The vector function creates a vector field in the domain where these functions are defined.  Each of the component function is a function of the coordinates. The gradient of a vector of functions is defined using the outer-product of âˆ‡ and $\mathbf{f(\mathbf{x})}$:

 \begin{equation} \label{eqgvf}
 \boxed{
 \begin{split}
\nabla \mathbf{f}(\mathbf{x})&:= \begin{bmatrix} \frac {âˆ‚}{âˆ‚x}\\ \frac {âˆ‚}{âˆ‚y} \\ \frac {âˆ‚}{âˆ‚z} \end{bmatrix} âŠ— \begin{bmatrix} f_x (\mathbf{x})  \\ f_y (\mathbf{x}) \\ f_z (\mathbf{x}) \end{bmatrix} \\
&= \begin{bmatrix} \frac {âˆ‚}{âˆ‚x}\\ \frac {âˆ‚}{âˆ‚y} \\ \frac {âˆ‚}{âˆ‚z} \end{bmatrix} \begin{bmatrix} f_x (\mathbf{x})  & f_y (\mathbf{x}) & f_z (\mathbf{x}) \end{bmatrix} \\ 
&= \left[\begin{matrix}\frac{âˆ‚}{âˆ‚ x} f_x(\mathbf{x}) & \frac{âˆ‚}{âˆ‚ x} f_y(\mathbf{x}) & \frac{âˆ‚}{âˆ‚ x} f_z(\mathbf{x})\\\frac{âˆ‚}{âˆ‚ y} f_x(\mathbf{x}) & \frac{âˆ‚}{âˆ‚ y} f_y(\mathbf{x}) & \frac{âˆ‚}{âˆ‚ y} f_z(\mathbf{x})\\\frac{âˆ‚}{âˆ‚ z} f_x(\mathbf{x}) & \frac{âˆ‚}{âˆ‚ z} f_y(\mathbf{x}) & \frac{âˆ‚}{âˆ‚ z} f_z(\mathbf{x})\end{matrix}\right]
\end{split} }
\end{equation} 

Let us evaluate the gradient of a vector function, assume each of the component function is differentiable, such as the vector functions given below. 

```python
vf = Matrix([y*sp.sin(x), z*sp.sin(y), x*sp.cos(z)]) 
vf
```

### Divergence of a vector function

The divergence a vector of functions produces a scalar function. It coverts a 1st order tensor object to a $0$th order one. 

#### General formulation

Consider a differentiable vector field in 3D Cartesian coordinates defined by a vector of three scalar functions, given in Eq.(\ref{eqvf}) The divergence of a vector of functions is defined as the dot-product of âˆ‡ with $\mathbf{f}$:

\begin{equation} \label{eqDivf}
\boxed{
\begin{split}
\operatorname{div} \mathbf{f}& := \nabla \cdot \mathbf{f} \\ 
&=\left[ \frac{âˆ‚ }{âˆ‚ x}\;\;\; \frac{âˆ‚ }{âˆ‚ y}\;\;\; \frac{âˆ‚ }{âˆ‚ z} \right] \begin{bmatrix} f_x \\ f_y \\ f_z \end{bmatrix} \\ 
& =\frac{âˆ‚ f_x}{âˆ‚ x}+\frac{âˆ‚ f_y}{âˆ‚ y}+\frac{âˆ‚ f_z} 
{âˆ‚ z} \\
&= \frac{âˆ‚ f_i}{âˆ‚ x_i} = f_{i,i}
\end{split} }
\end{equation} 

This is a scalar function, because of the contraction of index $i$. The dot in Eq.(\ref{eqDivf}) stands for the dot-product of vectors: a del operator vector and a given vector of functions. This is a measure of the vector diverges away from a point. The divergence carries different physical meaning depending on application. For example, 

+ If the vector of functions is concentration of a species, it is the spatial rate of loss of mass at that point. The compensation to that loss should be the birth rate. 

+ If the vector is temperature, it is the spatial loss of heat at that point. The compensation to that loss should be the heat source. 

+ If the vector has stress components (all in a same direction) at a point, it is the loss of stresses in that direction. Such a loss must be compensated externally by an applied body force in that direction. This leads to equilibrium at the point together with the externally applied total body force (force density) there. 

Let us demonstrate this using the vector of 3 simplest functions. Consider first a vector with three stress components in a direction that are all constants (independent of the coordinates $x, y$, and $z$): 

```python
x, y, z = symbols('x, y, z')                      # define coordinates
a, b, c = symbols('a, b, c')                        # define constants

vc = Matrix([a, b, c])                      # constant vector function
vc
```

```python
vector.div(vc)                        #divergence of a vector function
```

The divergence of the stress vector with constants is zero. This is because a field of constant stays still and no loss will occur. This means also that the body force in that direction must be zero. We will see an example in the next Chapter. 

Consider now a vector with three stress components in a direction that are linear functions of $x, y$, and $z$:

```python
a1, b1, c1 = symbols('a1, b1, c1')                  # define constants
a2, b2, c2 = symbols('a2, b2, c2')                  # define constants
a3, b3, c3 = symbols('a3, b3, c3')                  # define constants

vl=Matrix([a1*x+b1*y+c1*z, a2*x+b2*y+c2*z, a3*x+b3*y+c3*z])   # linear  
vl
```

```python
vector.div(vl)                        #divergence of a vector function
```

Thus, if a vector of the stresses in a direction are linear functions, its divergence is a constant. This means that the body force in that direction must be a constant that are a sum of the slope in the respective coordinates. Therefore, if a solid is subjected to a constant gravity force in a direction, the stresses in that direction is linear. 

Generally speaking, the function of stresses  is one order higher than th body force function. 

This relationship of stress vector divergence and body force may be used to check stress solutions. We shall do so in some of the examples in the following chapters. Most importantly, it offers a convenient way to derive the equilibrium equations for solids. 

### curl of a vector function

Consider a vector function $\mathbf{f}$ defined in Eq.(\ref{eqvf}). Using the $\nabla$ gradient operator vector, the curl of a vector function can be defined in a similar way as the cross-product of two vectors. Using Eq.(\ref{eqcM}), we can write:  

\begin{equation} \label{eqcvf}
\boxed{
\begin{split}
\underbrace{\mathbf{\nabla\times \mathbf{f}}}_{\mathbf{\mathrm{curl} \;\mathbf{f}}} & := \begin{vmatrix}\mathbf{i} &\mathbf{j} &\mathbf{k} \\\frac{âˆ‚ }{âˆ‚ x}&\frac{âˆ‚ }{âˆ‚ y}&\frac{âˆ‚ }{âˆ‚ z} \\ f_{x} &f_{y} &f_{z} \end{vmatrix} \\
&=\begin{bmatrix}0 &-\frac{âˆ‚ }{âˆ‚ z} &\frac{âˆ‚ }{âˆ‚ y} \\\frac{âˆ‚ }{âˆ‚ z} &0 &-\frac{âˆ‚ }{âˆ‚ x} \\-\frac{âˆ‚ }{âˆ‚ y} &\frac{âˆ‚ }{âˆ‚ x} &0 \end{bmatrix} \begin{bmatrix} f_{x} \\ f_{y}\\ f_{z} \end{bmatrix} \\
&= \begin{bmatrix}\frac{âˆ‚f_z}{âˆ‚y} - \frac{âˆ‚f_y}{âˆ‚z}\\ \frac{âˆ‚f_x}{âˆ‚z} - \frac{âˆ‚f_z}{âˆ‚x}\\ \frac{âˆ‚f_y}{âˆ‚x} - \frac{âˆ‚f_x}{âˆ‚y}\end{bmatrix} 
\end{split} }
\end{equation} 

As shown $\mathrm{curl} \;\mathbf{f}$ is a vector. It is a measure of the components of **rate of rotation** or **rate of circulation** in the coordinate directions at a point in a vector function field. In solid mechanics, a typical example is a bar subject to pure torsion in the axial direction. The rate of the torque in the axial direction of the bar is the curl of the shear stress vector function on the cross section of the bar. The rates of the torque in other two direction will be zero. We will discuss this in Chapter 9. Here, let us take a look at the curl of two simple vector functions created earlier: the constant vector function and the linear vector function, which can offer some intuitive understanding of the curl of a vector function field.   

```python
vector.curl(vc)   # curl of constant vector functions, defined earlier
```

As shown: 1) the curl of a vector function is a vector; 2) when the vector function field is a constant, the curl will be zero, because there is no rotation at all at any point in the field. 

```python
vector.curl(vl)     # curl of linear vector functions, defined earlier
```

For linear vector functions, the curl will be constant in all three directions, because these three vector functions varying linearly in its own way, creating rotations at any point in the field. The condition for linear vector function field has no rotation with respect to $x$-axis would be $b_3 = c_2$, no rotation with respect to $y$-axis is $a_3 = c_1$, and no rotation with respect to $z$-axis is $a_2 = b_1$. The proof is given in the following codes. 

```python
vl_0curl = Matrix([a1*x+a2*y+a3*z, a2*x+b2*y+c2*z, a3*x+c2*y+c3*z])   
vl_0curl                     # linear vector function without rotation
```

```python
vector.curl(vl_0curl)        # linear vector function without rotation
```
