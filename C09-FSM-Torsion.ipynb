This Chapter discusses torsional bar members with various cross-sectional shape. The objective is to find the relationship between the **twist angle** (an important measure of deformation) and the applied **torque** (a typical force application), and **stresses** (critical for failure analysis) on the cross-section of the members. Because our focus is purely on torsional deformation, the members or components should have a **slender** geometry, like a bar or rod. Thus, we will call all this type of torsional members and components a **bar**, because it captures the major geometric feature, allowing proper assumptions to be made. The cross-section of the bar can be arbitrary, so long its cross-sectional dimensions are sufficiently smaller than its length. The torques applied about the axial axis of the bar. 

The **major challenge** in dealing with torsional bars is its three-dimensional (3D) nature: the torque is applied about is axial axis (z-axis). The stresses occurred, on the other hand, are on its cross-sectional plane (x-y plane). The **art** of the techniques is to convert the 3D problem into a 1D, or at most 2D one, so that solutions can be found analytically. Detailed formulations will be provided in this chapter on how this is done. 

If the bar is bulky, it needs to be treated in principle as a true three-dimension (3D) problem. The most effective techniques will be numerical, including the finite element method (FEM)\cite{liu2013finite}, smoothed FEM \cite{liusmoothed}, and even the meshfree methods \cite{liu2002mesh}. All these techniques require intensive formulation and substantial coding. It is thus beyond the topic of this book. 

This chapter is written in reference to the textbooks \cite{boresi2002advanced, gere1999mechanics, Timoshenko1970, xuzl1979, megson1996structural}. Our study begins with the simplest and the most fundamental case: bars with circular cross-section which are the most widely used in engineering. 

### Bars with circular cross-section

#### Problem description

Consider a solid cylindrical bar with circular cross-section. It has a cross-sectional area $A$ and length $L$. It is defined with the Cartesian coordinates $(x, y, z)$. Let $z$ being the axial coordinate and it is perpendicular to the cross-section. It is subjected to torque $T$ about the $z$-axis. An originally straight-line AB on the lateral surface of the bar becomes a helical curve AB*, due to the twist deformation of the bar, as shown in Fig.\ref{image:bar_torsion}. Our task is to find out how the twist angle relates to the applied torque, and the stresses resulted in the bar.  

<img src="./images/bar_torsion.png" alt=" " width="500"/><br>
\begin{figure}
  \includegraphics[width=5cm]{}
  \caption{\label{image:bar_torsion} A bar with circular cross-section subjected to a torque, undergoing a twist deformation.}
\end{figure}

#### Assumptions

Clearly, any cross-section of the bar is under shearing, which can results shear stresses. The shearing deformation accumulates to twist angle that may change along the $z$-axis. To formulate this problem properly, we assume:

1) The plane of the cross-sections of the bar remains perpendicular to the $z$-axis. 

2) The rotation is small, and cross-section rotates as a rigid plane about $z$-axis. 

3) The rotation 𝛩 of a given section, relative to the base plane $z = 0$, depends linearly on the distance $z$ from the base plane. 

Based on these assumptions, we have 

\begin{equation} \label{eqbeta}
𝛩 = 𝜗_t z
\end{equation} 

where $𝜗_t$ is the twist angle per unit length, and is a constant (does not change with $x, y$ and $z$). It is also sometimes called torsional curvature (rate of angular change with respect to $z$). 

These assumptions practically fixed the dependence of the problem on $z$. The problem becomes 2D with displacements on the cross-section of the bar. 

#### Displacements on the cross-section

Because the cross-section undergoes a rotation about $z$-axis, and the rotation is small, the displacements on the cross-section can be shown in Fig.\ref{image:uv_torsion}.

<img src="./images/uv_torsion.png" alt=" " width="300"/><br>
\begin{figure}
  \includegraphics[width=5cm]{}
  \caption{\label{image:uv_torsion} Displacement components on the cross-section of a circular bar subjected to a torque.}
\end{figure}

Because the cross-section rotates rigidly, these displacement components can be obtained by simple geometry relations. It is written as,

\begin{equation} \label{equvw0}
u = -y𝛩; \;\; v = x𝛩; \;\; w=0 
\end{equation} 
or
\begin{equation} \label{equvw}
u = -y𝜗_t z; \;\; v = x𝜗_t z; \;\; w=0 
\end{equation} 

It appears that the displacements depend on $x$, $y$, and $z$: a 3D problem. It is controlled by one parameter: $𝜗_t$ or $𝛩$. 

#### Shear strains and stresses on the cross-section

Using the small strain-displacement relations for 3D solid given in Chapter 4, and $𝜗_t$ is independent of $x,y,z$, we can find the strains by hand in the bar: 

\begin{equation} \label{eqEzxy0}
ε_{xx}=\frac{∂ u}{∂ x}=0; \;\; ε_{yy}=\frac{∂ v}{∂ y}=0; \;\; ε_{zz}=\frac{∂ w}{∂ z}=0; \;\; ε_{xy}=\frac{1}{2}(\frac{∂ u}{∂ y}+ \frac{∂ v}{∂ x})=0  
\end{equation}
and 
\begin{equation} \label{eqEzxy}
2ε_{zx}=𝛾_{zx}=\frac{∂ u}{∂ z}+ \frac{∂ w}{∂ x}=-y𝜗_t; \;\; 2ε_{zy}=𝛾_{zy}=\frac{∂ w}{∂ y}+ \frac{∂ v}{∂ z}=x𝜗_t  
\end{equation}

It is clear that there are only two nonzero strain components. All these can also be obtained using the following codes. First, defined the displacement vector: 

```python
x, y, z = symbols("x, y, z")             # define symbolic coordinates
X = Matrix([x, y, z])
𝜗t = symbols("𝜗_t")                           # twist angle per length

U = Matrix([-y*𝜗t*z, x*𝜗t*z, 0]) # Assumed displacement vector [u,v,w]
U
```

```python
# Compute the strains (see Chapter 4): 
U_g = gr.grad_vf(U, X)                         # displacement gradient  
ε = (U_g + U_g.T)/2                          # strain tensor,symmetric 
ε
```

The codes finds again that there are only two non-zero strains, and they all are shear strains. 

For isotropic materials, there can only be two shear stresses on the cross-section of the bar, which relate to shear strains by only the shear modulus $G$ (see Chapter 5). These are:

\begin{equation} \label{eqSzxy}
σ_{zx}=G𝛾_{zx}=-yG𝜗_t; \;\; σ_{zy}=G𝛾_{zy}=xG𝜗_t  
\end{equation}

where the shear modulus $G$ is a material constant. Notice that these stresses are independent of $z$, meaning that these stresses are uniform along the axial axis of the bar. 

With this simple stress state, we can perform some interesting analysis to it, using the following code. The stress tensor at any point in a torsion bar is given as follows. 

```python
G = symbols("G") 
σ_tensor = G*(2*ε)                                      # 𝛾_ij = 2𝜀_ij
σ_tensor  
```

##### Divergence of the stress states in a circular bar

This set of stresses satisfy the equilibrium equation for 3D solids given in Chapter 6 (considering static problem with no body force). This is because its divergence (see Chapter 3) are zeros: 

```python
div_σ_tensor = gr.div_tensorF(σ_tensor, X)   # a vector of 3 functions
div_σ_tensor                   # Or use: vector.div_tensor(σ_tensor.T)
```

Clearly, the stress tensor of torsion is a divergence free tensor. Therefore, the equilibrium equations is satisfied for free body forces in the bar. 

##### Curl of the stress vector field in a circular bar

The stresses on the $z$-plane in a circular bar under torsion can form a vector function with three components: $σ_{zx}, σ_{zy}, σ_{zz}$. Let us take a look at the curl (see Chapter 2) of this vector field formed by these torsional stresses. 

```python
# Define a vector field as symbolic functions of (x, y, z)
σ𝑧 = symbols("σ_𝑧") 
σ𝑧 = σ_tensor[2,:]                       # get the vector on z-surface 
σ𝑧
```

```python
curl_F = gr.curl_vf(σ𝑧, X)
curl_F                     # Curl of the torsional stress vector field 
```

This means that

* the rate of rotation of the torsional stress vector field around $x$- and $y$-axis are all zero at any point on the cross-section. The field has zero circulation around these two axes. 

* It has a positive constant circulation at any point in the cross-section around $z$-axis. The shear stresses are in the $x-y$ plane, and the torque (curl) is with respected to the $z$-axis. This is a typical feature of a circular bar under pure torsion.  

Also, the sum of these two shear stress components are in the circumferential direction on the cross-section of the bar, which is the red arrow in Fig.\ref{image:uv_torsion}. This can be confirmed by viewing the stress state in the polar coordinate ($r$, 𝜃). 

#### Distribution of shear stress on the cross-section

Again, because the cross-section undergoes a rigid body rotation about $z$-axis, and the rotation is small, the displacements on the cross-section in the polar coordinate ($r$,𝜃) are:

\begin{equation} \label{equrp}
u_r = 0; \;\; u_𝜃 = r 𝜗_t z; \;\; w=0 
\end{equation} 

The only one nonzero displacement $u_𝜃$ that is in the circumferential 𝜃-direction (the red arrow r𝛩 in Fig.\ref{image:uv_torsion}). This means again that the problem is 1D. 

The stress in the polar coordinates can be obtained using the stress tensor transformation rule: 

\begin{equation} \label{eqSTM3}
\mathbf{σ}'=\mathbf{T}\mathbf{σ}\mathbf{T}^\top
\end{equation}

where $\mathbf{σ}'$ is the stress tensor in the polar coordinate system, $\mathbf{σ}$ is the stress tensor in the Cartesian coordinates with components given in Eq.(\ref{eqSzxy}), and $\mathbf{T}$ is the transformation matrix defined in Eq.(2.\ref{eqTMdef}). 

Let us use the following Sympy code to get it done. 

```python
# Compute the stress tensor on the cross-section of the bar: 

𝜗, 𝜃  = "\u03D1", "\u03B8"                 # Define symbol in hex code
G, 𝜗t = symbols("G, 𝜗_t")                  # Define symbolic variables
r, 𝜃  = symbols("r, 𝜃")                        # for polar coordinates

# write the stresses in terms of r and 𝜃:
σzx = -r*sp.sin(𝜃)*G*𝜗t                                      # y=rsin𝜃
σzy =  r*sp.cos(𝜃)*G*𝜗t                                      # x=rcos𝜃

S = Matrix([[ 0 ,  0 , σzx],   # stress state for torsion w.r.t z-axis
            [ 0 ,  0 , σzy],
            [σzx, σzy,  0 ]])
S    
```

Readers are not to be confused about the symbol 𝜃 and $𝜗_𝑡$. The former is the coordinate for the polar system, and the latter is the twist angle per unit length. 

Perform coordinate transformation for the stresses.  

```python
# coordinate transformation for the stresses: 
T = gr.transferMts(𝜃)  # transformation matrix for rotation 𝜃 w.r.t. z
Sp=T@S@T.T           # S: stress tensor; @: the shorthand of np.matmul 
gr.simplifyM0(Sp)                     # simply the terms in the output
```

We found that stress components are all zero very where in the polar coordinates: 

\begin{equation} \label{eqS0}
σ_{rr}=σ_{r𝜃}=σ_{rz}=σ_{𝜃𝜃}=σ_{zz}=0
\end{equation}

The only one nonzero component is: 

\begin{equation} \label{eqSzrp}
σ_{z𝜃}=rG𝜗_t 
\end{equation}

which changes linearly in $r$ and is in the circumferential direction, as shown in Fig.\ref{image:uv_torsion}. The largest shear stress is registered by the surface of the bar. Because there is only one shear stress on the cross-section of a bar, the stress is often denoted as $\tau$, when dealing with torsion. 

\begin{equation} \label{eqTau}
𝜏 = rG𝜗_t  
\end{equation}

The strain should also have just one component: $2ε_{z𝜃}=r𝜗_t$ changing also linearly in $r$. 

Notice in the coordinate transformation, we used an arbitrary angle 𝜃. However, the transformed results do not depend on 𝜃 at all. The means that the result obtained is applicable to any location on the cross-section. This is the consequence of the rigid cross-section rotation assumption. The problem is essentially one dimensional. 

#### Boundary condition on the lateral surface of the bar. 

Because the only nonzero stress is in the circumferential direction, it will not result in any stress on the lateral surface of the bar. Therefore, the zero-stress boundary condition on the lateral surface of the bar is also satisfied. 

A rigorous proof is given by the code right above Eq.(\ref{eqS0}) with gives for any 𝜃: $σ_{rr}=σ_{r𝜃}=σ_{rz}=0$. These are all the possible stress components on the lateral surface of a circular bar, and they all are zero regardless of 𝜃. 

Note also that $σ_{rr}=σ_{r𝜃}=σ_{rz}=0$ is true for any $r$. This means that any cylinder in the circular bar rotates together without an inter-surface stress. In other words, if we put cylinders together to form a circular bar, the assembly behaves the same as one solid circular bar in terms of pure torsion, if they all have the same twist angle. 

Note that in usual applications, the shear stress distribution on the cross-section at the two ends of the bar may not be linearly distributed. In those cases, the solution is still applicable far away from the ends, based on the Saint-Venant principle. This implies that our formulation applicable in general for slender bars for locations away from the two ends. 

#### Equilibrium on the cross-section, twist-torque relation

We found the stresses and know their distribution on the cross-section. The amplitude of the stresses must depend on the level of the torque applied. For example, if the torque is zero, the stresses must be zero. We need now to establish the relationship between the stresses and the applied torque. For this, naturally, we must invoke equilibrium conditions.

Let us now examine the equilibrium status on the cross-section of the bar subjected to a torque $T$. The sum of all moments with respect to $z$-axis should be zero. Thus, the torque $T$ must be resulted from these two shear stresses. 

<img src="./images/SzxyOnBar.png" alt=" " width="300"/><br>
\begin{figure}
  \includegraphics[width=5cm]{}
  \caption{\label{image:SzxyOnBar} Shear stresses on the cross-section of a circular bar result in a torque.}
\end{figure}

With the help of Fig.\ref{image:SzxyOnBar}, we have 

\begin{equation} \label{eqMz}
\begin{split}
\sum M_z = 0: T&=\int_A (x σ_{zy} - y σ_{zx}) dA =\int_A (x^2G𝜗_t + y^2 G𝜗_t) dA \\
&=\int_A G𝜗_t \underbrace{(x^2+ y^2)}_{r^2} dA =G𝜗_t \underbrace{\int_A r^2 dA}_{J} =G𝜗_t J 
\end{split}
\end{equation}

where $J$ is the **polar second moment of area**. For circular cross-section, we have

\begin{equation} \label{eqJ}
\begin{split}
J=\int_A r^2 dA = \int_A r^2 \underbrace{rd𝜃}_{dA} dr = 2𝜋 \int_0^R r^3 dr = \frac{𝜋R^4}{2} 
\end{split}
\end{equation}

#### On polar  second moment of area

The polar second moment of area $J$ is a property of the cross-section of a bar, and it is the most important one for all bars subject to torsion. The twist angle and the shear stress all depend on it. For circular cross-section, we have a very nice Eq.(\ref{eqJ}) for computing it. It is derived using the polar coordinate system, as shown above. The word "second" stands for $r^2$ in the integral in Eq.(\ref{eqJ}). 

For other shapes of cross-section, the formulas can be very complicated, and often the polar coordinate system is not usable in the derivation. We call such a $J$  **equivalent** polar second moment of area. We need to use the word "polar" because the second moment of area in Cartesian coordinate is a different one. A better unified name might be **torsional second moment of area** for bar with any type of cross-section.  

Our strategy in deal with bars with other types of torsional bars is to find the equivalent polar second moment of area of the cross-section of the bar. Once this is done, the formulas for computing the twist angle and the shear stress will be the same. 

#### Twist-angle and torque relations

Equation (\ref{eqMz}) gives the twist-angle per-unite-length, $𝜗_t$:

\begin{equation} \label{eqtT}
𝜗_t = \frac{T}{GJ} 
\end{equation}

If the bar is fixed at $z=0$, the (total) twist-angle at an arbitrary location $z$ becomes: 

\begin{equation} \label{eqtTz}
𝛩 = \frac{T}{GJ}z 
\end{equation}

The total twist angle for a bar with length $L$ becomes, 

\begin{equation} \label{eqtTL}
\boxed{𝛩 = \frac{L}{GJ}T  } \;\; \text{ for 1D bars subject to torque } T
\end{equation}

where $\frac{GJ}{L}$ is the **torsional stiffness of the bar**, and $GJ$ the **torsional stiffness per length**. 

Comparing the solution for a 1D bar subjected to axial force $F$, which is given in Chapter 7:

\begin{equation} \label{equL}
\boxed{ u_R = \frac{L}{EA }F  } \;\; \text{ for 1D bars subject to axial force } F
\end{equation}

The formula is exactly the same, if replacing 

\begin{equation} \label{eqEG}
u_R \;\text{ by }\;  𝛩; \;\;\; E \;\text{ by }\; G ; \;\;\; A \;\text{ by }\; J
\end{equation}

Equation (\ref{eqtTL}) can also be written as: 

\begin{equation} \label{eqTtL} 
\boxed{T = \frac{GJ}{L}𝛩  }
\end{equation}

In this case,  $\frac{L}{GJ}$ is the **torsional flexibility of the bar**. 

Using Eq.(\ref{eqTau}), we have the shear-stress torque relation: 

\begin{equation} \label{eqTauT}
𝜏 = rG𝜗_t  = \frac{Tr}{J}
\end{equation}

The maximum shear stress is registered at the furthest point from the $z$-axis, $r=R $ (by the lateral surface of the bar): 

\begin{equation} \label{eqTTmax}
𝜏_{max} = \frac{T R}{J}
\end{equation}

All these formulas are useful in design torsional bars. These formulas are also the base formulas for many other types of bars. The difference will be on the calculation of $J$. 

#### Continuous formulation

Consider a small cell $dz$ in the bar, as shown in Fig.\ref{image:bar_torsion}. The twist angle for it becomes,

\begin{equation} \label{eqtTdz}
d𝛩 = \frac{T(z)}{GJ} dz; \;\; \text{ or } \;\; \frac{d𝛩}{dz} = \frac{T(z)}{GJ} 
\end{equation}

This is the differential equation for twist-torque relation, which allows the torque to be a function of $z$, meaning that one can apply torque in a distributed manner. In this case, the equilibrium $\sum M_z = 0$ is enforce locally at any location $z$ in the integral manner over the cross-section of the bar. Equation (\ref{eqtTdz}) is a simple 1D DE of 1st order and can be solved with ease.  

In general, even $J$ and/or $G$ can also be a function of $z$, but the bar must be "long" to justify being 1D. 

#### Bars with hollow cylindrical cross-section

In practice, bars designed for taking torsional loads are often with a hollow cylindrical cross-section, as shown in Fig.\ref{image:CylinderBar}. This is because the materials near the axial axis does not take much stress, as shown in Fig.\ref{image:uv_torsion}, and can be removed. 

<img src="./images/CylinderBar.png" alt=" " width="300"/><br>
\begin{figure}
  \includegraphics[width=5cm]{}
  \caption{\label{image:CylinderBar} Bar with hollow cylindrical cross-section.}
\end{figure}

In such a case, all the formulations derived so far still apply. The only change is the calculation of the  polar second moment of area $J$. It is calculated using 

\begin{equation} \label{eqJab}
J=\int_A r^2 dA =  2𝜋 \int_{R_I}^{R_O} r^3 dr = \frac{𝜋({R_O}^4-{R_I}^4)}{2} 
\end{equation}

where $R_I$ is the inner radius, and $R_O$ is the outer radius of the cylinder. The maximum shear stress is registered at the furthest point from the $z$-axis, $r={R_O}$ (the outer radius): 

\begin{equation} \label{eqTbmax}
𝜏_{max} = \frac{T{R_O}}{J}
\end{equation}

The cross-section areas is computed using

\begin{equation} \label{eqAab}
A = 𝜋({R_O}^2-{R_I}^2) 
\end{equation}

#### Bars with thin-wall cross-sections

If the hollow cylinder is very thin with thickness $t$, we have ${R_I}≈{R_O}$, and ${R_O}-{R_I}=t$. Equation (\ref{eqJab}) gives, 

\begin{equation} \label{eqJR}
J= \frac{𝜋({R_O}^4-{R_I}^4)}{2} = \frac{𝜋({R_O}^2+{R_I}^2)({R_O}+{R_I})({R_O}-{R_I})}{2}≈ 2𝜋R_m^3t
\end{equation}

where $R_m=\frac{{R_O}+{R_I}}{2}$ is the mean radius of the cylinder. 

Equation (\ref{eqJR}) can be reformulated as: 

\begin{equation} \label{eqJRt}
J≈2𝜋tR_m^3 =  \frac{4t (𝜋R_m^2)^2}{2𝜋R_m} = \frac{4t A_m^2}{l_m} 
\end{equation}

where $A_m$ is the area of the circle with mean radius $R_m$, and is computed using

\begin{equation} \label{eqAm}
A_m = 𝜋R_m^2 
\end{equation}

and $l_m$ is the length of the mean perimeter of the circle with $R_m$.

It is found that the following equation works for thin-wall cylinders with arbitrary cross-section shape, as long as the cross-section is enclosed. We thus have the general formula: 

\begin{equation} \label{eqJRta}
\boxed{J= \frac{4t A_m^2}{l_m} }
\end{equation}

where $A_m$ is the area formed by the middle line of the cross-section, $l_m$ is the length of the mean perimeter of the cross-section, and $t$ is the thickness of the wall. This is because cross-sections regardless the shape all have these three geometric features: $A_m$, $l_m$, and $t$. The $J$ value should be proportional to $t$, inversely proportional to $l_m$, and $A_m^2$ is needed for dimension match, similar to the thin wall circular cylinder. Detailed rigorous (but lengthy) derivation leading to Eq.(\ref{eqJRta}) can be found in the next section and in Sections 9.2 and 9.6.

Note that when the cross-section is not closed (have a cut) the $J$ value is drastically reduced (see examples in Section 9.6).

#### Example: A shaft with two segments subject to two torques

Consider a shaft consisting of two uniform bars perfectly connected together, as shown in Fig.\ref{image:Shaft2Bars}. It is fixed a point A. Both bars have solid circular cross-section. It is loaded at point B with torque $T_1$, and at point C with $T_2$. The diameters of these two bars are denoted as $d_1$ and $d_2$, and their length are denoted as $L_1$ and $L_2$, respectively, for bar-1 and bar-2. Both bars are made of Grade-304 stainless steel with shear modulus $G=$88 GPa. 

<img src="./images/Shaft2Bars.png" alt=" " width="500"/><br>
\begin{figure}
  \includegraphics[width=5cm]{}
  \caption{\label{image:Shaft2Bars} A shaft consisting of two uniform bars perfectly connected together.}
\end{figure}

1. Derive the formula for computing the twist angle at C.

2. Derive the formula for computing the maximum shear stress and its location for bar-1. 

3. Determine the ratio of these two diameters $d_1/d_2$, so that the maximum shear stress these two bars are the same. 

4. Let $T_1=100$ Nm, $T_2=50$ Nm, $L_1=0.5$ m and $L_2=1.0$ m. Compute items 1, 2 and 3. 

**Solution to 1**:    

First, the 2nd polar moment of area for these two bars are: 

\begin{equation} \label{eqJ12}
J_1= \frac{𝜋(d_1/2)^4}{2}=\frac{𝜋d_1^4}{32}; \;\;\; J_2=\frac{𝜋d_2^4}{32}
\end{equation}

The twist angle at C in reference to B is

\begin{equation} \label{eqt2}
𝛩_{BC} = \frac{T_2L_2}{GJ_2} = \frac{32T_2L_2}{𝜋Gd_2^4}
\end{equation}

In bar-1, the torque is $(T_1+T_2)$, and hence the twist angle at B in reference to A is

\begin{equation} \label{eqt1}
𝛩_{AB} = \frac{32(T_1+T_2)L_1}{𝜋Gd_1^4}
\end{equation}

The total twist angle at C becomes:

\begin{equation} \label{eqt12}
\begin{split}
𝛩_C= 𝛩_{AB} + 𝛩_{BC}  &= \frac{32(T_1+T_2)L_1}{𝜋Gd_1^4} + \frac{32T_2L_2}{𝜋Gd_2^4} \\
&= \frac{32}{𝜋G}\bigg(\frac{(T_1+T_2)L_1}{d_1^4} +\frac{T_2L_2}{d_2^4} \bigg)
\end{split}
\end{equation}

**Solution to 2**:  compute the maximum shear stress and its location for bar-1

For bar-1, we have $R_1=\frac{d_1}{2}$. Thus, 

\begin{equation} \label{eqssmax1}
𝜏_1^{max} = \frac{(T_1+T_2)d_1}{2J_1}=\frac{16(T_1+T_2)}{𝜋d_1^3}
\end{equation}

**Solution to 3**:  determine the ratio of these two diameters $d_1/d_2$:

For bar-2, we have $R_2=\frac{d_2}{2}$. Hence, 

\begin{equation} \label{eqssmax2}
𝜏_2^{max} = \frac{T_2d_2}{2J_2}= \frac{16(T_2)}{𝜋d_2^3}
\end{equation}

The ratio $d_1/d_2$ is found when $𝜏_1^{max}=𝜏_2^{max}$:

\begin{equation} \label{eqdr}
\frac{d_1}{d_2} = \sqrt[3]{\frac{T_1+T_2}{T_2}} 
\end{equation}

### Bars with enclosed thin-wall cross-sections

Consider a long bar with a uniform cross-section that is a hollow with an enclosed thin-wall, as shown in Fig.\ref{image:ThinWallT}. The axis of the bar is denoted as $z$ that is perpendicular to the plane of the cross-section, and passing through its centroid O. The thickness $t$ of the wall needs not to be constant, and can change at difference location denoted by a curve coordinate $s$, and local coordinate $r$ orthogonal to $s$. Thus, $r$, $s$, and $z$ axes form a local   orthogonal coordinate system with the right-hand rule. 

The shape of the cross-section can be arbitrary but closed. The bar is subjected to a torque $\mathbf{T}$ with respect to the $z$-axis, resulting a twist 𝛩 of the cross-section by at $z=L$.

Before our analysis, we make the following assumptions:

1. The wall thickness is very small compared to the other dimensions of the cross-section.
2. The shear stress direction is along the meddle line of the thin-wall boundary.
3. The cross-section rotates rigidly with respect to the axis of the bar under the action of a torque, so that the displacements in the $x-y$ plane behave like a rigid disk. The displacement in the $z$-direction, $w$, is not constrained. 
4. The bar is loaded only with a torque with respect to its axis.

<img src="./images/ThinWallT.png" alt=" " width="850"/><br>
\begin{figure}
  \includegraphics[width=5cm]{}
  \caption{\label{image:ThinWallT} A long bar with a uniform cross-section that is a hollow with an enclosed thin-wall. It is fixed at $z=0$ and subjected to a torque $\mathbf{T}$ with respect to the $z$-axis of the bar, and the entire cross-section at the right-end undergoes a rigid rotation 𝛩.}
\end{figure}

Based on these assumptions, the stresses on the cross-section (the $z$-plane) become:

\begin{equation} \label{eqsz}
\begin{split}
&𝜎_{zr}=0, \;\; \text{ thin-wall, and circumferential surface is free of stresses}  \\
&𝜎_{zz}=0, \;\; \text{ the bar is not loaded in the axial direction} 
\end{split}
\end{equation}

The only nonzero stress on the cross-section is $𝜎_{zs}$. For convenience, it is denoted as 𝜏. Its direction is in the tangent direction of the middle line of the thin-wall, and its positive direction follows the standard right-hand rule for rotation with respected to the $z$-axis. This creates a flow of shear stress along the cross-section of the bar. Because the wall is thin, the shear stress 𝜏 is uniform across the thickness (or 𝜏 is treated as the averaged shear stress) at any fixed location $s$. 

#### Concept of shear flow

The **shear flow** is quite important concept when dealing with shear stresses in thin-walled structural members, and will be used for multiple times in this book. It usually denoted as $q$. To introduce this concept, let us cut an infinitely small free-body diagram from the thin-walled bar at an arbitrary location $s$, shown in Fig.\ref{image:ThinWallT}(a). We can  look at the equilibrium of all the forces in the $z$-direction: 

\begin{equation} \label{eqFz0}
\sum F_z = 0: \; 𝜏_{s+ds} t_{s+ds} dz −𝜏_{s} t_{s} dz = 0 
\end{equation}
which is 
\begin{equation} \label{eqts}
 𝜏_{s+ds} t_{s+ds} =  𝜏_{s} t_{s} 
\end{equation}

This means that the product $𝜏_{s}t_{s}$ does not change when $s$ changes. It is constant along $s$. This conserved quantity is call **shear flow** defined as: 

\begin{equation} \label{eqq}
q =  𝜏_{s} t_{s} 
\end{equation}

The shear stress $𝜏_{s}$ and the wall thickness $t_{s}$ may change with $s$, but $q$ is constant for a given bar with thin-walled cross-section subjected to torque $\mathbf{T}$! 

#### Torque and shear-flow relation 

Because the shear stress is generated by the torque, we have 

\begin{equation} \label{eqTtau}
T = \oint \underbrace{𝜏_{s} t_{s}}_{q\text{:constant}} r_{s} ds = 2q \underbrace{\oint \underbrace{\frac{1}{2}r_{s} ds}_{dA}}_{A_m} = 2qA_m
\end{equation}

where $A_m$ is the area enclosed by the middle line of the wall on the cross-section, and $r$ is "radius" at point $s$ which changes with $s$ (which is different from the circular cross-section). It is the shortest distance from point O in the $r$-axis direction. Clearly, this formula holds only if the cross-section is closed, allowing the closed loop integration.  We obtain now a very simply but important equation, known as the Bredt-Batho formula: 

\begin{equation} \label{eqT2qA}
\boxed{ T = 2qA_m} \;\; \text{ or }\;\; \boxed{ q=\frac{T}{2A_m} }
\end{equation}

This means that the torque is twice the shear flow multiplied by the mean cross-section area. With Eq.(\ref{eqT2qA}), the shear stress, as a function of $t_s$ and hence of $s$, can be easily obtained for a given torque: 

\begin{equation} \label{eqtauT}
\boxed{𝜏_s = \frac{T}{2A_m t_s } }
\end{equation}

Note here that $t_s$ is in general a function of $s$. It is clear that **the maximum shear stress is registered where $t_s$ is smallest**. 

#### Twist-angle and torque relation

The final piece of formulation is the relation between the twist-angle and the torque. This clearly leads to the stiffness of the bar under torsion. Naturally, it must involve the shear strain and stress relation. 

Based on the strain-displacement relations similar to Eq.(\ref{eqEzxy}), and the fact that $z$-axis and $s$-axis are always orthogonal, we have

\begin{equation} \label{eqgzs}
\begin{split}
&𝛾_{zr} =0, \;\; \text{ because } \;\; 𝜎_{zr}=0 \\
&𝛾_{zs} = 2𝜀_{zs}= \left(\frac{∂ w}{∂ s} + \frac{∂ u_s}{∂ z}  \right) 
\end{split}
\end{equation}

Because the cross-section rotates rigidly, the displacement $u_s$ relates the twist angle 𝛩 as (see Fig.\ref{image:ThinWallT}): 

\begin{equation} \label{equsrt}
u_s = r_s 𝛩 
\end{equation}

Note that $r_s$ is in general a function of $s$, but independent of $z$. Substituting Eq.(\ref{equsrt}) into Eq.(\ref{eqgzs}), and using the constitutive equation for shear strain and stress, we obtain: 

\begin{equation} \label{eqgzs1}
𝛾_{zs} = \frac{𝜏_s}{G} = \frac{q}{G t_s} = \left(\frac{∂ w}{∂ s} + \frac{∂ [r_s 𝛩] }{∂ z}  \right) = \left(\frac{∂ w}{∂ s} + r_s\frac{∂𝛩}{∂ z}  \right)
\end{equation}
which gives,
\begin{equation} \label{eqqGt}
\frac{q}{G t_s} = \left(\frac{∂ w}{∂ s} + r_s\frac{∂𝛩}{∂ z}  \right)
\end{equation}

Integration of the foregoing equation along $s$ over the entire cross-section of the bar gives,

\begin{equation} \label{eqintq}
 \oint \frac{q}{Gt_s} ds=  \oint  \frac{∂ w}{∂ s} ds +  \oint r_s\frac{∂𝛩}{∂ z} ds  
\end{equation}

Now, because the cross-section rotates rigidly, 𝛩 does not depend on $s$, we have 

\begin{equation} \label{eqintq1}
 \oint \frac{q}{Gt_s} ds=  \underbrace{w \big|_{s=0}^{s=l_m}}_{=0}  + \frac{∂𝛩}{∂ z} \underbrace{\oint r_s ds}_{2A_m}  
\end{equation}

where $l_m$ is the perimeter length of the middle line of the cross-section, $A_m$ is the area enclosed by the middle line of the cross-section. Because the cross-section is enclosed, and $s$ comes back to its original location after looping around, $w |_{s=0}^{s=l_m} =0$. Note also that $q$ is independent of $s$, Eq.(\ref{eqintq1}) becomes, 

\begin{equation} \label{eqtdz}
\boxed{
 \frac{∂𝛩}{∂ z}= \frac{q}{2A_mG} \oint \frac{1}{t_s} ds }
\end{equation}

where $\frac{∂𝛩}{∂ z}$ is the twist angle per unit length $𝜗_t$.  

Using Eq.(\ref{eqT2qA}), Eq.(\ref{eqtdz}) can also be written as

\begin{equation} \label{eqtdzJ}
\boxed{
 \frac{∂𝛩}{∂ z} = \frac{T}{4A_m^2G} \oint \frac{1}{t_s} ds = \frac{T}{GJ}}
\end{equation}

where $J$ is the **equivalent polar moment of area** of the cross-section, defined as 

\begin{equation} \label{eqJt}
\boxed{
J = \frac{4A_m^2}{\oint \frac{1}{t_s} ds} }
\end{equation}

If the thickness $t_s$ is piecewise constant: $t_1, t_2,...,t_n$ with the corresponding lengths: $l_1, l_2,...,l_n$, we shall have: 
 

\begin{equation} \label{eqJtili}
\boxed{
\oint \frac{1}{t_s} ds = \sum_i^n \frac{l_i}{t_i} }
\end{equation}

If the bar has a uniform wall thickness $t$, the foregoing equation is simplified to 

\begin{equation} \label{eqJl}
\boxed{
J = \frac{4tA_m^2}{l_m} }
\end{equation}

where $l_m$ is the perimeter of the middle line of the thin wall cross-section. This is Eq.(\ref{eqJRta}) we obtained earlier by simple induction or guessing.  

If the cross-section is circular with radius $R$ (that is constant), we have $l_m=2\pi R_m$, and $A_m=\pi R_m^2$, we have Eq.(\ref{eqJR}). Readers may find the formula for $J$ for a rectangular thin-wall cross-section.  

If Eq.(\ref{eqbeta}) holds: $𝛩 = 𝜗_t z$, Eq.(\ref{eqtdzJ}) leads to Eq.(\ref{eqtT}) and also Eq.(\ref{eqtTL}).

### Bars with non-circular cross-sections *

#### Setting of the problem

Consider a torsion bar member with non-circular cross-section subjected to a torque $T$. It is defined in coordinates $(x, y, z)$ with $z$-axis being the axial axis of the bar. The cross-section of the bar is assumed uniform along $z$-axis. 

Because the non-circular shape, the cross-section may not remain plane and can warp, meaning the displacement $w$ may vary with $x$ and $y$. The rotation of the bar changes along with z-axis. Therefore, the problem is three dimensional. The method used to perform analysis is the so-called Saint-Venant’s semi-inverse method \cite{boresi2002advanced}. It approximates the displacements in such a way that it leads to a stress state $(σ_{zx}, σ_{zy})$ on the cross-section, under torque $T$, aiming to convert the problem to a two-dimensional one. Our formulation and notation follow largely the textbook [4]. Codes are developed for some of the derivations. 

#### Assumed displacements 

Consider a point B at $(x, y, z)$ on a cross-section in an undeformed bar. When it is deformed subjecting $T$, the cross-section rotates by 𝛩 about $z$-axis with respect to the origin, and B is displaced by displacements $(u, v, w)$ and moves to B*, as shown in Fig.\ref{image:bar_torsionW}.

<img src="./images/bar_torsionW.png" alt=" " width="500"/><br>
\begin{figure}
  \includegraphics[width=5cm]{}
  \caption{\label{image:bar_torsionW} A bar with non-circular cross-section subjected to a torque. Warping may occur on the cross-section.}
\end{figure}

The semi-inverse method assumes:

1. The rotation is the principal cause of displacements $(u, v)$, which decouples $u$ and $v$ from $w$.

2. The twist angle changes linearly with $z$: $𝛩 = 𝜗_tz$, where $𝜗_t$ is the angle of twist per unit length and is a constant (does not change with $(x, y, z)$).

Based on these assumptions, the displacements can be written as:   

\begin{equation} \label{equvwP}
u=−𝑦𝜗_t𝑧; \;\; v=𝑥𝜗_t𝑧; \;\; w=𝜗_t𝜓(𝑥,𝑦)
\end{equation}

where 𝜓 is the warping function representing the off-cross-section deformation on the bar cross-section. Notice in the circular cross-sectional bars, 𝜓 is assumed to be zero. In non-circular bars, it may not be zero and we may need to find the warping function. 

We have two unknowns $𝜗_t$, and 𝜓 that depends on $(x, y)$. We have converted the 3D problem to 2D. 

#### Shear strains in non-circular bars 

Using the small strain-displacement relations given in Chapter 4, and $𝜗_t$ is a constant, we can find the strains in the bar: 

\begin{equation} \label{eqEzxyn0}
\begin{split}
ε_{xx}&=\frac{∂ u}{∂ x}=0; \;\; ε_{yy}=\frac{∂ v}{∂ y}=0; \;\; ε_{zz}=\frac{∂ w}{∂ z}=0 \\
ε_{xy}&=\frac{1}{2}(\frac{∂ u}{∂ y}+ \frac{∂ v}{∂ x})=0 \\
2ε_{zx}&=𝛾_{zx}=\frac{∂ u}{∂ z}+ \frac{∂ w}{∂ x}=𝜗_t\bigg(\frac{∂ 𝜓}{∂ x} -y \bigg) \\
2ε_{zy}&=𝛾_{zy}=\frac{∂ w}{∂ y}+ \frac{∂ v}{∂ z}=𝜗_t\bigg(\frac{∂ 𝜓}{∂ y} +x \bigg)  
\end{split}
\end{equation}

The codes to obtain these equations are as follows. 

```python
x, y, z = symbols("x, y, z")             # define symbolic coordinates
𝜓 = sp.Function("𝜓")(x, y)
X = Matrix([x, y, z])
𝜗t = symbols("𝜗_t")                           # twist angle per length

U = Matrix([-y*𝜗t*z,x*𝜗t*z,𝜓*𝜗t]) # Assumed displacement vector[u,v,w]
U
```

```python
# Compute the strains (see Chapter 4): 
U_g = gr.grad_vf(U, X)                         # displacement gradient  
ε = (U_g + U_g.T)/2                          # strain tensor,symmetric 
2*ε                                                      # change to 𝛾
```

It is clear that the only nonzero strains are: 

\begin{equation} \label{eqEzxyn}
\boxed{
\begin{split}
𝛾_{zx} =𝜗_t\bigg(\frac{∂ 𝜓}{∂ x} -y \bigg) \\
𝛾_{zy} =𝜗_t\bigg(\frac{∂ 𝜓}{∂ y} +x \bigg)  
\end{split} }
\end{equation}

It is easy to confirm that the compatibility conditions given in Chapter 4 are satisfied. Note that if 𝜓 is zero, Eq.(\ref{eqEzxyn}) becomes Eq.(\ref{eqEzxy}). 

Differentiate $𝛾_{zx}$ with respect to $y$, and $𝛾_{zy}$ with respect to $x$, and then subtracting the results of these two, the warping function 𝜓 is eliminated, leading to

\begin{equation} \label{eqgzxy}
\boxed{
\frac{∂ 𝛾_{zx}}{∂ y} -\frac{∂ 𝛾_{zy}}{∂ x} = 2𝜗_t }
\end{equation}

This means that if $𝛾_{zx}$ and $𝛾_{zy}$ satisfies the forgoing equation, any function 𝜓 is admissible. Thus, we need not to worry about 𝜓 (if we do not want to find it out). Instead, the strains resulted from the stresses to-be-found must satisfy Eq.(\ref{eqgzxy}). 

#### Equilibrium equations at any point in torsion bar and stress function

Because only $𝛾_{zx}$ and $𝛾_{zy}$ are non-zero, the non-zero stresses are $σ_{zx}$ and $σ_{zy}$ (for isotropic materials). The equilibrium equations for 3D solids given in Chapter 6 are reduced to (without body force): 

\begin{equation} \label{eqEqtorn}
\frac{∂ σ_{zx}}{∂ z} =0; \;\;\; \frac{∂ σ_{zy}}{∂ z} =0 \\
\frac{∂ σ_{zy}}{∂ y}+\frac{∂ σ_{zx}}{∂ x} =0
\end{equation}

The first two equations mean that $σ_{zx}$ and $σ_{zy}$ are independent of $z$, and they shall satisfy the last equation in Eq.(\ref{eqEqtorn}). Such stresses can be written in terms of the so-called Prandtl **stress function** $𝜙(x, y)$ as:

\begin{equation} \label{eqsxyp}
σ_{zx}= \frac{∂ 𝜙(x, y)}{∂ y}; \;\;\; σ_{zy}= - \frac{∂ 𝜙(x, y)}{∂ x} 
\end{equation}

It is easy to confirm the last equation in Eq.(\ref{eqEqtorn}) is satisfied, by substituting Eq.(\ref{eqsxyp}) into it. 

The problem becomes finding the stress function $𝜙(x, y)$, subjected to proper boundary conditions and overall equilibrium on the cross-section of the bar. 

#### Boundary conditions for the stress function

Because the lateral surface $s$ of a torsion bar is free of stresses, the component of the shear stress $τ$ at any point on the cross-section by the surface S in its surface normal direction $\mathbf{n}$ must be zero. 

<img src="./images/PhiBC.png" alt=" " width="400"/><br>
\begin{figure}
  \includegraphics[width=5cm]{}
  \caption{\label{image:PhiBC} Stress components on the cross-section by the lateral surface S of a bar. They can be expressed in both the Cartesian and polar coordinates, and the coordinate transformation rule for vectors apply.}
\end{figure}

Considering an infinitely small curve segment $ds$ for a point P on the boundary of the cross-section shown Fig.\ref{image:PhiBC}. The corresponding infinitely small line segments along the Cartesian coordinates are denoted as $dx$ and $dy$, which cuts off $ds$ on the boundary. Based on the geometry of the infinitely small triangular cell on the boundary, we shall have on boundary S:    

\begin{equation} \label{eqsca}
\sin𝜃 = -\frac{dx}{ds}; \;\;\; \cos𝜃 =  \frac{dy}{ds}
\end{equation}

Notice the minus sign is due to the increase in $ds$ (counter-clockwise), results in decrease in $dx$. 

In order to obtain the shear stresses on the cross-section plane $z$ by boundary surface S, we can simply perform coordinate transformation. Because these shear stresses are all on the same cross-sectional surface-z, and the coordinate transformation is with respect to $z$-axis. This means that $(σ_{zx}, σ_{zy})$ and $ (σ_{zr}, σ_{z𝜃})$ carry the same first index $z$. The coordinate transformation rule for vectors (only one index is transfered) apply, which gives

\begin{equation} \label{eqtxyrt}
\begin{bmatrix} σ_{zr} \\ σ_{z𝜃} \end{bmatrix}
=  \underbrace{\left[\begin{matrix} \cos 𝜃 & \sin 𝜃 \\ -\sin 𝜃 & \cos 𝜃 \end{matrix}\right] }_{\mathbf{T}} \begin{bmatrix} σ_{zx} \\ σ_{zy} \end{bmatrix}
= \left[\begin{matrix}σ_{zx} \cos 𝜃 + σ_{zy} \sin 𝜃 \\-σ_{zx} \sin 𝜃 + σ_{zy} \cos 𝜃 \end{matrix}\right]
\end{equation} 

Here we used transformation matrix $\mathbf{T}$ for 2D, its formation is discussed in Chapter 2. Notice that this transformation angle is between the $x$-axis and the normal vector $\mathbf{n}$ at a point on the boundary surface S. 

We can also just follow the standard tensor transformation rule for stress tensor, given in Eq.(\ref{eqSTM3}). The code is as simple as follows. 

```python
# Define the stress tensor on the cross-section of the bar: 

σzx, σzy, 𝜃 = symbols("σ_zx, σ_zy, 𝜃")   # Define symbolic variables

S = Matrix([[ 0,  0, σzx],     # stress state for torsion w.r.t z-axis
            [ 0,  0, σzy],
            [σzx, σzy, 0]])

# Create sympy transformation matrix for rotation 𝜃, w.r.t. z:
T = gr.transferMts(𝜃)  

# Perform the 2nd order tensor transformation for the stresses: 
Sp=T@S@T.T                  
gr.simplifyM0(Sp)                     # simply the terms in the output
```

Compared the result with that in Eq.(\ref{eqtxyrt}), these two shear stresses $ (σ_{zr}, σ_{z𝜃})$ in the $(r,𝜃)$ coordinates are exactly the same.  

The first equation in Eq.(\ref{eqtxyrt}) gives, 

\begin{equation} \label{eqszxyBC}
\begin{split}
σ_{zr} = σ_{zx} \cos𝜃 + σ_{zy}\sin 𝜃 
\end{split}
\end{equation}

Because there is no shear stresses on the lateral surface $s$ of the bar, we must have $σ_{zr}=0$. Substituting Eq.(\ref{eqsca}) into Eq.(\ref{eqszxyBC}), leads to

\begin{equation} \label{eqssca}
σ_{zx} \frac{dy}{ds}-σ_{zy}\frac{dx}{ds}=0
\end{equation}

Next, substituting Eq.(\ref{eqsxyp}) into Eq.(\ref{eqssca}), we obtain:

\begin{equation} \label{eqdpBC}
\frac{∂ 𝜙}{∂ y} \frac{dy}{ds}+\frac{∂ 𝜙}{∂ x}\frac{dx}{ds}= \boxed{\frac{d 𝜙}{d s} = 0}
\end{equation}

Notice the change from partial derivatives of 𝜙 to total derivative. This means that 𝜙 must be a **constant** on the boundary of the cross-section of the bar. Further, because the stresses are given by partial derivatives of 𝜙, it can be set to zero on the boundary without affecting the stress solution. Therefore, the boundary condition for 𝜙 is found as,

\begin{equation} \label{eqpBC}
\boxed {𝜙 = 0} \;\;\; \text{ at any point on the boundary of the cross-section of a solid bar} 
\end{equation}

#### Shear stress on the cross-section by the boundary 

By the boundary, because $σ_{zr}=0$, the only nonzero stress components in the polar coordinates is $σ_{z𝜃}$, and is often denoted as $\tau$. It is given by the 2nd equation in Eq.(\ref{eqtxyrt}):  

\begin{equation} \label{eqtauBC}
\tau = σ_{z𝜃} = -σ_{zx} \sin 𝜃 + σ_{zy} \cos 𝜃  
\end{equation}

In other words, $σ_{zx}$ and $σ_{zy}$ are the two components of $\tau$, and we have

\begin{equation} \label{eqxytau}
σ_{zx}= -\tau \sin 𝜃 \\ σ_{zy} = \tau \cos 𝜃 
\end{equation}

When the foregoing equation is substituted to Eq.(\ref{eqtauBC}), justifies our argument. Reader may observe the foregoing equation together with Fig.\ref{image:PhiBC}. 

Next, substituting Eq.(\ref{eqsxyp}) into Eq.(\ref{eqtauBC}), we obtain

\begin{equation} \label{eqtaup}
\tau = -\frac{∂ 𝜙}{∂y} \sin 𝜃 - \frac{∂ 𝜙}{∂x} \cos 𝜃  = -\frac{∂ 𝜙}{∂r} 
\end{equation}

In obtaining Eq.(\ref{eqtaup}), we used coordinate transformation again, but for the gradient vector of 𝜙. 

Equation (\ref{eqtaup}) means that $\tau$ is the negative slope of the stress function 𝜙 with respect to $r$ (in the normal direction of bar's lateral surface S with normal $\mathbf{n}$). This finding is important when using the so-called membrane analogy for analyzing bars with complicated cross-section, which will be discussed later. Most importantly, it tell us that 

> the maximum absolute shear stress along the lateral surface is at location where normal slope of 𝜙 is the largest. 

#### Equilibrium conditions for the stress function

We found the expressions for the stresses on the cross-section (in terms of the stress function). The amplitude of the stresses must depend on the level of the torque applied. Therefore, we must examine the equilibrium conditions for the stresses obtained using the stress function using Eq.(\ref{eqsxyp}), in relation to the torque on the cross-section. These conditions are 

\begin{equation} \label{eqEqC}
\begin{split}
\sum F_x &= \int_A σ_{zx} dxdy = \int_A \frac{∂ 𝜙}{∂ y} dxdy= 0 \\
\sum F_y &= \int_A σ_{zy} dxdy = -\int_A \frac{∂ 𝜙}{∂ x} dxdy = 0 \\
\sum M_z &= \int_A (xσ_{zy}- yσ_{zx})dxdy = -\int_A (x\frac{∂ 𝜙}{∂ x}+y\frac{∂ 𝜙}{∂ y})dxdy= T
\end{split}
\end{equation}

where $T$ is the torque on the cross-section. We need to find the conditions on stress function 𝜙 so that the forgoing equilibrium conditions are satisfied. 

First, we evaluate the 2nd equation in Eq.(\ref{eqEqC}). Consider the strip $dy$ across the cross-section along $x$-direction, as shown in Fig.\ref{image:Int_dy}.

<img src="./images/Int_dy.png" alt=" " width="300"/><br>
\begin{figure}
  \includegraphics[width=5cm]{}
  \caption{\label{image:Int_dy} Integration along $x$-direction over $dy$ strip.}
\end{figure}

Because 𝜙 does not vary in the $y$-direction for this strip, we have

\begin{equation} \label{eqEqdy}
\begin{split}
-\int_A \frac{∂ 𝜙}{∂ x} dxdy &= -dy\int_A \frac{d 𝜙}{d x} dx= -dy\int_{𝜙(x_A)}^{𝜙(x_B)} d 𝜙 \\
&= -dy[𝜙(x_B)-𝜙(x_A)]=0
\end{split}
\end{equation}

This means that the sum of all the forces in the $y$ direction vanishes. The same can be done for the 1st equation in Eq.(\ref{eqEqC}), by considering a $dx$ strip. Readers may get this done on piece of paper. The first two equations in Eq.(\ref{eqEqC}) are all satisfied. This makes sense, because all the shear stresses on the cross-section can only be circling around, and the total divergence must be zero.  

Consider now the 3rd equation in Eq.(\ref{eqEqC}). 

\begin{equation} \label{eqEqM}
\begin{split}
\sum M_z &= -\int_A (x\frac{∂ 𝜙}{∂ x}+y\frac{∂ 𝜙}{∂ y})dxdy \\
&= -\int_A \begin{bmatrix} x & y \end{bmatrix} \begin{bmatrix} \frac{∂ 𝜙}{∂ x} \\ \frac{∂ 𝜙}{∂ y} \end{bmatrix} dxdy 
= -\int_A \begin{bmatrix} x & y \end{bmatrix} \cdot \nabla 𝜙 dxdy \\ 
& = -\int_s \underbrace{𝜙}_{\;\;=0, \text{ on } s } \begin{bmatrix} x & y \end{bmatrix} \cdot
\mathbf{n} \; dxdy + \int_A 𝜙 \; \nabla \cdot \begin{bmatrix} x \\ y \end{bmatrix} dxdy \\
&= 2\int_A 𝜙  dxdy =T
\end{split}
\end{equation}

In the 2nd to last line we used the Gauss divergence theorem. We finally found that the integral of the stress function over the area of the cross-section should equal to half of the torque on the cross-section. In other words, the volume under $𝜙(x,y)$ is half of the torque T: 

\begin{equation} \label{eqEqT}
\boxed{
 2\int_A 𝜙  dxdy =T }
\end{equation}

This is an important condition for 𝜙, because it relates directly to the torque applied. 

#### Linear elastic solution

Assume the bar is made of a linear elastic isotropic material. The stresses relate to the strains by the Hooke’s law: 

\begin{equation} \label{eqssG}
σ_{zx}= \frac{∂ 𝜙}{∂ y}=G𝛾_{zx} \\
σ_{zy}= -\frac{∂ 𝜙}{∂ x}=G𝛾_{zy} 
\end{equation}
or 
\begin{equation} \label{eqgG}
𝛾_{zx} = \frac{1}{G} \frac{∂ 𝜙}{∂ y}\\
𝛾_{zy} = -\frac{1}{G} \frac{∂ 𝜙}{∂ x} 
\end{equation}

Substituting Eq.(\ref{eqgG}) to Eq.(\ref{eqgzxy}), we obtain:

\begin{equation} \label{eqpLap}
\boxed{
\frac{∂^2 𝜙}{∂ x^2} + \frac{∂^2 𝜙}{∂ y^2} = -2G𝜗_t }
\end{equation}

This is a Laplace equation. If the twist angle $𝜗_t$ is given and 𝜙 satisfies BC Eq.(\ref{eqpBC}), then Eq.(\ref{eqpLap}) uniquely determines $𝜙(x, y)$. The stresses can be found via Eq.(\ref{eqssG}), and the torque by Eq(\ref{eqEqT}). We thus have a logical and workable path to the solution. 

#### Techniques for constructing the stress function

Because 𝜙 must vanish on the boundary of the cross-section, we can make use of the equation that describes the boundary of the cross-section to construct 𝜙. Assuming the boundary of the cross-section is given by $f(x,y)=0$, 𝜙 can be written as

\begin{equation} \label{eqpBEq}
𝜙(x,y)= C f(x,y)
\end{equation}

where $C$ is a constant. If $f(x,y)$ satisfies

\begin{equation} \label{eqfLap}
\frac{∂^2 f}{∂ x^2} + \frac{∂^2 f}{∂ y^2} = \text{constant} 
\end{equation}

we can then adjust $C$ to have Eq.(\ref{eqpLap}) be satisfied. 

#### Example: Bars with elliptical cross-section

Consider a bar with elliptical cross-section, as shown in Fig.\ref{image:Ellipse}.

<img src="./images/Ellipse.png" alt=" " width="300"/><br>
\begin{figure}
  \includegraphics[width=5cm]{}
  \caption{\label{image:Ellipse} Elliptical cross-section for a torsional bar.}
\end{figure}. 

Stress function 𝜙 can be written as:
 
\begin{equation} \label{eqpelli}
𝜙 = C \bigg( \frac{x^2 }{a^2} + \frac{y^2 }{b^2} -1 \bigg) 
\end{equation}

Substituting it to Eq.(\ref{eqpLap}), we found

\begin{equation} \label{eqB}
C = -  \frac{a^2 b^2G𝜗_t}{a^2+b^2} 
\end{equation}

The shear stresses can be found via Eq.(\ref{eqssG}):

\begin{equation} \label{eqselli}
σ_{zx} =\frac{∂ 𝜙}{∂ y} = \frac{2Cy}{b^2}=-\frac{2a^2 G𝜗_ty}{a^2+b^2}\\
σ_{zy} =-\frac{∂ 𝜙}{∂ x} = -\frac{2Cx}{a^2}=\frac{2b^2 G𝜗_tx}{a^2+b^2}
\end{equation} 

The maximum shear stress is found as

\begin{equation} \label{eqsellimax}
\tau _{max}  = σ_{zy} \bigg|_{y=b}= \frac{2a^2 b G𝜗_t}{a^2+b^2} 
\end{equation}

Torque $T$ is obtained by substituting Eq.(\ref{eqpelli}) into Eq.(\ref{eqEqT}): 

\begin{equation} \label{eqelliT}
 T = 2\int_A 𝜙 dA = \frac{2C}{a^2} \int_A x^2 dA + \frac{2C}{b^2} \int_A y^2 dA - 2C \int_A dA \\
 =\frac{2C}{a^2} I_{yy}  + \frac{2C}{b^2} I_{xx} - 2CA 
\end{equation}

where $A$ is the area, $I_{xx}$, $I_{yy}$ and the 2nd moment of area with respect to $x$- and $y$-axis of the cross-section of the bar. For an ellipse, the formulas are as follows.

\begin{equation} \label{eqElipIA}
I_{xx} = \frac{πab^3}{4 }; \;\;  I_{yy} = \frac{πba^3}{4 };  \;\; A = π ab
\end{equation}

Using these formulas given above, we have

\begin{equation} \label{eqElipT}
T = -πC ab
\end{equation}

The torque may be expressed either in terms of $τ_{max}$ or $𝜗_t$ by means of Eqs.(\ref{eqB}), (\ref{eqsellimax}) and (\ref{eqElipT}).

\begin{equation} \label{eqselimax}
\tau _{max}  =  \frac{2T}{πab^2} 
\end{equation} 

\begin{equation} \label{eqsellit}
𝜗_t = \frac{T(a^2+b^2)}{πGb^3a^3} = \frac{T}{GJ} 
\end{equation}

where $GJ$ is the torsional stiffness per unit length, with $J$ given by

\begin{equation} \label{eqelliJ}
\boxed{
J = \frac{πb^3a^3}{a^2+b^2} }
\end{equation}

This is the equivalent polar moment of area of ellipses. When $a=b$, Eq.(\ref{eqelliJ}) is reduced to Eq.(\ref{eqJ}) for the circular cross-section. 

#### Comparison of torsional stiffness between elliptic and circular bars

Let us conduct an analysis on torsional stiffness of elliptic and circular solid bars made of the same material. Assuming the radius of the circular cross-section is $R$, we shall set $R^2=ab$ two ensure their areas are the same. The stiffness ratio of these to bars becomes. 

\begin{equation} \label{eqastiff}
𝛼_{\text{stiffness}} = \frac{J_{\text{ellipse}}}{J_{\text{circle}}}\bigg|_{R^2=ab} = \frac{\frac{πb^3a^3}{a^2+b^2}}{\frac{πR^4}{2}}\bigg|_{R^2=ab} = \frac{\frac{b^3a^3}{a^2+b^2}}{\frac{(ab)^2}{2}} = \frac{2ab}{a^2+b^2} \le 1 
\end{equation}

In the last step, we used inequality $a^2+b^2\ge 2ab$ (because $(a-b)^2\ge 0$). This means that the stiffness of an elliptic bar is always smaller that the circular bar with the same area. They are equal only when $a=b$.  

#### Example: Bars with equilateral triangle cross-section 

Consider a bar with equilateral triangle cross-section, as shown in Fig.\ref{image:T3a}. It is subjected to torque $T$. 

<img src="./images/T3a.png" alt=" " width="300"/><br>
\begin{figure}
  \includegraphics[width=5cm]{}
  \caption{\label{image:T3a} Equilateral triangle cross-section for a torsional bar.}
\end{figure}

Stress function 𝜙 can be written as:
 
\begin{equation} \label{eqpt3}
𝜙 = \frac{G𝜗_t}{2h} \bigg(x-\sqrt{3}y-\frac{2h}{3}\bigg) \bigg(x-\sqrt{3}y+\frac{2h}{3}\bigg) \bigg(x+\frac{h}{3}\bigg) 
\end{equation}

Following the same process for the ellipse bar, the maximum shear stress is found at the center of the edges as

\begin{equation} \label{eqst3max}
\tau _{max}  = \frac{15\sqrt{3}T}{2h^3} = \frac{20T}{a^3} 
\end{equation}

The twist angle per unit length is found as 

\begin{equation} \label{eqt3t}
𝜗_t = \frac{15\sqrt{3}T}{Gh^4} = \frac{T}{GJ} 
\end{equation}

where $GJ$ is the torsional stiffness per unit length for the equilateral triangle cross-section, with $J$ being the equivalent polar 2nd moment of area given by

\begin{equation} \label{eqt3J}
J = \frac{h^4}{15\sqrt{3}} = \frac{\sqrt{3}a^4}{80}
\end{equation}

For cross-sections of more complex shape, analytical solutions are difficult to obtain. Approximate solutions may be obtained by Prandtl’s membrane analogy for some of those.   

### Prandtl's elastic membrane analogy for torsion *

#### General formulation

Considering a membrane subjected to pressure $p$, resulting in a uniform membrane tension $S$, and undergoes a small deflection, as shown in Fig.\ref{image:Membrane_p}. It is fixed on its boundary $s$. Our formulation and notation follow largely the textbook [4]. Codes are developed for computations and result plots.

<img src="./images/Membrane_p.png" alt=" " width="350"/><br>
\begin{figure}
  \includegraphics[width=5cm]{}
  \caption{\label{image:Membrane_p} A membrane subjected to pressure and undergoes a small deflection. A infinitely small cell at a point is taken to examine the equilibrium state. Top: top-down view. Bottom: side section view.}
\end{figure}

A infinitely small cell of $dxdy$ at a point is taken to examine the equilibrium state. Assume the membrane undergoes a tension stress $S$ and its deflection is denoted as $w$. 

On edge A-D, the force in the membrane is 

\begin{equation} \label{eqfAD}
-S dy \sin\alpha ≈  -S dy \tan\alpha =  -S dy \frac{∂w}{∂x}
\end{equation}

On edge B-C, the force is 

\begin{equation} \label{eqfBC}
S dy \tan \bigg(𝛼 +\frac{∂𝛼}{∂x}dx \bigg)= S dy \frac{∂}{∂x}\bigg(w+\frac{∂w}{∂x}dx\bigg)
\end{equation}

Similarly, on edge A-B, the force in the membrane is 

\begin{equation} \label{eqfDC}
-S dx \frac{∂w}{∂y}
\end{equation}

On edge D-C, it is 

\begin{equation} \label{eqfAB}
S dx \frac{∂}{∂y}\bigg(w+\frac{∂w}{∂y}dy\bigg)
\end{equation} 

Sum of all these four forces with the pressure should be zero, and dividing $dxdy$ for all terms, we obtain the equilibrium equation in terms of $w$: 

\begin{equation} \label{eqmEq}
\boxed{
\frac{∂^2w}{∂x^2}+\frac{∂^2w}{∂y^2} = - \frac{p}{S} }
\end{equation} 


This equation has the same form as Eq.(\ref{eqpLap}) for the stress function. They both are Laplace equation, but with different constant on the right-hand-side. Therefore, the membrane deflection and the stress function differ in a constant factor $c$: 

\begin{equation} \label{eqwp}
w = c 𝜙 
\end{equation}

where $c$ is found by equaling terms in the right-hand-side of Eq.(\ref{eqmEq}) Eq.(\ref{eqpLap}), which is

\begin{equation} \label{eqcwp}
c =  \frac{p}{2G𝜗_tS} 
\end{equation}

Finding the stress function for torsion bar becomes solving for the deflection of a pressured membrane with a shape same as the cross-section of the bar.  

Examples are given in the following sections. 

#### A bar with narrow rectangular cross-section *

Consider a bar with a narrow rectangular cross-section subjected to a torque $\mathbf{T}$, as shown in Fig.\ref{image:N_rectangleT}. The torque will create a force couple on the cross-section that results in shear stress distributed there. Because of its narrow shape, it is expected that the shear stress will form two major flows: one near the top surface and another near the bottom surface, with mutually opposite direction. We shall find a way to quantify it. 

<img src="./images/N_rectangleT.png" alt=" " width="350"/><br>
\begin{figure}
  \includegraphics[width=5cm]{}
  \caption{\label{image:N_rectangleT} A bar with narrow rectangular cross-section subjected to a torque.}
\end{figure} 

Using the membrane analogy, we consider a membrane with narrow rectangular shape and subjected to pressure, as shown in Fig.\ref{image:N_rectangle}.  

<img src="./images/N_rectangle.png" alt=" " width="350"/><br>
\begin{figure}
  \includegraphics[width=5cm]{}
  \caption{\label{image:N_rectangle} Narrow rectangular cross-section of a bar subjected to a pressure and undergoes a small deflection. It has a width of $2b$ and height of $2h$. }
\end{figure} 

Based slender geometry of the deflected membrane, we can assume membrane deflection is independent of $x$, and varies in a parabolic way with respect to $y$. 

\begin{equation} \label{eqww0}
w=w_0 \left( 1−\frac{y^2}{h^2} \right) 
\end{equation}

where $w_0$ is the maximum deflection at $y=0$. This deflection satisfies the boundary condition on $y=\pm h$. Except for the small regions near $x=±b$. Hence, this is a good approximation. The approximation error reduces with decreasing $h/b$. The Laplace of the assumed $w$ becomes,

\begin{equation} \label{eqnrEq}
 \frac{∂^2w}{∂y^2} = - \frac{w_0}{h^2}
\end{equation} 

It is now a 1D problem. The membrane analogy leads to $w_0 /h^2 = 2cG 𝜗_t$. The stress function is found as 

\begin{equation} \label{eqnrp}
𝜙 =G 𝜗_t h^2 \left( 1−\frac{y^2}{h^2} \right) 
\end{equation}

The shear stresses can be found via Eq.(\ref{eqssG}):

\begin{equation} \label{eqsnrect}
σ_{zx} =\frac{∂ 𝜙}{∂ y} = - 2G 𝜗_ty\\
σ_{zy} =-\frac{∂ 𝜙}{∂ x} =0
\end{equation}

It is clear that $σ_{zx}$ is linearly in $y$ and anti-symmetric with respect to $y=0$. 

A actual distribution of the shear stress on a narrow rectangular cross-section is give in Fig.\ref{image:shearOnNRect}. This is produced using our code to be introduced later. 

<img src="./images/shearOnNRect.png" alt=" " width="600"/><br>
\begin{figure}
  \includegraphics[width=5cm]{}
  \caption{\label{image:shearOnNRect} A actual case of shear stress on narrow rectangular cross-section of a bar subjected to torsion ($b=10h$). }
\end{figure} 

As shown, the shear stress flows near the top and bottom surfaces forms a force couple giving the torque on the cross-section of the bar, as shown in Fig.\ref{image:N_rectangleT}. The resulted two forces are located about $h/3$ away from the top and bottom surfaces. 

The maximum shear stress is found at $y=\pm h$:

\begin{equation} \label{eqsnrecmax}
\tau _{max}  = 2G 𝜗_t h
\end{equation}

Torque $T$ is obtained using Eqs.(\ref{eqEqT}) and (\ref{eqnrp}): 

\begin{equation} \label{eqnrectT}
 T = 2\int_A 𝜙 dA = 2\int_{-b}^b \int_{-h}^h  𝜙 dxdy  =G 𝜗_t \underbrace{\tfrac{1}{3}(2b)(2h)^3}_{J} =GJ𝜗_t
\end{equation}

where $J$ is the equivalent polar 2nd moment of area given by 

\begin{equation} \label{eqnrectJ}
\boxed{
 J=\frac{1}{3}(2b)(2h)^3 }
\end{equation}

> Note that Eq.(\ref{eqnrectJ}) can be used for any cross-section that has a narrow channel shape, and it can be curved. This is because of the membrane analogy, the deformed membrane will have the same volume regardless whether it curved or not, so long it is a narrow channel. Therefore, Eq.(\ref{eqnrectJ}) is very useful to calculate the stiffness per length $GJ$ for such bars. 

Using Eq.(\ref{eqsnrecmax}) and (\ref{eqnrectT}), the maximum shear stress can now be written in terms of $T$:

\begin{equation} \label{eqsnrecmaxT}
\tau _{max}  = \frac{3T}{(2b)(2h)^2} = \frac{2Th}{J} 
\end{equation}

Finally, the twist angle per length can be written for given $T$: 

\begin{equation} \label{eqtnrecmaxT}
𝜗_t = \frac{3T}{G(2b)(2h)^3} = \frac{T}{GJ} 
\end{equation}

We mention that the boundary conditions are not satisfied on $x=±b$. The error is expected small, if $h/b$ is small.  

#### Bars with cross-sections consists of narrow rectangles *

Many torsion bars are with rolled composite sections consisting of a number of narrow rectangles. The most typical one is the I-beams, as shown in Fig.\ref{image:I-BeamCrossSection}. This type of bars analyzed easily making use of the results from the previous section. 

<img src="./images/I-BeamCrossSection.png" alt=" " width="480"/><br>
\begin{figure}
  \includegraphics[width=5cm]{}
  \caption{\label{image:I-BeamCrossSection} A typical I-beam with three narrow rectangles. The image on the left is from Poutrelle métallique at wikipedia commons under the CC BY 3.0 license.}
\end{figure}

Assuming the cross-section of a bar is formed with $n$ narrow channels, its $J$ value can be obtained simply by 

\begin{equation} \label{eqnJ}
 J=c\frac{1}{3}\sum_{i=1}^n(2b_i)(2h_i)^3
\end{equation}

This addition is justified by, again, the membrane analogy, because the total volume of the deflection (hence the stress function) for a composite section with narrow channels are just simple addition of the individual channels. Each channel acts alone having a stress distribution like that shown in Fig.\ref{image:shearOnNRect}, forming an individual torque. The sum of the torques of all the channels of the cross-section gives the total torque $T$. 

In Eq.(\ref{eqnJ}), we introduced a correction coefficient $c$. It can be used for proper adjustment when a rectangle is not narrow. If $b_i>10h_i$ for a rectangular part, then $c≈1$. If $b_i<10h_i$, $c$ may be reduced to, say, $c≈0.9$. It is a rough estimation. As will be seen in the next section with computer codes, we can use more precise formula. 

Note also that when computing the maximum shear stress in a composite section, we shall use

\begin{equation} \label{eqsnmaxT}
\tau_{max}  = \frac{2Th_{max}}{J} 
\end{equation}

where $h_{max}$ is $\max_{i}(h_i)$. This is because wider channel has larger membrane deflection and hence larger $𝜙$ values. 

### Bars with rectangular cross-section*

#### Strategy for solution 

Consider torsion bars with the rectangular section of width $2h$ and depth $2b$, as shown in Fig.\ref{image:Rctangle_hb}. This types of bars require more intensive analysis for proper design against torsion. Our formulation and notation follow largely the textbook \cite{boresi2002advanced}. Codes are developed for computations and result plots.  

<img src="./images/Rctangle_hb.png" alt=" " width="200"/><br>
\begin{figure}
  \includegraphics[width=5cm]{}
  \caption{\label{image:Rctangle_hb} A typical rectangular cross-section for bars subjected to torsion.}
\end{figure}

We need to find the stress function 𝜙, and then solve the torsion problem for stresses and twist angle. Because of the symmetry of the cross-section and the Laplace operator is also symmetric, 𝜙 must be an even function of both $x$ and $y$. The problem can be mathematically defined by

\begin{equation} \label{eqpRect}
\begin{split}
\nabla^2 𝜙 &=  -2G 𝜗_t \;\;\; &\text{  PDE in the domain of the cross-section} \\
 𝜙 &= 0  \;\;\; &\text{ on the boundary of the cross-section} 
\end{split}
\end{equation}

One effective approach to solve this problem is to use the method of separation of variables. To do so, we first make the Laplace PDE homogeneous (with zero right-hand-side). This is done by writing 𝜙 in the following form, assuming $h\le b$. 

\begin{equation} \label{eqpGw}
𝜙 =G 𝜗_t (h^2-x^2) + w(x, y) 
\end{equation}

where $w$ is an unknown function to be determined. It is easy to check that the first term in the foregoing equation satisfies the PDE in Eq.(\ref{eqpRect}). It is thus a particular solution to Eq.(\ref{eqpRect}). Now, substituting Eq.(\ref{eqpGw}) into Eq.(\ref{eqpRect}) gives, 

\begin{equation} \label{eqwRect}
\begin{split}
\nabla^2 w &=0 \;\;\; \text{ governing PDE} \\
w &=
    \begin{cases}
        0               & \text{ for } x=\pm h\\
        G 𝜗_t (h^2-x^2) & \text{ for } y=\pm b
    \end{cases}
\end{split}
\end{equation}

We need next to solve the foregoing equation for $w(x,y)$, using the method of separation of variables. Note if assuming $h>b$, we would use $G 𝜗_t (b^2-y^2)$ instead as the particular solution to Eq.(\ref{eqpRect}). This is just for a better convergence in the solution to be obtained later. It is not a really big issue. 

#### Method of separation of variables

We write the unknown function as the product of two unknown functions $f(x)$ and $g(x)$: 

\begin{equation} \label{eqwfg}
w(x,y) = f(x)g(y)
\end{equation}

Substitute Eq.(\ref{eqwfg}) into  Eq.(\ref{eqwRect}), we obtain,

\begin{equation} \label{eqwfg2}
\nabla^2 w = g(y)f''(x) + f(x)g''(y) =0
\end{equation}

which can be rewritten as 

\begin{equation} \label{eqwfg2l}
\frac{f''(x)}{f(x)} = -\frac{g''(y)}{g(x)}=-𝜆^2
\end{equation}

The first term is a function of  only $x$ and the second is a function of  only $y$. Thus, 𝜆 can only be a (unknown) constant independent of both $x$ and $y$. Equation (\ref{eqwfg2l}) gives two separate equations: 

\begin{equation} \label{eqfg22}
f''(x) + 𝜆^2 f(x) =0 \\
g''(y) - 𝜆^2 g(y) =0
\end{equation}

Each of these equations is a 1D differential equation (DE). The solutions to them are, respectively, 

\begin{equation} \label{eqfg2s}
f(x) =A \cos𝜆x + B\sin𝜆x \\
g(x) =C \cosh𝜆y +D\sinh𝜆y 
\end{equation}

The solutions contain four constants. Because the original stress function 𝜙 must be an even function, both $f(x)$ and $g(x)$ must be even. Thus, $B=D=0$. Substitute Eq.(\ref{eqfg2s}) into  Eq.(\ref{eqwfg}), we obtain,

\begin{equation} \label{eqwA}
w(x,y) = A \cos𝜆x  \cosh𝜆y
\end{equation}

where $A$ (absorbed $C$) is a to-be-determined constant, in addition to constant 𝜆. This is done using boundary conditions. 

To satisfy the boundary condition at $x=\pm h$, we set 

\begin{equation} \label{eqlmd}
𝜆 = \frac{n\pi}{2h}, \;\; n =1, 3, 5,...
\end{equation}

Equation (\ref{eqwA}) becomes,

\begin{equation} \label{eqwAn}
w(x,y) = \sum_{n =1, 3, 5,...} A_n \cos\frac{n\pi x}{2h} \cosh\frac{n\pi y}{2h}
\end{equation}

Using the boundary condition at $y=\pm b$, we have

\begin{equation} \label{eqwAnG}
\sum_{n =1, 3, 5,...} A_n \cos\frac{n\pi x}{2h} \cosh\frac{n\pi b}{2h}=G 𝜗_t (h^2-x^2)
\end{equation}

To solve the foregoing equation for $A_n$, we multiply $\cos\frac{n\pi x}{2h}$ on both sides and integrate along $x$ from $–h$ to $h$, and use formulas for cosine function, which gives, 

\begin{equation} \label{eqwAnI}
A_n \cosh\frac{n\pi b}{2h}=\frac{G 𝜗_t}{h} \int_{-h}^h (x^2-h^2) \cos\frac{n\pi x}{2h} dx = \frac{-32G 𝜗_t h^2 (-1)^{(n-1)/2)}}{n^3\pi^3}  
\end{equation}

These above-integrals can be done using the following code. 

```python
x, h = symbols("x, h")
n = symbols('n', integer=True, odd=True)

# The integral = on the left-hand-side
integrate(((sp.cos(n*sp.pi*x/(2*h))**2)), (x, -h, h))   
```

```python
# The integral = on the right-hand-side
integrate((x**2-h**2)*sp.cos(n*sp.pi*x/(2*h)), (x, -h,h))
```

Finally, $A_n$ is found as 

\begin{equation} \label{eqAn}
A_n = \frac{-32G 𝜗_t h^2 (-1)^{(n-1)/2)}}{n^3\pi^3 \cosh\frac{n\pi b}{2h}}, \;\; n=\text{ odd}
\end{equation}

#### Final solution of the stress function

The stress function becomes, 

\begin{equation} \label{eqpF}
𝜙 =G 𝜗_t (h^2-x^2)- \frac{32G 𝜗_t h^2}{\pi^3} \sum_{n =1, 3, 5,...} \frac{(-1)^{(n-1)/2}}{n^3 \cosh\frac{n\pi b}{2h}} \cos\frac{n\pi x}{2h} \cosh\frac{n\pi y}{2h}
\end{equation}

Note that in the denominator of the coefficient of the series above, there is $n^3$ term. Also, the cosh function has the following Taylor expansion: 

\begin{equation} \label{eqcosh}
cosh x = 1 + \frac{x^2}{2!} + \frac{x^4}{4!} + …
\end{equation}

Therefore, the coefficients in Eq.(\ref{eqpF}) goes to zero quite fast when $b/h$ is large. When $b/h$ goes to infinity, Eq.(\ref{eqpF}) reduces to the one for narrow rectangles discussed in the previous section. 

#### Shear stress solution

Using Eq.(\ref{eqsxyp}), the shear stresses can be found as follows. 


\begin{equation} \label{eqsxRct}
σ_{zx}= \frac{∂ 𝜙}{∂ y}=-\frac{16G 𝜗_t h}{\pi^2} \sum_{n =1, 3, 5,...} \frac{(-1)^{(n-1)/2}}{n^2 \cosh\frac{n\pi b}{2h}} \cos\frac{n\pi x}{2h} \sinh\frac{n\pi y}{2h}
\end{equation}

\begin{equation} \label{eqsyRct}
σ_{zy}=-\frac{∂ 𝜙}{∂ x}=2G 𝜗_t x-\frac{16G 𝜗_t h}{\pi^2} \sum_{n =1, 3, 5,...} \frac{(-1)^{(n-1)/2}}{n^2 \cosh\frac{n\pi b}{2h}} \sin\frac{n\pi x}{2h} \cosh\frac{n\pi y}{2h}
\end{equation}

#### Torque twist-angle solution and  equivalent polar 2nd moment of area

The torque is found using Eq.(\ref{eqEqT}) as 

\begin{equation} \label{eqTRct}
T = 2\int_{-b}^b \int_{-h}^h 𝜙 dxdy = GJ𝜗_t
\end{equation}

where 

\begin{equation} \label{eqJRct}
J=  (2b)(2h)^3 \underbrace{\left[\frac{1}{3}- \frac{192}{3\pi^5}\frac{h}{b} \sum_{n =1, 3, 5,...} \frac{1}{n^5} \tanh \bigg(\frac{n\pi}{2}\frac{b}{h}\bigg)\right]}_{k_1} =  (2b)(2h)^3 k_1
\end{equation}

where $k_1$ is a coefficient that depends only on $b/h$. When $b/h \to ∞$, $k_1 \to \frac{1}{3}$, because summation term in the foregoing equation approaches to zero. 

#### The maximum shear stress

The maximum slope of the stress function (membrane) on the rectangle section occurs at $x = ±h, y = 0$, where $σ_{zx}= 0$ and $σ_{zy}$ has the maximum value $(h \le b)$:

\begin{equation} \label{eqsymRct}
\tau_{max} = σ_{zy}\bigg|_{x=h\\y=0} =2G 𝜗_t h \underbrace{\left[1-\frac{8}{\pi^2} \sum_{n =1, 3, 5,...} \frac{1}{n^2 \cosh\frac{n\pi b}{2h}} \right]}_{k}    = 2G 𝜗_t h k
\end{equation}

If we introduce $k_2 = \frac{k_1}{k}$, the maximum shear stress can be computed alternatively using 

\begin{equation} \label{eqtmaxk2}
\tau_{max} = \frac{T}{k_2(2b)(2h)^2)} 
\end{equation}

Coefficient $k_1$ and $k$ can be computed using the code given below. Tabulated $k_1$ and $k_2$ are given in \cite{boresi2002advanced}. 

#### Python code for computing $k_1$

```python
# Define the function for computing k1: 
def k1_f(b_h, n_end):
    sum1 = 0.0
    for n in range(1, n_end, 2):
        sum1 += np.tanh(n*np.pi*b_h/2)/n**5
    return 1.0/3.0 - 192.0/(3*b_h*np.pi**5)*sum1

# Define the function for computing k: 
def k_f(b_h, n_end):
    sum1 = 0.0
    for n in range(1, n_end, 2):
        sum1 += 1./(np.cosh(n*np.pi*b_h/2))/n**2
    return 1.0 - 8.0/(np.pi**2)*sum1
```

J= (2h)^4 k_{1(b=h)} = 0.140577 (2h)^4
\end{equation}

Here we used $k_{1(b=h)} = 0.140577$ obtained earlier. Compared to the polar moment of area of a square bar, using perpendicular axis theorem (see, Chapter 10), $J_z = 2 I_{xx}=\frac{(2h)^4}{6}$, the corresponding $k_1$ should be calculated using,

k_1^{square}= \frac{1}{6} = 0.166667

\end{equation}

For the same infinitely long channel bar subjected to torsion, we have $k_1=1/3$, which is 4 times $k_1^{narrow}$.

#### Comparison of torsional stiffness  between rectangular and square bars

Let us conduct an analysis on torsional stiffness of rectangular bar of $(2b)\times (2h)$ ($b>h$) and square bars of $(2a)\times (2a)$. They both are solid and made of the same material. We set $(2a)^2=(2b)(2h)$ or $ a^2=bh$ to ensure their areas are the same. The stiffness ratio of these two bars becomes. 

\begin{equation} \label{eqars}
\begin{split}
𝛼_{\text{stiffness}} &= \frac{J_{\text{rectangle}}}{J_{\text{square}}}\bigg|_{a^2=bh} = \frac{(2b)(2h)^3k_1}{(2a)^4k_{1(b=h)}}\bigg|_{a^2=bh} \\
&= \frac{bh^3k_1}{(bh)^2k_{1(b=h)}} = \frac{h}{b} \frac{k_1}{k_{1(b=h)}} 
\end{split}
\end{equation}

To find out how $𝛼_{\text{stiffness}}$ changes with $b/h$, we use the following code to plot the curve. 

```python
n_end = 25; dx = 0.01
b_h  = np.arange(1.0, 10.0+dx, dx)                     # b/h as x_data

X_data = [b_h]                      # put the x (b/h) values to a list
k1_data = []
for i in range(len(b_h)):
    k1_data.append(k1_f(b_h[i], n_end))              # function values 
    
𝛼_stiffness = [k1_data[i]/b_h[i]/k1_data[0] for i in range(len(b_h))]
Y_data = [𝛼_stiffness]                # put the function values to a list

print(f"b/h={b_h[-1]}, 𝛼_stiffness={𝛼_stiffness[-1]}")
labels=['$𝛼$_stiffness']                           # define the label
gr.plot_fs(X_data, Y_data, labels, 'b/h',   '$𝛼$_stiffness', 'a_stiff')
                                  # x-label,   y-label,       file-name
```

It is clear that $𝛼_{\text{stiffness}}<1$ for $b>h$. This means that the torsional stiffness of a rectangular bar is smaller than that of the square bar with the same area. They are equal when $h=b$. When $b/h=10$, the stiffness reduces to only about 22%. This significant reduction can be explained using Fig.\ref{image:N_rectangleT}. When the height is very small, the distance between these force couple (one near the top and another near the bottom) of the rectangular channel becomes small. In addition, for the same twist angle per bar length, the shear stress on both top and bottom surface is also small. The double reduction effect results in significantly reduction in torque, and hence the stiffness. 

This finding is similar to that given Section 9.3.11, when comparing the narrow elliptic and circular bars. 

#### Comparison of torsional stiffness between circular and square bars

Let us also compare the torsional stiffness of a circular bar of radius $R$ with a square bars of $(2a)\times (2a)$. Assume both bars are solid and made of the same material. We set $\pi R^2=(2a)(2a)$ or $R^2=\frac{4}{\pi}a^2$ to ensure their areas are the same. The stiffness ratio of these two bars becomes: 

\begin{equation} \label{eqacs}
\begin{split}
𝛼_{\text{stiffness}} &= \frac{J_{\text{circular}}}{J_{\text{square}}}\bigg|_{R^2=\frac{4}{\pi}a^2} 
= \frac{\pi R^4/2}{(2a)^4k_{1(b=h)}}\bigg|_{R^2=\frac{4}{\pi}a^2} \\
&= \frac{\pi (\frac{4}{\pi}a^2)^2}{2(2a)^4 k_{1(b=h)}} 
= \frac{1}{2\pi k_{1(b=h)}} = \frac{1}{2\pi \times 0.140577} = 1.132
\end{split}
\end{equation}

This means that the torsional stiffness of a circular bar is about 13.2% higher than that of the square bar with the same area. 

#### Code to compute the stress function on a rectangular cross-section

We write the following code to compute the stress function for torsion bar with rectangular cross-section, using Eq.(\ref{eqpF}). 

It computes effectively a series solution for any function governed by a 2D Laplace PDE with a constant right-hand-side, and fixed boundary condition. 

```python
def sf_phi(b, h, Gt, x, y, n_end = 21):
    '''Compute the stress function phi for torsion bar with rectangular
    cross-section. It is effectively a series solution to 2D Laplace PDE
    with a constant right-hand-side, and fixed boundary condition. 
    b:       half-height of the domain
    h:       half-width  of the domain
    Gt:      constant at the right-hand-side of the Laplace.
    n_end:   terms used in the series of the solution
    x and y: point at which the solution to be found.'''
    
    sum1 = 0.0
    for n in range(1, n_end, 2):
        sum1 += (-1)**((n-1)/2)/(n**3*np.cosh(n*np.pi*b/(2*h)))*\
                 np.cos(n*np.pi*x/(2*h))*np.cosh(n*np.pi*y/(2*h))
            
    return Gt*(h**2-x**2) - (32*Gt*h**2/np.pi**3)*sum1
```

#### Code to compute the shear stresses on a rectangular cross-section

We write the following code to compute the shear stresses on a rectangular cross-section bar subjected to torsion, using Eqs.(\ref{eqsxRct}) and (\ref{eqsyRct}).  

It computes effectively a series solution for the derivatives of the function governed by any 2D Laplace PDE with a constant right-hand-side, and fixed boundary condition. 

```python
def stress_rect(b, h, Gt, x, y, n_end = 21):
    '''compute the shear stresses on a rectangular cross-section 
    bar subjected to torsion. It is effectively a series solution for 
    the derivatives of the function governed by any 2D Laplace PDE 
    with a constant right-hand-side, and fixed boundary condition. 
    b:       half-height of the domain
    h:       half-width  of the domain
    Gt:      constant at the right-hand-side of the Laplace.
    n_end:   terms used in the series of the solution
    x and y: point at which the solution to be found.
    
    Return: sigam_zx, sigam_zy
    '''
    factor = 16*Gt*h/np.pi**2
    sum_x = 0.0;    sum_y = 0.0
    for n in range(1, n_end, 2):
        coef = (-1)**((n-1)/2)/(n**2*np.cosh(n*np.pi*b/(2*h))) 
        sum_x += coef*np.cos(n*np.pi*x/(2*h))*np.sinh(n*np.pi*y/(2*h))
        sum_y += coef*np.sin(n*np.pi*x/(2*h))*np.cosh(n*np.pi*y/(2*h))
 
    sigam_zx =        - factor*sum_x
    sigam_zy = 2*Gt*x - factor*sum_y 
    
    return sigam_zx, sigam_zy
```

#### Plot of the shear stresses on a rectangular cross-section

```python
from matplotlib import cm
plt.rcParams['figure.dpi'] = 500  
h, b, Gt, n_end = 2., 1., 1., 21              # arbitrarily set values

xl, xr, dx = -h, h, 0.2
yl, yr, dy = -b, b, 0.1

X = np.arange(xl, xr+dx, dx)
Y = np.arange(yl, yr+dy, dy)
X, Y = np.meshgrid(X, Y)
xn, yn = X.shape
Szx = X*0
Szy = Y*0

for xk in range(xn):
    for yk in range(yn):
        Szx[xk, yk], Szy[xk, yk] = stress_rect(b,h,Gt,X[xk,yk],Y[xk,yk])

magnitude = np.sqrt(Szx**2 + Szy**2)
plt.quiver(X, Y, Szx, Szy, magnitude, scale=20, cmap=cm.jet)#'viridis')
plt.axis('equal')  #plt.colorbar()
plt.savefig('images/shearOnRect.png', dpi=500)    # save plot to a file
plt.show()
```

It is seen that the shear stress vector is circulating on the cross-section (the $x-y$ plan) of the bar. The total circulation is the curl of the stress vector integrated over the cross-section, relating in the torque with respect to the $z$-axis. Because of the circulating nature, the total stress along both $x$- and $y$-direction will be zero. The equilibrium conditions are all satisfied. In addition, the shear stress is always along the edges of the cross-section, there will be no shear stress on the circumferential surface of the bar. The traction boundary conditions are also satisfied.

#### Solution strategy

Consider a bar with a cross-section that is a hollow enclosed thin-wall, as shown in Fig.\ref{image:ThinWall}. The thickness of the wall needs not to be constant, and the shape of the cross-section can be arbitrary. Our formulation and notation follow largely the textbook \cite{boresi2002advanced}. Codes are developed for computations.  

Before our analysis, we make the following assumptions:

1. The wall thickness is small compared to the other dimensions of the cross-section.
2. The shear stress is constant through the wall thickness (It may not be constant along the boundary, unless the thickness $t$ is constant). 

<img src="./images/ThinWall.png" alt=" " width="350"/><br>
\begin{figure}
  \includegraphics[width=5cm]{}
  \caption{\label{image:ThinWall} A hollow enclosed thin-wall cross-section for bars subjected to torsion (top), and the idealized deformed membrane $w$ on the cross-section (bottom). }
\end{figure}

Based on the membrane analogy studied earlier, we know that the stress function $𝜙$ relates to the deformed membrane $w$ just by a constant factor. Because $𝜙$ must satisfy Eq.(\ref{eqdpBC}) on the boundary of the cross-section, it must be either zero or a constant there. We set $𝜙=0$ on the outer boundary of the thin-wall. On the inner boundary it must be a constant, and it is noted as $𝜙_1$. Correspondingly, $w$ must also be zero on the outer boundary, and a constant on the inner boundary and it is noted as $w_1$. This leads to the drawing on the bottom of Fig.\ref{image:ThinWall}.  

Using Eq.(\ref{eqtaup}), we have $𝜏=-\tfrac{∂ 𝜙}{∂n}$, where $n$ is normal to contour of $w$ (or 𝜙), we shall have

\begin{equation} \label{eqtauc}
𝜏=\frac{1}{c} \underbrace{\left(-\frac{∂ z}{∂n}\right)}_{\tan 𝛼} = \frac{1}{c} \tan 𝛼  = \frac{1}{c} \frac{w_1}{t} 
\end{equation}

Introducing $q$ as follows. 

\begin{equation} \label{eqqtt}
q = 𝜏t  = \frac{w_1}{c} =  𝜙_1
\end{equation}

Clearly, $q$ has a unit of [Nm], and is called **shear flow**. It must be a constant, because $𝜙_1$ is a constant. Equation (\ref{eqtauc}) shows that the shear stress 𝜏 is inversely proportional to the thickness of the wall $t$, and hence the maximum shear stress is at the smallest $t$ on the wall. 

For a thin-wall member with perimeter segments $l_1$, $l_2$,…,$l_n$, each of which with constant thickness $t_1$, $t_2$,…,$t_n$, the corresponding shear stresses become: $𝜏_1 = q/t_1, 𝜏_2 = q/t_2$, …, $𝜏_n = q/t_n$. 

#### Formula for twist angle 

We need to relate the twist angle with the shear stress. For this we use the again the membrane analogy. Considering the equilibrium of force in the $z$-direction, we have

\begin{equation} \label{eqFz00}
\sum F_z =0: pA-\oint S \tan 𝛼 dl =0 
\end{equation}

where $dl$ is the infinitely small length segment of the mean perimeter of the cross-section. Equation (\ref{eqFz00}) is rewritten as  

\begin{equation} \label{eqFzn}
\begin{split}
pA&=S\oint  \frac{w_1}{t} dl =0 \\
\frac{pA}{S} &=w_1 \oint \frac{1}{t} dl \\
\frac{pA}{cS} &=\frac{w_1}{c} \oint \frac{1}{t} dl \\
\frac{pA}{cS}&=𝜙_1 \oint \frac{1}{t} dl \\
2GA𝜗_t&=q \oint \frac{1}{t} dl  
\end{split}
\end{equation}

We thus have: 

\begin{equation} \label{eqttt}
\begin{split} 
\boxed{𝜗_t = \frac{q}{2GA}\oint \frac{1}{t} dl  = \frac{1}{2GA}\oint 𝜏 dl }\\
\end{split}
\end{equation}

This equation is the same as Eq.(\ref{eqtdzJ}).

If $t$ is piecewise constant, and there is $n$ pieces, we have

\begin{equation} \label{eq1_t}
\oint \frac{1}{t} dl = \frac{l_1}{t_1}+ \frac{l_2}{t_2} + \cdots + \frac{l_n}{t_n}=\sum_i^n \frac{l_i}{t_i}
\end{equation}

The twist angle becomes, 

\begin{equation} \label{eqt1_t}
𝜗_t = \frac{q}{2GA} \sum_i^n \frac{l_i}{t_i}
\end{equation} 

Further, if the thickness of the wall is uniformly $t$, we have

\begin{equation} \label{eqt_t}
𝜗_t = \frac{ql}{2GAt}  =  \frac{𝜏l}{2GA}
\end{equation} 

where $l$ is the length of mean perimeter of the cross-section. 

#### Torque stress relation

Using now Eq.(\ref{eqEqT}), the torque is twice the volume of the stress function. We obtain torque written in stress:

\begin{equation} \label{eqTq}
  T=2\int_A 𝜙  dxdy = 2A𝜙_1 = 2Aq = 2A𝜏t  
\end{equation}

where $A$ is the area enclosed by the mean perimeter of the cross-section. This is Eq.(\ref{eqT2qA}). 

The twist angle can be written in torque: 

\begin{equation} \label{eqtwT}
𝜗_t = \frac{𝜏l}{2GA}= \frac{Tl}{4GtA^2}
\end{equation} 

#### Equivalent polar 2nd moment of area

This can be written in the conventional form $𝜗_t = \tfrac{T}{GJ}$, if we write $J$ as: 

\begin{equation} \label{eqJt_t}
J \;\; \underset{\text{  general }}{=} \;\; \frac{4A^2}{\oint \frac{1}{t} dl} \;\; \underset{\text{ uniform }}{=} \;\;\frac{4tA^2}{l} 
\end{equation} 

This is Eq.(\ref{eqJl}). 

Assuming the thin-wall cross-section is a thin hollow cylinder with uniform thickness $t$, Eq.(\ref{eqJt_t}) becomes,

\begin{equation} \label{eqJtcy}
J =\frac{4tA^2}{l} ≈ \frac{4t(𝜋R)^2}{2𝜋R}=  2𝜋R^3t
\end{equation} 

where $R$ is the mean radius. This formula is Eq.(\ref{eqJR}) we obtained earlier. This means that Eq.(\ref{eqJt_t}) is the general form for computing the equivalent polar moment of area for closed thin-wall cross-sections. 

If the thin-wall cross-section is a hollow rectangle with uniform thickness $t$, Eq.(\ref{eqJt_t}) becomes,

\begin{equation} \label{eqJtrct}
J =\frac{4tA^2}{l}\underset{\text{ rectangle }}{≈} \frac{4t(hb)^2}{h+b} \underset{\text{ sqaure }}{=} 8b^3t 
\end{equation} 

where $h$ is the mean half-height, and $b$ is the mean half-width of the hollow rectangle.  

#### Example: Bars with thin-wall circular cross-sections 

Consider two bars of thin-wall circular cross-section. Bar-1's cross-section is closed, and bar-2's cross-section has a small cut, as shown in Fig.\ref{image:TwoCircularThinWalls}.

<img src="./images/TwoCircularThinWalls.png" alt=" " width="300"/><br>
\begin{figure}
  \includegraphics[width=5cm]{}
  \caption{\label{image:TwoCircularThinWalls} Two thin-wall circular cross-section. One is closed (left), and one has a small cut (right).}
\end{figure}

Its dimensions are: 
> outer diameter is 22 mm; inner diameter is 18 mm. 
1. Calculate the $J$ values of bar-1 without assuming it is a thin-wall. 
2. Calculate the $J$ values of bar-1 assuming it is a thin-wall. 
3. Calculate the $J$ values of bar-2 assuming it is a narrow channel. 
4. Assume the material shear stress is allowed at 150MPa. Calculate the torque, based on the three assumptions in items 1,2, and 3. 
5. Assume the material shear stress is allowed at 150MPa. Calculate the twist angle per unit length, based on the three assumptions in items 1,2, and 3. 

**Solutions**: 

We write the following code the solve and answer these questions. 

```python
d_outer = .022; d_inner = .018
t = (d_outer - d_inner)/2                                 # thickness
r = d_outer/2                            # radius of the outer circle
R_mean = (d_outer + d_inner)/4
A_mean = np.pi*R_mean**2
l = 2*np.pi*R_mean

J_exact = np.pi/32*(d_outer**4-d_inner**4)     # Q1
J_thinW = 4*t*A_mean**2/l                      # Q2
k1 = k1_f(l/t, 21)
J_chnnl = k1*l*t**3                            # Q3

print(f"k1 = {k1} ")
print(f"J_exact = {J_exact} (mm^4)")
print(f"J_thinW = {J_thinW} (mm^4)")
print(f"J_chnnl = {J_chnnl} (mm^4)")
print(f"J_chnnl/J_thinW = {J_chnnl/J_thinW}")
```

It is found that the torsional stiffness of the bar with a cut is only ~1.3% of the closed one. 

```python
# Q4
𝜏_allow = 70.0e6          # Pa
T_exact = 𝜏_allow * J_exact / r
T_thinW = 𝜏_allow * J_thinW / R_mean       # or use: 2*A_mean*𝜏_allow*t
T_chnnl = 𝜏_allow * J_chnnl / t

print(f"T_exact = {T_exact} (Nm)")
print(f"T_thinW = {T_thinW} (Nm)")
print(f"T_chnnl = {T_chnnl} (Nm)")
print(f"T_chnnl/T_thinW = {T_chnnl/T_thinW}")
```

It is found that the loading capacity of the bar with a cut is only ~6.5% of the closed one. 

```python
# Q5
G = 77.5e9                # Pa
𝜗_exact = T_exact / J_exact / G
𝜗_thinW = T_thinW / J_thinW / G     
𝜗_chnnl = T_chnnl / J_chnnl / G

print(f"𝜗_exact = {𝜗_exact} (Nm)")
print(f"𝜗_thinW = {𝜗_thinW} (Nm)")
print(f"𝜗_chnnl = {𝜗_chnnl} (Nm)")
print(f"𝜗_chnnl/𝜗_thinW = {𝜗_chnnl/𝜗_thinW}")
```

It is found that the twist angle of the bar with a cut is about 5 times the closed one, for the same allowable stress level. 

In conclusion, a cut significantly reduces the loading capacity and the stiffness of a torsional bar with a thin-wall cross-section.
