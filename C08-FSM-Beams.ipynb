## Solution to bending beams

### Thin beam theory

The other widely used structure member is the beam, as shown in Fig.\ref{image:beam}. It is similar to 1D bar studied in the previous chapter in term of slender geometry. The difference is that a beam is subjected to forces in the transverse $y$ direction, or moments with respect to $z$ axis, resulting bending deformation. 
 
\begin{figure}
  \includegraphics[width=5cm]{./images/beam.png}
  \caption{\label{image:beam} A 1D beam member. It is subjected to forces in the transverse $y$ direction, resulting bending deformation.}
\end{figure} 

The primary variable for the beam is the deflection, the displacement in the $y$ direction $v$, on its neutral axis, where the normal stress $𝜎_xx$ is zero. Because the deflection $v$ changes only along the $x$ axis, it is essentially one-dimensional (1D) problem, after a proper simplification process. 

For a 2D problem, there are two unknown displacement components: $u$ and $v$. To simplify the problem to 1D, we must found a way to relate one of the displacement, say $u$, to $v$. This needs to be done in a way that captures the major mechanics behavior of the beam under bending.  

To quantify the bending behavior of the beam, we need to make some assumptions based some theory. One of which is the so-called **Euler-Bernoulli beam theory**. It describes well the behavior of thin beams under bending, and hence is most widely used. This theory makes two major assumptions: 

 1) The cross-section of the beam remains straight during bending. 
 
 2) The cross-section of the beam (perpendicular to its axis) remains perpendicular to its axis during bending. 

Figure \ref{image:beamDeflection} is a schematic draw of a deflected beam.  

<img src="./images/beamDeflection.png" alt=" " width="450"/><br>
\begin{figure}
  \includegraphics[width=5cm]{}
  \caption{\label{image:beamDeflection} Beam under bending. This neural axis that representing the beam has a deflection, while the cross-section of the beam has a rotation. The slope of the deflection cure and the rotation angle is the same due to the Euler-Bernoulli assumptions.}
\end{figure} 

### Normal stress and the resulted moment

These two assumptions means that the shear strain $\varepsilon_{xy}$ is negligible. Otherwise the cross-section and the neutral axis will not be perpendicular, when the beam is bent. This leads to 

\begin{equation} \label{eqexy}
\varepsilon_{xy}=\frac{1}{2} (\frac{∂ u}{∂ y} +\frac{∂ v}{∂ x}) = 0
\end{equation}

where $\frac{∂ v}{∂ x}$ is the **slope** of the deflection. Because of the 2nd assumption, the rotation of the cross-section $\theta$ is the same as the slope of the deflection (see Fig.\ref{image:beamDeflection}). We thus have: 

\begin{equation} \label{eqtheta}
𝜃 = \frac{∂ v}{∂ x} 
\end{equation}

Equation (\ref{eqexy}) becomes:  

\begin{equation} \label{eqexyt}
\frac{∂ u}{∂ y} = - \theta 
\end{equation}

Integration of Eq.(\ref{eqexyt}) with respect to $y$ gives 

\begin{equation} \label{equytc}
u=-y \theta + c
\end{equation}

where $c$ is an integral constant. Because at the neutral axis ($y=0$) does not deform in the $x$ direction, and hence $u=0$. This condition leads to $c=0$. Equation (\ref{equytc}) becomes, 

\begin{equation} \label{equyt}
u=-y \theta =-y \frac{∂ v}{∂ x} 
\end{equation}

The displacement $u$ is written in the derivative of deflection $v$. Thus, two unknown displacements become only one, which is deflection $v$. We need to write other variables also in terms of only $v$. 

Using Eqs.(\ref{equyt}), the normal strain can now be evaluated using 

\begin{equation} \label{equxx}
\varepsilon_{xx}= \frac{∂ u}{∂ x} =-y \frac{∂^2 v}{∂ x^2} 
\end{equation}

The normal stress $\varepsilon_{xx}$ is written in terms of the second derivative of deflection $v$. It is linear in $y$, meaning that the normal strain distribution along $y$ is a straight line. 

Note that there is only one independent variable $x$ involved for beams under bending. Therefore, the partial derivative can be replaced by total derivative. 

For this 1D beam, the normal stress relates the normal strain by the Hook's law, and hence can be obtained using

\begin{equation} \label{eqsxx}
𝜎_{xx}=E\varepsilon_{xx}=-y E\frac{d^2 v}{d x^2} 
\end{equation}

This means the normal stress changes also linearly with $y$, as shown in Fig.\ref{image:normalStress}.  

<img src="./images/normalStress.png" alt=" " width="398"/><br>
\begin{figure}
  \includegraphics[width=5cm]{}
  \caption{\label{image:normalStress} Normal stress (positive) distribution along the cross-section of beams, and the resultant moment (positive).}
\end{figure} 


The linearly distributed normal stress results in moment $M_z$ on the cross-section, which can be evaluated using,

\begin{equation} \label{eqMz0}
M_z=-\int_A 𝜎_{xx} \mathrm{d}A=E\; \underbrace{(\int_A y^2 \mathrm{d}A)}_{I_z}\;\frac{d^2 v}{d x^2}  =EI_z\frac{d^2 v}{d x^2}
\end{equation}
or
\begin{equation} \label{eqMz}
\boxed{
M_z=EI_z\frac{d^2 v}{d x^2} }
\end{equation}

Note that on a positive plane, positive moment must be counter-clockwise. The minus sign is because the positive normal stress results in negative moment. We see now the moment is related to the deflection in 2nd derivative. In Eq.(\ref{eqMz}), $I_z$ is the **second moment of area** of the cross-section of the beam about $z$ axis:

\begin{equation} \label{eqIz}
\boxed{
I_z = \int_A y^2 \mathrm{d}A }
\end{equation}

This is the general equation for computing $I_z$ of the cross-section of a beam. Its value dependents on shape and dimensions of cross-section over which the integration is performed. Notice that because $y$ is squared, the height of the cross-section influences significantly the $I_z$ value.  

Using Eqs.(\ref{eqMz}) and (\ref{eqsxx}), the normal stress relates to the moment as

\begin{equation} \label{eqsxxMz}
\boxed{
𝜎_{xx}=-y\frac{M_z}{I_z} }
\end{equation}

This is a useful equation for computing normal stress on the cross-section of the beam for given moment. We see that the normal stress is proportional to the moment on the cross-section, meaning that larger moment results in large normal stress. Because the moment relates to the deflection $v$ in Eq.(\ref{eqMz}). The normal stress is also related to the only known displacement $v$. 

### On second moment of area of beam cross-section

The normal stress is inversely proportional to $I_z$ of the cross-section, meaning that larger $I_z$ reduces the normal stress. In other words, we would like to have beam having large $I_z$. 

For beams with rectangular cross-section with height $h$ and base $b$, $I_z$ can be computed easily using Eq.(\ref{eqIz}): 

\begin{equation} \label{eqIzR}
I_z = \int_A y^2 \mathrm{d}A = \int_{-h/2}^{h/2} y^2 b \mathrm{d}y= 2\int_{0}^{h/2} y^2 b \mathrm{d}y= 2b \frac{1}{3} y^3 \bigg|_0^{h/2} = \frac{bh^3}{12}
\end{equation}

To make $I_z$ larger, we can increase $b$ and/or $h$. Because $h$ has a cubic power, increase $h$ is far more effective than increase $b$ for given same amount of material. This means that we would like to make the beam as tall as possible, which is what the engineering designer is doing. However, there is a limit to it, because beam with too large $h/b$ ratio will have stability issues, which will be discussed in Chapter 12??.

A better alternative is to design the beam cross-section with a better shape. The objective is to move the material as farther as possible away from the neutral axis of the beam. A typical such a design is the I-beam. This kind of study relates to the property of an area, which will discussed in more detail in Chapter 11??. 

### Equilibrium Equation for beams

We now need to establish the equilibrium equation of beams subjected to loading. This is done using the usual trick of free body diagram, which is created by isolating a small representative cell from the beam with loads on. Figure \ref{image:beamEquilibrium} shows such a cell. 

<img src="./images/beamEquilibrium.png" alt=" " width="280"/><br>
\begin{figure}
  \includegraphics[width=5cm]{}
  \caption{\label{image:beamEquilibrium} A small representative cell from a beam subjected to loading, including possible inertia force. All the forces (moments) shown a positive by our sign convention.}
\end{figure}  

where 𝜌 is the density of the material. The body-force is assume to be $b_y$ that can change along $x$. It carries a unit of $[N/m]$. When the force is dynamic or the beam is subjected to dynamic excitation, there can also be inertia force, which depends on the mass of the representative cell and its acceleration, as shown in Fig. \ref{image:beamEquilibrium}. On the cross-section, there are possible moments and shear forces caused by these forces in the transversal direction. Both moments and shear forces may vary along $x$, and thus have increments corresponding to the coordinate increment $dx$. 

First, all these forces in the $y$ direction must be in equilibrium. This gives

\begin{equation} \label{eqeqy0}
dV+(b_y(x)-\rho A \ddot{v}) dx = 0  
\end{equation}
Or
\begin{equation} \label{eqeqy}
\frac{dV}{dx}=-b_y(x)+\rho A \ddot{v} 
\end{equation}

Next, all the moments about $z$ axis must be in equilibrium. This leads to

\begin{equation} \label{eqeqMz}
dM_z+dV dx - (b_y(x)-\rho A \ddot{v}) dx \; \frac{1}{2}dx  = 0  
\end{equation}

Because the 3rd term is 2nd order in the small $dx$, it can be neglected. We obtain, 

\begin{equation} \label{eqdMV}
\boxed{
\frac{dM_z}{dx}=-V }
\end{equation}

This means that the shear force on the cross-section is the negated gradient of the moment.  

Using Eq.(\ref{eqMz}), we found the shear force in terms of deflection. 

\begin{equation} \label{eqVv}
\boxed{
V=- EI_z\frac{d^3 v}{d x^3} }
\end{equation}

The shear force relates to the deflection in its 3rd derivative. Substituting Eq.(\ref{eqVv}) to (\ref{eqeqy}), we finally arrived at the dynamic equilibrium equation for beams. 

\begin{equation} \label{eqv4byd}
EI_z\frac{d^4 v}{d x^4}+\rho A \ddot{v}=b_y 
\end{equation}

For static cases, we have

\begin{equation} \label{eqv4by}
\boxed{
\frac{d^4 v}{d x^4} =\frac{b_y }{EI_z} }
\end{equation}

where $EI_z$ is often called bending stiffness of the beam. This is the well known **beam equation** based on the Euler-Bernoulli theory. It is quite similar to Eq.(7.\ref{eqEEq1ds}), but it is a 4th order differential equation (DE). This 4th order DE can be solved through integration as long as the body force $b_y$ is integrable. The integration will result in four integral constants, which need to be determined using boundary conditions (BCs). The process is essentially the same as we did in the previous chapter for 2nd order DEs, but a little more complicated because of the higher order differentials. 

In this section, we put subscript $z$ for $I$ and $M$ to indicate the bending is with respect to $z$-axis. In the following sections, we may drop subscript $z$ to lighten the subscripts.  

### Boundary conditions for beams

#### Possible BCs for beams

Based on the thin beam theory, at a cross-section (a point on its neutral axis to be more precise) has two displacement degrees of freedom (DoFs): deflection $v$ and rotation $\theta$. Their corresponding force variables are the shear force $V$ and moment $M$. Thus, there are a total of four possible variables: $v$, $\theta$, $V$, and $M$, at any cross-section. These four variables always have paired complementary relations: $v$ with $V$, $\theta$ with $M$. This means that if we know $v$ at a cross-section, we do not in general know $V$, and if we know $V$ we do not generally know $v$. The same apply to $\theta$ with $M$. Thus, we can only pre-scribe two variables at a cross-section.  

Now, any given beam will have two boundaries: left- and right-ends. Therefore, at each of these two boundary points. The possible boundary conditions for a boundary point can be given in Fig.\ref{image:BeamBCs}. As seen, we cannot pre-scribe any complementary pair of variables at a boundary point.

\end{figure}  

Because there always be two boundary point for a given beam, we always have four boundary conditions. This is just the right number we need to solve the 4th order DE given in Eq.(\ref{eqv4by}).  

#### Admissible BCs for beams

It is important to note that in a set of four BCs, there most be at least two conditions on these two displacement DoFs. This is because we must constrain the beam for rigid body movements, or we will not have a unique solution. For example, we can not have both ends of the beam have condition (d) shown in Fig.\ref{image:BeamBCs}. Because the beam can move freely without any force being generated. If condition (d) is given for one end, the other end must be condition (a), which is the often encountered cantilever beam. Also, (d) cannot go with (b) or (c), because such cases have only one displacement variable is constrained. The other rigid motion is still possible.  

On the other hand, we can have over pre-scription with displacement DoFs, because our method is displacement based. It works naturally well for indeterminate beams. For example, it is perfectly fine for beam with both ends clamped that is known as c-c condition. Simple-clamp (s-c), and roller-clamp (r-c) are also fine, meaning that unique solutions can be found.    

### Example: a cantilever beam subjected to a force at its right-end

Considering a simplest case of cantilever beam fixed at the left-end, and subjected to a concentrated force $F$ (N) at the right end, without body force, as shown in Fig.\ref{image:CantileverF}. The beam has length $L$ and a uniform cross-section with its 2nd moment of area given as $I$. 

<img src="./images/CantileverF.png" alt=" " width="450"/><br>
\begin{figure}
  \includegraphics[width=5cm]{}
  \caption{\label{image:CantileverF} A cantilever beam fixed at the left-end, subjected to a concentrated force at the right-end.}
\end{figure}

We derive, by hand, the formulas for solutions to a beam as functions of $x$, so that we can have a good idea on how Eq.(\ref{eqv4by}) can be solved. We will then write a Python code that can be used to derive formulas for more complicated problems. The code follows essentially the same process, but allows computer to do the heavy lifting of derivation. 

**Solution**: 
Because there is no body force, Eq.(\ref{eqv4by}) is reduced to: 

\begin{equation} \label{eqv40}
\frac{d^4 v}{d x^4} = 0
\end{equation}

The general solution of the deflection $v$ that satisfies Eq.(\ref{eqv40}) can be found by integration for four times, which gives, 

\begin{equation} \label{eqvcs}
v = \frac{1}{6} c_3 x^3 + \frac{1}{2} c_2 x^2+ c_1 x + c_0
\end{equation}

where $c_i, i=0,1,2,3$ are integral constants. Equation (\ref{eqvcs}) is the **general solution** for any beam without body force. It has a total of four constants, because the DE is of 4th order. For given a specified problem defined in this example, we shall use the boundary conditions (BCs) to determine these constants. 

At $x=0$, the beam is fixed and hence we have the following BCs: 

\begin{equation} \label{eqbcx0}
\begin{cases}
       v = 0 &  \\
       𝜃 = 0 &  
    \end{cases} \;\; \text{ at } x=0 
\end{equation}

To satisfy Eq.(\ref{eqbcx0}), $c_0=0$ and  $c_1=0$. Equation (\ref{eqvcs}) is reduced to

\begin{equation} \label{eqvc32}
v = \frac{1}{6} c_3 x^3 + \frac{1}{2} c_2 x^2 
\end{equation}

At $x=L$, we have the following BCs:

\begin{equation} \label{eqbcxL}
\begin{cases}
       V  = F &  \\
       M  = 0 &  
    \end{cases} \;\; \text{ at } x=L 
\end{equation}

Using Eqs.(\ref{eqVv}) and (\ref{eqMz}), we have

\begin{equation} \label{eqVM}
\begin{cases}
      -EI c_3 = F &  \\
       EI  ( c_3 L + c_2 ) = 0 &  
    \end{cases} \;\; \text{ at } x=L 
\end{equation}
which gives: 
\begin{equation} \label{eqc23}
\begin{cases}
      c_3 =-\frac{F}{EI } & \\
      c_2 =\frac{FL}{EI }  &  
    \end{cases} \;\; \text{ at } x=L 
\end{equation}

Substituting the foregoing equation to Eq.(\ref{eqvc32}), we obtain the solution for the deflection as follows. 

\begin{equation} \label{eqvF}
v = -\frac{F}{6EI } x^3 + \frac{FL}{2EI } x^2 
\end{equation}

Once the deflection is found, we can easily find all other solutions (typical for a displacement method). The rotation is then obtained using Eq.(\ref{eqtheta}):

\begin{equation} \label{eqtcanti}
𝜃 = \frac{∂ v}{∂ x} = -\frac{F}{2EI } x^2 + \frac{FL}{EI } x 
\end{equation}

The moment is found using Eq.(\ref{eqMz}):

\begin{equation} \label{eqMcanti}
M =EI \frac{d^2 v}{d x^2} = F(L- x) 
\end{equation}

The shear force is found using Eq.(\ref{eqVv}):

\begin{equation} \label{eqVcanti}
V=- EI \frac{d^3 v}{d x^3}= F   
\end{equation} 

The correctness of these solutions can be easily confirmed using the general solver: gr.solver1D4(), which will presented in great detail in the next section.  

```python
F, L = symbols('F, L')                # variables used in this example
by = 0                                              # body force [N/m] 
v = gr.solver1D4(E, I, by, l, v0, 𝜃0, vl, 𝜃l, V0, M0, Vl, Ml,key='c-f')
print(f'Is solution correct? {by/E/I==v[4]}; The solution v(x) is:')

dic = {l:L, v0:0, 𝜃0:0, Vl:F, Ml:0}      # parameters for this example
vx=[v[i].subs(dic) for i in range(len(v))]
vx                            # solution outputs are: v, 𝜃, Mx, Vx, qy 
```

The normal stress is found using Eq.(\ref{eqsxxMz})

\begin{equation} \label{eqsxxcanti}
𝜎_{xx}=-y\frac{F(L- x)}{I}
\end{equation}

The same procedure can find solutions for beams subjected to body forces, as long as the force function is integrable analytically. We shall use Sympy to do the derivation for us. 

### A Python beam solver

Our Python beam solver is developed based on the following considerations. 

1. Because Eq.(\ref{eqv4by}) is a 4th order DE, and we need to integrate it four times. This results in 4 integral constants. For beams of length $l$, we have two boundaries at $x=0$ and $x=l$. At each boundary, we have two of the four possible variables: deflection $v$, rotation $theta$, shear force $V$ and moment $M$. Therefore, we shall have just enough 4 equations to determine these integral constants for obtaining the solution. After these 4 constants are determined, we can have analytic formulas for $v, \frac{dv}{dx}, \frac{d^2v}{dx^2}, \frac{d^3v}{dx^3}$, and $\frac{d^4v}{dx^4}$. These formulas allow us to compute all the unknowns. The symbolic code function solver1D4() given below is written to complete this task. 

2. The solver1D4() accepts the following combinations of boundary conditions: c-c, s-c, c-s, f-c, and c-f. In which c stands for clamp, s for simple, and f for free. These set of BCs are all admissible. Note that when we say displacement it includes also rotation. When we say force, it includes also moment. 

3. As discussed before, because the mechanics problem is complimentary, the displacement variables are naturally paired with the force variables. For example, for the s-c BC, $v$ and $M$ are specified at $x=0$, and $v$ and $\theta$ are specified at $x=l$. This gives 4 equations for determining these 4 integral constants. Their corresponding counterpart $V$ and $\theta$ at $x=0$, and $V$ and $M$ at $x=l$ can then be computed after the beam equation is solved.  

4. Note that using symbolic computations, such as Sympy, we can set all the variables specified on the BCs as symbolic variables, and then derive general formulas for a solution set. Such a set formulas can then be used to compute many problems of beams with complicated loading and even multiple spans. Therefore, in our beam solver, all the variables specified on the BCs as defined as symbolic variables. If some or all of these variables are zero, we can simply set so to these solution formulas derived. Therefore, it is flexible and useful.  

### Method 1: step-by-step approach integration

#### Python code of beam solver

The following code uses step-by-step approach to perform the integration four times, and then uses BCs of  symbolic variables to determine the integral constants. The code is self-explanatory with comments given. 

```python
def solver1D4(E, I, by, l, v0, 𝜃0, vl, 𝜃l, V0, M0, Vl, Ml, key='c-c'):
    '''Solves the Beam Equation for integrable distributed body force: 
           u,x4=-by(x)/EI, with various boundary conditions (BCs): 
           c-c, s-c, c-s, f-c, c-f. 
    Input: EI: bending stiffness factor; by, body force; l, the length 
           of the beam; v0, 𝜃0, V0, M0, deflection, rotation, 
           shear force, moment at x=0; vl, 𝜃l, Vl, Ml, those at x=l. 
    Return: u, v_x, v_x2, v_x3, v_x4   upto 4th derivatives of v
    ''' 
    c0, c1, c2, c3 = symbols('c0, c1, c2, c3')   #integration constant
    EI = E*I                 # I is the Iz in our choosen coordinates. 
    
    # Integrate 4 times:     
    v_x3= sp.integrate(by/EI,(x, 0, x))+ c0  #ci: integration constant   
    v_x2= sp.integrate( v_x3,(x, 0, x))+ c1          
    v_x = sp.integrate( v_x2,(x, 0, x))+ c2   
    v   = sp.integrate(  v_x,(x, 0, x))+ c3   
    
    # Solve for the 4 integration constants: 
    if   key == "s-s":
        cs=sp.solve([v.subs(x,0)-v0, v_x2.subs(x,0)-M0/EI,
                     v.subs(x,l)-vl, v_x2.subs(x,l)-Ml/EI],[c0,c1,c2,c3])
    elif key == "c-c":
        cs=sp.solve([v.subs(x,0)-v0, v_x.subs(x,0)-𝜃0,
                     v.subs(x,l)-vl, v_x.subs(x,l)-𝜃l], [c0,c1,c2,c3])
    elif key == "s-c":
        cs=sp.solve([v.subs(x,0)-v0, v_x2.subs(x,0)-M0/EI,
                     v.subs(x,l)-vl, v_x.subs(x,l)-𝜃l], [c0,c1,c2,c3])
        #print('solver1D4:',cs[c0],cs[c1],cs[c2],cs[c3])
    elif key == "c-s":
        cs=sp.solve([v.subs(x,0)-v0, v_x.subs(x,0)-𝜃0,
                     v.subs(x,l)-vl, v_x2.subs(x,l)-Ml/EI],[c0,c1,c2,c3])
    elif key == "c-f":
        cs=sp.solve([v.subs(x,0)-v0, v_x.subs(x,0)-𝜃0,
           v_x3.subs(x,l)+Vl/EI, v_x2.subs(x,l)-Ml/EI], [c0,c1,c2,c3])
    elif key == "f-c":
        cs=sp.solve([v_x3.subs(x,0)-V0/EI, v_x2.subs(x,0)-M0/EI,
                     v.subs(x,l)-vl, v_x.subs(x,l)-𝜃l], [c0,c1,c2,c3])
    else:
        print("Please specify boundary condition type.")
        sys.exit()
        
    # Substitute the constants back to the integral solutions   
    v   =   v.subs({c0:cs[c0],c1:cs[c1],c2:cs[c2],c3:cs[c3]})
    v   =   v.expand().simplify().expand()
    
    v_x = v_x.subs({c0:cs[c0],c1:cs[c1],c2:cs[c2],c3:cs[c3]})
    v_x = v_x.expand().simplify().expand()
    
    v_x2=v_x2.subs({c0:cs[c0],c1:cs[c1],c2:cs[c2],c3:cs[c3]})
    v_x2=v_x2.expand().simplify().expand()
    
    v_x3=v_x3.subs({c0:cs[c0],c1:cs[c1],c2:cs[c2],c3:cs[c3]})
    v_x3=v_x3.expand().simplify().expand()
    
    v_x4 = sp.diff(v_x3,x).expand()
    
    print("Outputs form solver1D4(): v, 𝜃, M, V, qy")
    return v,v_x,(EI*v_x2).expand(),(-EI*v_x3).expand(),(v_x4).expand()
```

Let us use this function can find solutions for some bending problems. We define first necessary symbolic variables. 

```python
E, I, EI, A, l, 𝜉 = symbols('E, I, EI, A, l, 𝜉 ', nonnegative=True)
x, q = symbols('x, q ')  # geometry & force
c0, c1, c2, c3 = symbols('c0, c1, c2, c3')       #integration constant

# for displacement boundary conditions (DBCs): 
v0, 𝜃0, vl, 𝜃l = symbols('v0, 𝜃0, v_l, 𝜃_l')     # DBCs

# for force boundary conditions (FBCs): 
V0, M0, Vl, Ml = symbols('V0, M0, V_l, M_l')     # FBCs 

by = sp.Function('b_y')(x)                       # for force function
```

In such an analysis we can set all the variable units to 1, without affect the generality. We will practice this in the following analysis. After the solutions are found using the dimensionless functions, one can re-install physical solutions simply by multiplying the corresponding units, using Eq.(\ref{eqnormal}).

### Python code for find the maximum and minimum values

We write a code to find the maximum and minimum values of solutions. 

```python
def roundS(expr, n_d):           # to be used in maxminS() 
    '''To limit the number of digits to keep in a variable. 
    Usage: roundS(expr, n_d), where n_d: number of digits to keep.
    '''
    return expr.xreplace({n.evalf():round(n,n_d) 
                          for n in expr.atoms(sp.Number)})
```

```python
def maxminS(f, title="value"):
    '''Find maximum location and values of a symbolic function
    '''
    ndg = 8           # number of digits
    df = sp.diff(f, x)
    ddf = sp.diff(df, x)
    df0_points = sp.solve(df, x)     #find stationary points
    df0s = [roundS(point.evalf(),ndg) for point in df0_points]
    print(f"Stationary points for {title}: {df0s}")

    for point in df0s:
        fv = roundS((f.subs(x,point)).evalf(), ndg)
        ddfv = ddf.subs({x:point})
        if ddfv < 0:
            print(f"At x={point}, local maximum {title}={fv}")
        elif ddfv > 0:
            print(f"At x={point}, local minimum {title}={fv}")
        else:
            print(f"At x={point}, max or min {title}={fv}")

    df0s.append(0)        # Add in points on the boundaries
    df0s.append(1)
    fs_df0 = [f.subs(x, point).evalf() for point in df0s]
    f_max = roundS(max(fs_df0), ndg)
    x_max=[pnt for pnt in df0s if ma.isclose(f.subs(x,pnt),max(fs_df0))]

    print(f"\nAt x={x_max}, Max {title}={f_max}\n")
```

Because the location of the key values, maximum and/or minimum values, depends only on the function part of the solution. We can set all the parameters to unit and then find these key values.

### Example 5: Clamped-Clamped beam subjected to a point force at the middle span

Consider a beam with length $l$ with clamped-clamped (c-c) BC, as shown in Fig.\ref{image:beamC-C_P}a.  The beam has the same Young's modulus $E$, and its cross-section has a second moment of area $I$. It is subjected to a point force at the middle span A ($x=l/2$), with $x$-axis starts at the left-end of the beam and being leftwards. Derive formulas for solutions, including deflection, rotation, moment, and shear force in the beam. 

<img src="./images/beamC-C_P.png" alt=" " width="280"/><br>

\begin{figure}
  \includegraphics[width=5cm]{}
  \caption{\label{image:beamC-C_P} A beam with clamped-clamped (c-c) displacement boundary conditions (DBCs). It is subjected to a concentrated force $P$ at middle span.}
\end{figure}   

Derive the solutions, including deflection, rotation, moment, and shear force in the beam. 

**Solution**: 

This problem looks simple, but it needs some considerations. First, because force $P$ is a point force, the solution may not be continuous, and needs to be given piecewise. Since the problem is symmetric with respect to to $x=l/2$, the rotation at A must be zero. Thus, we can consider only a half of the beam, say the left half, with also the force being halved. Now, we have a beam with length of $l/2$ and clamp-roller (c-r) BCs, as shows in Fig.\ref{image:beamC-C_P}b. The general solution for the c-r beam can be found using our beam solver with c-c BC and then set the rotation at A to zero. The process and codes are as follows.   

```python
title = [["Deflection", "Rotation"], ["Moment", "Shear force"]]
# Consider constant distrubted force  
by = 0                              # one may try other force funtions
                                    # the length is halved 
v = solver1D4(E, I, by, l/2, v0, 𝜃0, vl, 𝜃l, V0, M0, Vl, Ml,key='c-c')
print(f'Is solution correct? {by/E/I==v[4]}; The solution u(x) is:')
v[0]                     # solution in a general displacement function
```

```python
# Set DBCs at the left-end and rotation at A to zero: 

vx=[v[i].subs({v0:0,𝜃0:0,𝜃l:0}) for i in range(len(v))]
vx
```

This set of solution has an unknown $v_l$. At point A, we can find the shear force using the solution obtained: 

```python
fy2=vx[3].subs(x,l/2).expand()              # at point A  
gr.printM(fy2, 'Shear force at point A, fy2:',n_dgt=4)
```

The shear force at point A must be $P/2$, we then have: 

```python
P = symbols("P")
sln = sp.solve(fy2 - P/2, vl)
gr.printM(sln[0], 'deflection at A, v_l = ',n_dgt=4)
```

Finally, we put $v_l$ back to the solution set: 

```python
# substitute v_l to the solution obtained earlier:  

sln_final =[vx[i].subs({vl:sln[0]}) for i in range(len(v))]
sln_final
```

This is the final solution to our problem, which can be found in textbooks and the open literature. 

### Plot piecewise solutions

For the foregoing problem, the solution has two pieces: one for left-span and one for the right-span of the beam. We write the following code to plot the solutions. 

```python
# Set the parameters to unit values: 

sln_x =[sln_final[i].subs({P:1,l:1,E:1,I:1}) for i in range(len(v))]
sln_x
```

```python
# set parameters: #P = 1.0; E = 1.0; I = 1.0
l1 = 0.5; l2 = 0.5

dx = 0.02
x1  = np.arange(0.0, l1, dx)            # coordinates in l1, left span
x2  = np.arange(0.0, l2, dx)           # coordinates in l2, right span
X2 = -x2 + l1   # coord. transfer, right-span is opposite to left-span
signs=[1, -1, 1, -1]  # symmetry of solutions: v is sym, 𝜃 is anti-sym

# convert Sympy solution func. to numpy arrays. 
n = len(sln_x)-1
x = symbols("x")
sln = [sp.lambdify(x, sln_x[i], 'numpy') for i in range(n)]
sx1 = [         sln[i](x1)  for i in range(n)]
sx2 = [signs[i]*sln[i](X2) for i in range(n)]

# The shear force function could be a constant. Need to generate array
if type(sx1[-1]) != np.ndarray:         # in case, func. is a constant
    sx1[-1] = float(sx1[-1])*np.ones_like(x1)    
    sx2[-1] = float(sx2[-1])*np.ones_like(X2)

X  = np.concatenate((x1, l1+x2))   # put the results in a single array
sx = [np.concatenate((sx1[i], sx2[i])) for i in range(n)]

sx_max = [np.max(abs(sx[i])) for i in range(n)]  # absolute max values
print(f"The absolute maximum solutions: {sx_max}") 
```

```python
# Plot the distribution curves: 
y_data = [sx[i]/sx_max[i] for i in range(n)]        # put function values to a list
labels=["Deflection", "Rotation", "Moment", "Shear force"]  
gr.plot_fs(X, y_data, labels, 'x',     'Solutions',   'temp')
                                    # x-label, y-label, file-name
```

The discontinuity of the solutions especially for the moment and shear force are clearly shown.  

### Example 6: Simply supported beam with two spans

Consider a beam with two spans. Span-1 has a length $l_1$, and span-2 has a length $l_2$. All the members has the same Young's modulus $E$, and its cross-section has a second moment of area $I$. The beam is simple-simple-clamp supported (s-s-c), as shown in Fig.\ref{image:beam_S-S-C_q}. It is subjected to a distributed force $q$ over span-2. 

1. Derive formulas for solutions, including deflection, rotation, moment, and shear force in the beam. 
2. Plot the distribution of all these solutions. 

<img src="./images/beam_S-S-C_q.png" alt=" " width="380"/><br>

\begin{figure}
  \includegraphics[width=5cm]{}
  \caption{\label{image:beam_S-S-C_q} A beam with two spans. It is simple-simple-clamp supported.}
\end{figure}   

**Solution**: 

Because the beam has two spans, the solution will be piecewise. Let the $x$-axis starts at point A for span-1, and $x$-axis starts at point B for span-2. Both are positive rightwards. 

For each of the span, we can find the general solution using our beam solver solver1D4(). The process and codes are as follows.   

```python
# Define variables: 
title = [["Deflection", "Rotation"], ["Moment", "Shear force"]]
E, I, l1, l2, q = symbols("E, I, l1, l2, q")
x = symbols("x")

# for displacement boundary conditions (DBCs): 
v10, 𝜃10, v1l, 𝜃1l = symbols('v10, 𝜃10, v_1l, 𝜃_1l') # DBCs for span-1
v20, 𝜃20, v2l, 𝜃2l = symbols('v20, 𝜃20, v_2l, 𝜃_2l') # DBCs for span-2

# for force boundary conditions (FBCs): 
V10, M10, V1l, M1l = symbols('V10, M10, V_1l, M_1l') # FBCs for span-1
V20, M20, V2l, M2l = symbols('V20, M20, V_2l, M_2l') # FBCs for span-2
```

#### For span-1
Let us seek for the general solution for span-1. 

```python
# Distrubted force is zero for span-1 
by = 0                                                     # BC is s-s
v1 = solver1D4(E,I,by,l1, v10,𝜃10,v1l,𝜃1l, V10, M10, V1l, M1l,key='s-s')
print(f'Is solution correct? {by/E/I==v1[4]}; The solution v(x) is:')
v1[0]                     # solution of general displacement function
```

```python
# Set known BCs for span-1: 

v1x=[v1[i].subs({v10:0,M10:0,v1l:0}) for i in range(len(v1))]
v1x                            # solution of v, 𝜃, Mx, Vx, qy
```

This set of solutions has an unknown $M_{1l}$. At point B, we can find the rotation using the general solution: 

```python
𝜃1B = v1x[1].subs(x,l1).expand()       # at point B of span-1   
gr.printM(𝜃1B, 'rotation at point B of span-1, 𝜃1B:',n_dgt=4)
```

#### For span-2
Next, let us seek for the general solution for span-2. 

```python
# General solution for span-2: (constant q)
                                                  # BC is simple-clamp
v2 = solver1D4(E,I,q,l2, v20,𝜃20,v2l, 𝜃2l, V20, M20, V2l, M2l,key='s-c')
print(f'Is solution correct? {by/E/I==v2[4]}; The solution u(x) is:')
v2[0]                    # solution of general displacement function
```

```python
# Set known BCs for span-2: 

v2x=[v2[i].subs({v20:0,v2l:0,𝜃2l:0}) for i in range(len(v2))]
v2x[1]                                        # solution of 𝜃 
```

This set of solutions has an unknown $M_{20}$. At point B of span-2, we can find the rotation using the general solution: 

```python
𝜃2B = v2x[1].subs(x,0).expand()       # at point B of span-2   
gr.printM(𝜃2B, 'rotation at point B of span-2, 𝜃2B:',n_dgt=4)
```

#### Continuity and equilibrium conditions

At point B of the beam, we must have the following continuity conditions: 

\begin{equation} \label{eqPointB}
\begin{split}
𝜃_{1B} &= 𝜃_{2B} \;\; &\text{ continuity } \\
-M_{1B} + M_{2B} &= 0 \;\; &\text{ equilibrium in moment}
\end{split}
\end{equation} 

We now use these conditions to solve for unknowns $M_{1B}$ and $M_{2B}$. using the following code. 

```python
sln = sp.solve([𝜃1B - 𝜃2B, -M1l + M20], [M1l, M20])
sln                   # solution for these unknowns
```

#### Final solutions

Submitting these unknown moments back to the general solution, gives the final solutions.  

```python
# Final solution in span-1: 
vs1x = [v1x[i].subs({M1l:sln[M1l]}) for i in range(len(v1x))]
vs1x[0] #.subs({l2:l1})           # solution of devflection v 
```

```python
vs1x[1]                             # solution of rotation 𝜃
```

```python
vs1x[2:]         # solution of moment M and shear force V
```

```python
# Final solution in span-2:  
vs2x=[v2x[i].subs({M20:sln[M20]}) for i in range(len(v2x))]
vs2x[0] #.subs({l2:l1})                     # solution of v 
```

```python
vs2x[1]                                     # solution of 𝜃
```

```python
vs2x[2:]          # solution of moment M and shear force V
```

This is the final solution to our problem, which may not be found in textbooks and the open literature. When setting $l_1=l_2$, one can obtain simplified solution that can be found in textbooks and the open literature. Readers may do so to check the solutions obtained.

#### Continuity and equilibrium conditions at point B

At point B of the frame, we must have the following continuity conditions: 

\begin{equation} \label{eqPBF}
\begin{split}
𝜃_{1B} &= 𝜃_{2B} \;\; &\text{ continuity } \\
-M_{1B} + M_{2B} &= 0 \;\; &\text{ equilibrium in moment} \\
\frac{-V_{1B} l_2}{EA} &= v_{1l} \;\; &\text{ elongation of member-2=deflection of member-1 } \\
\frac{-V_{2B} l_1}{EA} &= -v_{20} \;\; &\text{ elongation of member-1=deflection of member-2 } \\

\end{split}
\end{equation} 

We now use these conditions to solve for unknowns $𝜃_{1l}, v_{1l}, 𝜃_{20}, v_{20}$ and $v_{2l}$. using the following code. 

```python
eq1 = 𝜃1l-𝜃20
eq2 = -M1B+M2B
eq3 = -V1B*l2/(E*A)-v1l
eq4 = -V2B*l1/(E*A)+v20
eq5 = V2B-q*l2
sln = sp.solve([eq1, eq2, eq3, eq4, eq5], [𝜃1l, v1l, 𝜃20, v20, v2l])
sln[𝜃1l]                               # solution for these unknowns 
```

### Thermal stress *

#### Setting of the problem

For beams that are over-constrained (indeterminate), a temperature gradient across the beam thickness can also result in additional thermal stresses, when the material is not allowed to expand freely. The general constitutive equation is given in Eq.(5.\ref{eqSOrth0T}). For thin beams, a temperature gradient relates to the derivative of the rotation of the cross-section of the beam. The relationship can be derived as follows. 

Consider a free-body with length $dx$ from a thin beam with height $h$, as shown in Fig.\ref{image:beamThermal}. 

<img src="./images/beamThermal.png" alt=" " width="380"/><br>

\begin{figure}
  \includegraphics[width=5cm]{}
  \caption{\label{image:beamThermal} Free-body diagram for a thin beam experiencing a temperature across its thickness.}
\end{figure}

#### Governing equation for deflection caused by the temperature gradient

The temperature on the top surface of the beam is assumed $T$ and that on the bottom surface is $-T$, and it changes linearly across the thickness of the beam. Consider small displacements resulted from the temperature difference, the change in rotation $𝜃_T$ of the cross-section is evaluated as follows.  

\begin{equation} \label{eqt2T}
d𝜃_T=- \frac{du^+ dx - du^- dx}{h} =- \frac{𝛼 T - (-𝛼 T)}{h}dx  =- 𝛼 \underbrace{\frac{2T}{h}}_{T_g} dx
\end{equation}

where  $du^+ $ and $ du^-$ are, respectively, the incremental displacements on the top and bottom surface of the beam caused the the temperature there, $𝛼$ is the thermal expansion coefficient of the material of the beam, and $T_g$ is the temperature gradient. the negative sign on the front of the right-hand-side is because positive gradient results in negative (clockwise) rotation. Equation (\ref{eqt2T}) can be rewritten as: 

\begin{equation} \label{eqdtdxTg}
\frac{d𝜃_T}{dx} = - 𝛼T_g  
\end{equation}

where $\frac{d𝜃_T}{dx}$ is the curvature of the beam. This equation is similar to Eq.(7.\ref{eqBardT}}) for 1D bars. The curvature is equivalent to the strain, and the temperature gradient is to the temperature change.

Using Eqs.(\ref{eqtheta}) and (\ref{eqdtdxTg}), we obtain: 

\begin{equation} \label{eqvTTg}
\frac{∂^2 v_T}{∂ x^2} =-𝛼T_g  
\end{equation}

where $v_T$ is the deflection resulted from the temperature gradient. 

#### Solution of deflection resulted for temperature gradient

Assuming $T_g$ is constant along $x$-axis, the foregoing equation can be integrated twice, which gives

\begin{equation} \label{eqvTc}
v_T =-\frac{1}{2}𝛼T_g x^2 + c_0 x +c_1 
\end{equation}

Assuming the beam is clamped at $x=0$, we have $v_T =0$ and $\frac{dv_T}{dx}=0$. This leads to both $c_0 =0$ and $c_1=0$. Hence the deflection resulted from the temperature gradient is found as 

\begin{equation} \label{eqvTa}
v_T =-\frac{𝛼T_g}{2} 𝛼T_g x^2 
\end{equation}

If the beam is clamped at $x=l$, we have $v_T =0$ and $\frac{dv_T}{dx}=0$. This leads to both $c_0 =𝛼T_gl $ and $c_1=- \frac{𝛼T_g l^2}{2}$. Hence the deflection resulted from the temperature gradient is found as

\begin{equation} \label{eqvTal}  
v_T =-\frac{𝛼T_g}{2}  (l-x)^2 
\end{equation}

The code to find this expression is given as follows. 

```python
x, l = symbols("x, l")
vT, Tg, 𝛼, c0, c1 = symbols("v_T, T_g, 𝛼, c0, c1 ")
vT = -𝛼*Tg/2*x**2 + c0*x+c1
sln = sp.solve([vT.subs(x,l), vT.diff(x).subs(x,l)], [c0, c1])
sln
```

```python
vTl = vT.subs({c0:sln[c0], c1:sln[c1]})
vTl.factor(Tg) 
```

#### Solutions of indeterminate beams in temperature gradient

Since we have already the beam solver, and the just derived formulas for the deflection caused by a temperature gradient, finding the solutions of deflection, rotation, moment and shear force in beams is straightforward. The procedure is as follows. 

1. Find solution to beams for mechanical effects using the solver1D4(), as we did earlier, which gives the mechanical deflection $v$ for the beam. 

2. Add the thermal deflection $v_T$ to the mechanical deflection $v$. 

3. Enforce boundary conditions to the combined deflection to determine the relevant unknowns, and to obtain the solution for the final deflection. 

4. After the final deflection is obtain, using Eq.(\ref{eqtheta}) to compute the rotation, Eq.(\ref{eqVv}) to compute the shear force. 

5. The moment need to computed using Eq.(\ref{eqMz}) and (\ref{eqvTTg}):

\begin{equation} \label{eqMzT}
M_z= EI_z \left[\frac{d^2 v_{combined}}{d x^2} - \frac{∂^2 v_T}{∂ x^2}\right] = EI_z \left[\frac{d^2 v_{combined}}{d x^2} + 𝛼T_g\right]
\end{equation} 

<span style="display: inline-block; margin-left: 29px;"> This is because of the constitutive equation change due to the temperature gradient. In other words, we need to remove the thermal curvature to compute the moment. This treatment is similar to the equation derived in the previous chapter for 1D bars. </span>

6. Using Eq.(\ref{eqdMV}) to compute the shear force. 

Let us look at an example. 

#### Example 8 Simply-Clamped supported beams subjected to temperature gradient

Consider beams of length $l$ with cross-section having a second moment of area $I$. It is made of material with Young's modulus $E$. It is subjected to temperature gradient $T_g$ across its thickness. Its boundaries are simply-clamped supported (s-c). 

1. Derive formulas for solutions, including deflection, rotation, moment, and shear force in the beam. 
2. Plot the distribution of all these solutions, using unit as all its parameters. 

We write the following code to complete the task. First, define all necessary symbolic variables. 

```python
E, I, A, l = symbols('E, I, A, l ', nonnegative=True)
x, q = symbols('x, q ')  # geometry & force

# variables for displacement boundary conditions (DBCs): 
v0, 𝜃0, vl, 𝜃l = symbols('v0, 𝜃0, v_l, 𝜃_l')      

# variables for force boundary conditions (FBCs): 
V0, M0, Vl, Ml = symbols('V0, M0, V_l, M_l')     

by = sp.Function('b_y')(x)                       # for force function
```

##### Find the general displacement solution for a s-c beam

```python
by = 0                                         # no mechanical loading
v = solver1D4(E, I, by, l,v0,𝜃0,vl,𝜃l,V0,M0,Vl,Ml,key='s-c')
print(f'Is solution correct? {by/E/I==v[4]}; The solution u(x) is:')
v[0]
```

This is the general displacement solution for a s-c beam with variables on the boundaries. 

##### Add in  the displacement caused temperature gradient

```python
vT, Tg, 𝛼 = symbols("v_T, T_g, 𝛼")          # Tg: temperature graident
vT = - 𝛼*Tg/2*(l-x)**2             # disp. caused temperature gradient
v_ = vT + v[0]                                    # total displacement
v_
```

##### Find the combined displacement
For our s-c beam, the clamped boundary conditions at the right-end is known as zero. We use the BCs to simplify the general solution. 

```python
v_= v_.subs({vl:0,𝜃l:0,M0:0})      # set known values for BC variables
v_                                     # simplified total displacement
```

The combined displacement at the left-end must be zero. We thus use this condition to find variable $v_0$. 

```python
sln = sp.solve(v_.subs(x,0), v0)  # solve for displacement at left-end
sln
```

```python
v_T = v_.subs({v0:sln[0]})              # put the solution for v0 back
v_T                                     
```

This is the final solution for the displacement. It can then be used to find the other solutions: rotation, moment, and shear force. We put this set of solutions in a list: 

```python
v = [v_T, v_T.diff(x,1), E*I*(v_T.diff(x,2)+𝛼*Tg), -E*I*v_T.diff(x,3)]
v = [si.simplify() for si in v]   
v                                  # formulas for the set of solutions
```

This list contains the deflection, rotation, moments and shear force resulted from the temperature gradient. These are all functions of $x$, except the shear force that is constant.  

##### Plot the distribution of solutions

Finally, we plot the solutions using gr.plot2curveS() that finds also some of the extreme values (maximum, minimum, and values at the boundaries.)

To generate the plots for viewing the variation of the solutions, we set all the parameters to unit. 

```python
vx=[v[i].subs({E:1,I:1,l:1,q:1,Tg:1,𝛼:1}) for i in range(len(v))]
vx                            # This is a set functions of only x
```

```python
title = [["Deflection", "Rotation"], ["Moment", "Shear force"]]
for i in range(len(title)):
    gr.plot2curveS(vx[2*i:2*(i+1)], 0., 1., title=title[i]) 

plt.show()
```

It is seen that although there is no mechanical loading on the beam, the beam deforms with moments and shear force built up. This is purely due to the temperature gradient.

### Remarks

The beam solver solver1D4() developed in this chapter is a powerful tool for thin beams. It is quite easy to use and applicable to a wide range of problems, including to frames and to thermal effect analysis. 

In the existing textbooks and literature, there is a quite common approach by drawing shear force and moment diagrams, and then use Eq.(\ref{eqVv}) and/or (\ref{eqMz}) to obtained the deflection via integration. This may work for statically determinate beams, because it may be easier for drawing the shear force and moment diagrams. However, use of beam solver solver1D4() is much straightforward, and effortless. All the diagrams, deflection, rotation, moment and shear force can all obtained by one run, with much less manual operations.  

For large frame structures or frame with truss members, it is much easier to use more systematic approach based on the concept of stiffness matrix assembly, as we often do in the finite element method \cite{liu2013finite}. Using FEM, the number of members in a structure can be millions and even billions, because no manual operation for setting various types of condition are needed.
