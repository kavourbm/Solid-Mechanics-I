## Solution to axial loaded bars

Bar (or rod) is the most widely used structural member. It takes axial loads, that are applied along the axial direction of the bar. Solutions to it, in terms of both axial displacement and stress are useful in many applications. A typical example is the widely used truss structures. Therefore, a bar is also referred as a truss member. 

Consider a bar subjected to a distributed body-force $b_f$, as shown in Fig.\ref{image:bar1Dbf}. It is fixed at $x=0$. At $x=l$, it may be subjected to a concentrated force $f$ (force driven), or with a fixed given displacement (displacement driven). The problem is essentially one-dimensional (1D), meaning that our concern is mainly the unknown displacement and stress/strain varying with the coordinate $x$. 

<img src="./images/bar1Dbf.png" alt=" " width="450"/><br>
\begin{figure}
  \includegraphics[width=5cm]{}
  \caption{\label{image:bar1Dbf} A 1D bar fixed at the left-end, subjected to a distributed body-force $b_f$ (N/m), and a concentrated force at the right-end.}
\end{figure}

In theory, 1D problems can have three types, and they differ by a material constant, as discussed in Section 6.4.3??. Therefore, we consider only the 1D bar problems. The solution to the counterpart can be obtained by simply replace $E$ by $\frac{E}{1-\nu^2}$ for 1D y-constrained problems, by 
$\frac{E(1-\nu)}{(1-2\nu)(1+\nu)}$ for 1D yz-constrained problem. The governing equation for 1D bar problems has been derived in Chapter 6??, and is rewritten as follows.  

\begin{equation} \label{eqEEq1ds}
\frac{\partial^2 u}{\partial x^2} = \frac{-b_f}{EA}
\end{equation}

where $b_f$ ($N/m$) is the body-force within a 1D bar or truss member in the axial direction. For bars with uniform cross-section area $A$ ($m^2$), it relates to the body force per volume $b_x$ ($N/m^3$) as: 

\begin{equation} \label{eqbf}
b_f = b_x  A
\end{equation}

Equation (\ref{eqEEq1ds}) is a 2nd order differential equation (DE). It can then be solved easily by simply integrating Eq.(\ref{eqEEq1ds}) twice. As long as $b_f$ function can be integrated analytically, we can obtain the exact solutions for bars with different types of boundary conditions (BCs). Because the DE is of 2nd order, the integrations result in two unknown constants. Since the 1D bar has always 2 boundaries (left and right ends), which provide us the just-needed number of BCs to determine these two constants. Thus, we have a logical path to find the exact solution effectively. Let us using the following example to demonstrate this process. 

### Solution procedure

First, we use an example to derive, by hand, the formulas for displacement and stress solutions to a bar, considering a simplest case where the body force is zero. So that we can have a good idea on how Eq.(\ref{eqEEq1ds}) can be solved. We will then write a Python code that can be used to derive formulas for more complicated problems. The code follows essentially the same process, but allows computer to do the heavy lifting of derivation. 

#### Example: a bar subjected to a force at its right-end

Consider a  bar with length $L$ and a uniform cross-section area $A$. It is made of material with Young's modulus $E$. It is fixed at the left-end where $x=0$, and subjected to a force at its right-end denoted as $f_R$ at $x=L$. Derivate the formulas for the displacement and stress in the bar. 

**Solution**: 
Because there is no body force, Eq.(\ref{eqEEq1ds}) is reduced to: 

\begin{equation} \label{eq1d0}
\frac{\partial^2 u}{\partial x^2} = 0
\end{equation}

The general solution of $u$ that satisfies Eq.(\ref{eq1d0}) can be found by integration. Integrating once gives, 

\begin{equation} \label{eq1dc0}
\frac{\partial u}{\partial x } = c_0
\end{equation}

where $c_0$ is an integral constant. Integrating again gives,

\begin{equation} \label{eq1dc1}
u = c_0 x + c_1
\end{equation}

where $c_1$ is another integral constant. Equation (\ref{eq1dc1}) is the **general solution** for any bar without body force. It has a total of two constants. For given a specified problem defined in this example, we shall use the boundary conditions (BCs) to determine these constants. 

At $x=0$, the bar is fixed and hence we have the following BC: 

\begin{equation} \label{eq1dbc1}
u = 0, \;\; \text{ at } x=0 
\end{equation}

To satisfy Eq.(\ref{eq1dbc1}), $c_1$ in Eq.(\ref{eq1dc1}) has to be zero, which leads to

\begin{equation} \label{eq1dc0x}
u = c_0 x  
\end{equation}

At $x=L$, the bar loaded by $f_R$, and hence we have the following BC:

\begin{equation} \label{eq1dbc2}
𝜎 A = f_R, \;\; \text{ or } \;\; AE 𝜀= f_R, \;\; \text{ or } \;\; AE \frac{∂u}{∂x}= f_R, \;\; \text{ at } x=L  
\end{equation}

Substituting Eq.(\ref{eq1dc0x}) into Eq.(\ref{eq1dbc2}) we obtain

\begin{equation} \label{eq1dc0fr}
AE c_0 = f_R, \;\; \text{ or } \;\; c_0 = \frac{f_R}{AE } 
\end{equation}

Finally, the displacement solution is found as 

\begin{equation} \label{eq1dux}
u = \frac{f_R}{AE } x  
\end{equation}

At $x=L$, we have displacement $u_R$ as

\begin{equation} \label{equL}
\boxed{ u_R = \frac{L}{AE }f_R  }\;\; \text{ or } \;\; \boxed{ f_R = \frac{AE}{L}u_R  }
\end{equation}

where $\frac{AE}{L}$ is known as the **axial stiffness of a bar**, and $AE$ is the axial stiffness per length. 

The stress solution is next found: 

\begin{equation} \label{eq1dsx}
𝜎 = E 𝜀 = E \frac{∂u}{∂x} = \frac{f_R}{A} 
\end{equation}

Clearly this is correct solution, as expected. This is in fact the solution we found at the beginning of Chapter 6. 

The same procedure can find solutions for bars subjected to body forces, as long as the force function is integrable analytically. We shall use Sympy to do the derivation for us. 

#### Python code for derive solutions for bars 

We write a Python code function to get it done, in a step-by-step approach integrating Eq.(\ref{eqEEq1ds}), as we did in the previous example. The code is self-explanatory by comment lines. 

```python
def solver1D2(E, A, bf, l, bc0, bcl, key='disp'):
    
    '''Solves mechanics prolems for 1D bars subjected to body 
    force bf: u,xx=-bf(x)/EA, with BCs: u(0)=bc0; u(l)=bcl or fR=bcl.
    Input: bf, body force; l, the length of the bar; 
           bcl, condition at x=l; 
           key="disp", displacement condition at a boundary;
           otherwise a force condition
    Return: u, E*u_x, u_xx ''' 
    
    # Integrate twice to obtain the general solutions in u_x and u: 
    u_x = sp.integrate(-bf/E/A,(x))+ c0  #c0: integration constant
    u   = sp.integrate(u_x,(x))+ c1      #c1: integration constant
    
    # Solve for the integration constants: 
    if key == "disp":
        consts=sp.solve([u.subs(x,0)-bc0, u.subs(x,l)-bcl],  [c0, c1])
    else:
        consts=sp.solve([u.subs(x,0)-bc0,u_x.subs(x,l)-bcl/E/A],[c0,c1])
        
    # Substitute the constants back to the general solutions, to obtain
    # solutions that satisfy the BCs:
    u   =   u.subs({c0:consts[c0],c1:consts[c1]})
    u   =   u.expand().simplify()
    u_x = u_x.subs({c0:consts[c0],c1:consts[c1]})
    u_x = u_x.expand().simplify()
    u_xx = sp.diff(u_x,x)
    
    print(f'output from solver1D2(): u, 𝜎, u_xx')
    return u.expand(), (E*u_x).expand(), u_xx.expand()
```

Let us use this code function to find solutions for two 1D problems, one with force boundary condition on the right-end, and the other with a displacement condition there. For both cases, the bar may be subjected to a given body force. We fix the bar at $x=0$, because it is needed to have our mechanics problem **well defined**, meaning has a physic meaningful solution with nonzero strains or stresses. By "fix" we mean that the displacement is pre-described with a fixed value that can be zero or a finite number. 

### Example: Constant body-force and concentrated force at the right-end 

Consider a bar subjected to a constant body-force and a concentrated force $f$ at $x=l$, all in its axial direction, as shown in Fig.\ref{image:bar1Dbf}. We compute the full solution: axial displacement, strain, and stress in the bar. 

#### Method 1: step-by-step approach integration

```python
E, A, l = symbols('E, A, l', nonnegative=True)   #l: length of the bar
x, q, bf, c0, c1 = symbols('x, q, b_f, c0, c1')

fL, fR, uL,uR = symbols('f_L, f_R, u_L, u_R')      
# fL: force at the Left-end; fR: force at the Right-end
# uL: displacement at the Light-end; uR: displacement at the Right-end

bf = sp.Function('b_f')(x)                       # 1D body-force (N/m)
title = ["Displacement", "Stress"]

# Consider the distrubted force is a constant q 
bf = q #0 #(2*sp.sin(x)+8)/E        # one may try other force funtions
u = solver1D2(E, A, bf, l, uL, fR, key='force')

# chech whether the DE is satisfied. 
print(f'Is solution correct? {u[2] == -bf/E/A}') 
u                    # solutions: u; E*u,x; u,xx
```

We found a full set of formulas for solutions to the 1D bar fixed with $u_L$at left end at $x=0$, subjected to distributed force $q$, and point force $f_R$ at $x=l$. The formula for the displacement is 

\begin{equation} \label{eqbaru}
u(x)= u_{L} + \frac{f_{R} x}{A E} + \frac{q l x}{A E} - \frac{q x^{2}}{2 A E} 
\end{equation} 

It is quadratic in $x$, and has a maximum value at $x=l$: 

\begin{equation} \label{eqbarum}
u(x)= u_{L} + \frac{f_{R} l}{A E} + \frac{q l^2}{2A E} 
\end{equation} 

The formula for the stress is 

\begin{equation} \label{eqbars}
𝜎(x)=  \frac{f_{R}}{A} + \frac{q l}{A} - \frac{q x}{A} 
\end{equation} 

It is linear in $x$, and has a maximum value at $x=0$:

\begin{equation} \label{eqbarsm}
𝜎(x)=  \frac{f_{R}}{A} + \frac{q l}{A} =  \frac{f_{R}+q l}{A} 
\end{equation} 

This is expected, because the root of the bar receives the total force of $(f_{R}+q l)$.

Equations (\ref{eqbaru}) and (\ref{eqbars}) are useful formulas for computing the solution to a bar subjected to uniform body force. The stress solution is exactly the same as we obtained in the first example, by setting $q=0$. 
It is also exactly the same obtained in the 2nd example in Chapter 3 (by setting $f_R=0$).

To obtain solutions for a bar with specified parameters, we simply substitute these parameters with fixed values. For example: 

```python
ux_cq=[u[i].subs({E:1,A:1,l:1,q:1,uL:0,fR:0}) for i in range(len(u))]
ux_cq                            # solution for constant body force q
```

If the distributed body force is another type function of $x$, we simply change the expression for $b_f$ in the foregoing code. For example, if the force is a sine function, we use the following code to obtained formulas of solutions.

```python
# Consider the distrubted force is a constant q 
bf = q*sp.sin(sp.pi*x)  
u = solver1D2(E, A, bf, l, uL, fR, key='force')
print(f'Is solution correct? {u[2] == -bf/E/A}') 
u
```

```python
# For a specifed set of parameters (unit values): 
ux=[u[i].subs({E:1,A:1,l:1,q:1,uL:0,fR:0}) for i in range(len(u))]
ux
```

Here, the displacement at the left-end of the bar is set to zero. All the deformation and stress/strain in this case are all driven by the forces (body-force and the force at the right-end). This types of problem is known as **force-driven** problem. 

#### Method 2: use the Sympy differential equation solver 

One my use the built-in sp.dsolve() to solve the differential equation directly.  

```python
E, A, l = symbols('E, A, l', nonnegative=True)
x, q = symbols('x, q ')
fR, uL, uR = symbols('f_R, u_L, u_R')
u = sp.Function('u')(x)                         # 1D displacement (m)
title = ["Displacement", "Stress"]

# Consider the distrubted force is a constant q 
bf = q*sp.sin(sp.pi*x)     

# Define the 2nd differential equation
diff_eq = u.diff(x, 2) + bf/E/A 

# Define the boundary conditions
BCs = {u.subs(x, 0): uL, u.diff(x).subs(x, l): fR/E/A}

# Solve the differential equation using sp.dsolve()
solution = sp.dsolve(diff_eq, u, ics=BCs)   # This is an sp.Eq object

# chech whether the DE is satisfied. 
print(f'Is solution correct? {solution.rhs.diff(x,2) == -bf/E/A}') 
solution                                       # solution equation
```

The result obtained is the same as that obtained earlier. 

#### Python code for plotting the curves for the solutions

Next, we write the following code to plot the results. 

```python
def plot2curveS(u, xL=0., xR=1., title="f_title"):
    
    '''Print out maximum values and loctions, as well as stationary 
    points, and the values at the stationary points, and boundaries.
    '''
    x = symbols('x')
    dx = 0.01; dxr = dx*10       # x-step
    xi = np.arange(xL, xR+dx, dx)
    uf = sp.lambdify((x), u[0], 'numpy')   #convert Sympy f to numpy f 
    yi = uf(xi)
    if type(yi) != np.ndarray:              #in case, uf is a constant
        #type(yi) == int or type(yi) == float: # or len(yi)==1:  
        xi = np.arange(xL, xR+dxr, dxr)
        yi = float(yi)*np.ones_like(xi)

    fig, ax1 = plt.subplots(figsize=(5.,1.), dpi=300)
    fs = 8     # fontsize
    color = 'black'
    ax1.set_xlabel('location x', fontsize=fs)
    ax1.set_ylabel(title[0], color=color, fontsize=fs)
    ax1.plot(xi, yi, color=color)
    ax1.grid(color='r',ls=':',lw=.3, which='both') # Use both tick 
    ax1.tick_params(axis='x', labelcolor=color, labelsize=fs)
    ax1.tick_params(axis='y', labelcolor=color, labelsize=fs)

    vmax = yi[yi.argmax()]
    max_l = np.argwhere(yi == vmax)
    ax1.plot(xi[max_l],  yi[max_l], 'r*', markersize=4)
    print(f'Maximum {title[0]} value={vmax:.3e}, at x={xi[max_l][0][0]}')
    
    uf = sp.lambdify((x), u[1], 'numpy') #convert Sympy f to numpy f 
    xi = np.arange(xL, xR+dx, dx)
    yi2 = uf(xi)
    if type(yi2) != np.ndarray: # or len(yi2) == 1:
        xi = np.arange(xL, xR+dxr, dxr)
        yi2 = float(yi2)*np.ones_like(xi)

    m1, m2, m3 = np.partition(abs(yi2), 2)[0:3]
    msl=[np.where(abs(yi2)==m1)[0][0],np.where(abs(yi2)==m2)[0][0],
             np.where(abs(yi2)==m3)[0][0]]
    
    vmax = yi2[yi2.argmax()]
    max_l = np.argwhere(yi2 == vmax)
    print(f'Maximum {title[1]} value={vmax:.3e}, at x={xi[max_l][0][0]}')
        
    if abs(xi[msl[2]]-xi[msl[1]])<2*dx:
        if abs(yi2[msl[2]]-0.)<abs(yi2[msl[1]]-0.): msl.pop(1)
        else: msl.pop(2) 
    if len(msl) > 2:
        if abs(xi[msl[2]]-xi[msl[0]])<2*dx:
            if abs(yi2[msl[2]]-0.)<abs(yi2[msl[0]]-0.): msl.pop(0)
            else: msl.pop(2)  
    if len(msl) > 1:
        if abs(xi[msl[1]]-xi[msl[0]])<2*dx:
            if abs(yi2[msl[1]]-0.)<abs(yi2[msl[0]]-0.): msl.pop(0)
            else: msl.pop(1)
         
    ax2 = ax1.twinx()  # instantiate second axes sharing the same x-axis

    color = 'blue'
    ax2.set_ylabel(title[1], color=color, fontsize=fs) 
    ax2.plot(xi, yi2, color=color)
    ax2.plot(xi[max_l], yi2[max_l], 'r*', markersize=4)
    ax2.plot(xi[msl], yi2[msl], 'r*', markersize=4)
    ax1.plot(xi[msl],  yi[msl], 'r*', markersize=4)
    ax2.plot(xi[0], yi2[0], 'ro', markersize=2)
    ax1.plot(xi[0],  yi[0], 'ro', markersize=2)
    ax2.plot(xi[-1], yi2[-1], 'ro', markersize=2)
    ax1.plot(xi[-1],  yi[-1], 'ro', markersize=2)
    ax2.grid(color='r',ls=':',lw=.5, which='both') # Use both tick  
    ax2.tick_params(axis='x', labelcolor=color, labelsize=fs)
    ax2.tick_params(axis='y', labelcolor=color, labelsize=fs)
    np.set_printoptions(formatter={'float': '{: 0.3e}'.format})
    print(f'Extreme {title[0]} values={yi[msl]},\n    at x={xi[msl]}')
    print(f'Critical {title[1]} values={yi2[msl]},\n    at x={xi[msl]}')
    print(f'{title[0]} values at boundary ={yi[0], yi[-1]}')
    print(f'{title[1]} values at boundary ={yi2[0], yi2[-1]}\n')
```

### Bars subject to discontinuous body forces

#### Distributed forces

In deriving the equilibrium equations, we assumed that the stresses are sufficiently smooth and their functions are **differentiable** at least at first order. When the body force is discontinuous, the stresses may not be differentiable at the force discontinuity points.  Figure \ref{image:Barq1q2} shows a bar fixed at the left-end, subjected to two distributed body-forces $q_1$ and $q_2$ (N/m), and a concentrated force $f_R$ at the right-end. 

<img src="./images/Barq1q2.png" alt=" " width="450"/><br>
\begin{figure}
  \includegraphics[width=5cm]{}
  \caption{\label{image:Barq1q2} A 1D bar fixed at the left-end, subjected to two distributed body-force, and a concentrated force at the right-end.}
\end{figure}

In this case, we need to treat the bar being formed by two bars: bar-1 has length $l_1$ and bar-2 has length $l_2$. Because this problem is statically determined, the forces applied at the right-end of each of these two sub-bars can be obtained by simple equilibrium conditions: 

For bar-2: $f_{2R} = f_R $ that is the externally applied force at the right-end of the bar. 

For bar-1, $f_{1R} = f_R + q_2 l_2$. 

Using Eq.(\ref{eqbaru}), the displacement solution for bar-1 is found as 

\begin{equation} \label{equ1x}
u_1(x)= \frac{f_{1R} x}{A E} + \frac{l_1 q_1 x}{A E} - \frac{q_1 x^{2}}{2 A E} 
\end{equation} 

where we used boundary condition $u_{L}=0$. 

At $x=l_1$, we have

\begin{equation} \label{equ1l}
u_1(l_1)= \frac{f_{1R} l_1}{A E} + \frac{q_1 l_1^2}{2A E} = \frac{l_1}{A E}(f_{1R}+ \frac{q_1 l_1}{2})= \frac{l_1}{A E}(f_{R}+q_2 l_2+\frac{q_1 l_1}{2})  
\end{equation} 

The displacement solution for bar-2 is found as 

\begin{equation} \label{equ2x}
u_2(x)= u_1(l_1)+ \frac{f_{2R} x}{A E} + \frac{ q_2 l_2 x}{A E} - \frac{q_2 x^{2}}{2 A E} = u_1(l_1)+ \frac{f_{R} x}{A E} + \frac{ q_2 l_2 x}{A E} - \frac{q_2 x^{2}}{2 A E} 
\end{equation} 

Here we used boundary condition for bar-2 at its left-end: $u_{L}=u_1(l_1)$.

Using Eq.(\ref{eqbars}), the stress solution for bar-1 is found as

\begin{equation} \label{eqs1x}
𝜎_1(x)=  \frac{f_R + q_2 l_2}{A} + \frac{q_1 l_1}{A} - \frac{q_1 x}{A} 
\end{equation} 

Here we used: $f_{1R} = f_R + q_2 l_2$.  The stress solution for bar-2 is found as

\begin{equation} \label{eqs2x}
𝜎_2(x)=  \frac{f_R}{A} + \frac{q_2 l_2}{A} - \frac{q_2 x}{A} 
\end{equation} 

#### Point force 

For point forces, the discontinuities are at the points where the forces are applied. Figure \ref{image:BarfIfR} shows a bar fixed at the left-end, subjected to two point forces $f_I$ at an interior point, and $f_R$,  at the right-end. 

<img src="./images/BarfIfR.png" alt=" " width="450"/><br>
\begin{figure}
  \includegraphics[width=5cm]{}
  \caption{\label{image:BarfIfR} A 1D bar fixed at the left-end, subjected to two point forces one at an interior point A, and one at the right-end.}
\end{figure}

In this case, we also need to treat the bar as two bars: bar-1 has length $l_1$ and bar-2 has length $l_2$. This problem is also statically determined. The forces applied at the right-end of each of these two sub-bars can be obtained by simple equilibrium conditions: 

For bar-2: $f_{2R} = f_R $ that is the externally applied force at the right-end of the bar. 

For bar-1, $f_{1R} = f_R + f_I$. Here we need to look at the left-hand-side of point A where $f_I$ is applied. 

Using Eq.(\ref{eqbaru}), the displacement solution for bar-1 is found as 

\begin{equation} \label{eqPu1x}
u_1(x)= \frac{f_{1R} x}{A E}  = \frac{(f_R + f_I)  x}{A E}
\end{equation} 

where we used boundary condition $u_{L}=0$. At $x=l_1$, we have

\begin{equation} \label{eqPu1l}
u_1(l_1)= \frac{f_{1R} l_1}{A E} = \frac{(f_R + f_I) l_1}{A E} 
\end{equation} 

The displacement solution for bar-2 is found as 

\begin{equation} \label{eqPu2x}
u_2(x)=  \frac{(f_R + f_I) l_1}{A E} + \frac{f_{R} x}{A E}  
\end{equation} 

Here we used boundary condition for bar-2 at its left-end: $u_{L}=u_1(l_1)$.

Using Eq.(\ref{eqbars}), the stress solution for bar-1 is found as

\begin{equation} \label{eqPs1x}
𝜎_1(x)=  \frac{f_R + f_I}{A} 
\end{equation} 

Here we used: $f_{1R} = f_R + f_I$.  The stress solution for bar-2 is found as

\begin{equation} \label{eqPs2x}
𝜎_2(x)=  \frac{f_R}{A} 
\end{equation} 

### Statically indeterminate bars  

Because the problem is statically indeterminate, meaning the forces applied at the right-end of each of the bars cannot be obtained by simple equilibrium conditions, extra work is needed. Figure \ref{image:bar_q12c_c} shows a bar fixed at both ends, subjected to two distributed body-forces $q_1$ and $q_2$ (N/m). 

<img src="./images/bar_q12c_c.png" alt=" " width="450"/><br>
\begin{figure}
  \includegraphics[width=5cm]{}
  \caption{\label{image:bar_q12c_c} Top: a 1D bar fixed at both ends, subjected to two distributed body-force. Bottom: the constraint at the right-end is removed, and replaced by a known reaction force $R$}
\end{figure}

For this problem, we still need to treat the bar as two bars: bar-1 has length $l_1$ and bar-2 has length $l_2$, because of the body force discontinuity. This problem is statically indeterminate, and hence the forces applied at the right-end of each of these two sub-bars cannot be obtained by equilibrium conditions alone. Therefore, we remove the constraint at the right-end, and replace it with a unknown force called reaction force $R$. After this is done, the displacements at the right-end of the bar-2 can be found using the same procedure in the previous two examples. Using Eq.(\ref{equ2x}), and replace $f_R$ by $R$, we obtain: 

\begin{equation} \label{eqRu2x}
u_2(x)= \frac{l_1}{2A E}(2R+2q_2 l_2+q_1 l_1)+ \frac{R x}{A E} + \frac{ q_2 l_2 x}{A E} - \frac{q_2 x^{2}}{2 A E} 
\end{equation} 

At $x=l_2$, $u_2 = 0$, we then found, 

\begin{equation} \label{eqRl2}
R= \frac{-1}{2(l_1+l_2)} (q_1 l_1^2 + q_2 l_2^2+2 q_2 l_1 l_2)
\end{equation} 

It is clear that the reaction force is negative, for positive body forces. Consider a special case of $q_1 =q_2=q$, Eq.(\ref{eqRl2}) becomes, 

\begin{equation} \label{eqRlq}
R= \frac{-q}{2(l_1+l_2)} (l_1^2 + l_2^2+2 l_1 l_2)= \frac{-q}{2(l_1+l_2)} (l_1 + l_2 )^2= \frac{-q(l_1+l_2)}{2}
\end{equation} 


Finally, substituting Eq.(\ref{eqRl2}), to (\ref{equ1x}) (\ref{equ2x}) we obtain the displacement solution. Substituting Eq.(\ref{eqRl2}), to (\ref{eqs1x}) (\ref{eqs2x}) we obtain the stress solution.

The above technique works well for simple problems with a small number of over-constrains. For large structural system, the over-constraints can be very large in numbers. More effective approach such as the finite element method \cite{liu2013finite} should be used. 

### Thermal stress

#### Thermal displacement

Structural components may work in environment where the temperature changes. Because of the thermal expansion of materials and the over-constrained (indeterminate) of the bar, the material is not allowed to expand freely. This can result in additional stress, known as **thermal stress**. The general constitutive equation is given in Chapter 5. For 1D bars, the constitutive equation becomes, 

\begin{equation} \label{eqBardT}
𝜀_{Txx} = \alpha ∆T  
\end{equation}

where $𝜀_{Txx}$ is the thermal strain. The displacement corresponding to the temperature change can be found from the following differential equations.  

\begin{equation} \label{eqdudxT}
\frac{∂ u_T}{∂x} = 𝜀_{Txx} = \alpha ∆T  
\end{equation}

Integrating the foregoing equation gives, 

\begin{equation} \label{equTc}
u_T = \alpha ∆T x +c 
\end{equation}

where $c$ is the integral constant. Assuming $u_T =0$ at $x=0$, we found $c=0$, and we have,

\begin{equation} \label{eqbaruT}
u_T = \alpha ∆T x  
\end{equation}

This is the displacement at $x$, when the bar is fixed at $x=0$ and subjected to a temperature change $∆T$. 

#### Thermal forces

Such a thermal displacement can simply added into the displacement expression obtained earlier. For the example of fixed-fixed bar subjected to distributed body fores discussed in the previous section, we shall use the same trick of replacing the constraint on the left-end by a unknown reaction force $R$. Using Eq.(\ref{eqRu2x}), we have the combined displacement in bar-2 as 

\begin{equation} \label{eqRu2xT}
u_2(x)= \frac{l_1}{2A E}(2R+2q_2 l_2+q_1 l_1)+ \underbrace{𝛼 ΔT l_1}_{u_{T1}}+ \frac{R x}{A E} + \frac{ q_2 l_2 x}{A E} - \frac{q_2 x^{2}}{2 A E} + \underbrace{𝛼 ΔT x}_{u_{T2}}
\end{equation} 

where the last term is the extra displacement caused by the temperature change.  

At $x=l_2$, $u_2 = 0$, we then found, 

\begin{equation} \label{eqRl2T}
R= \frac{-1}{2(l_1+l_2)} (q_1 l_1^2 + q_2 l_2^2+2 q_2 l_1 l_2)-  A E (𝛼ΔT) 
\end{equation} 

The last term is the thermal forces resulted from the temperature change. Even if all the distributed forces are zero, there will still be a stress built up in the bar.  

After the reaction force is found, the displacements in these two bars should be obtained using Eq.(\ref{equ1x})-(\ref{equ2x}). The combined displacement formula for bar-1 becomes

\begin{equation} \label{equ1xT}
u_1(x)= \frac{(R+q_2 l_2) x}{A E} + \frac{l_1 q_1 x}{A E} - \frac{q_1 x^{2}}{2 A E} + 𝛼 ΔT x
\end{equation} 

At $x=l_1$, we have

\begin{equation} \label{equ1lT}
u_1(l_1)= \frac{(R+q_2 l_2) l_1}{A E} + \frac{q_1 l_1^2}{2A E} = \frac{l_1}{2A E}(2f_{R}+q_1 l_1)  + 𝛼 ΔT l_1
\end{equation} 

The combined displacement formula for bar-2 is 

\begin{equation} \label{equ2xT}
u_2(x)= u_1(l_1)+ \frac{R x}{A E} + \frac{ q_2 l_2 x}{A E} - \frac{q_2 x^{2}}{2 A E} + 𝛼 ΔT x
\end{equation} 

#### Computing thermal stresses

##### Approach-1: use reaction and external forces

The thermal stress can be obtained by treating the reaction force $R$ as the externally applied force, in addition to other externally applied forces. In this approach, the formulas for stress remain unchanged. For stress in the bar-2, for example, the formula becomes: 

\begin{equation} \label{eqBarST1}
𝜎_{xx} = \frac{R }{A} + \frac{ q_2 l_2 }{A} - \frac{q_2 x}{A}
\end{equation} 

The thermal effects to the stress solution are introduced thorough the reaction forces resulted from the temperature change. The thermal stress can be computed in addition to the mechanical stress for bars. 

##### Approach-2: use the combined displacement

Alternatively, we can use the combined displacement and the strain-displacement relation to compute the strain. Then, compute the stress using the constitutive equation. In this case, we need to use the constitutive given in Chapter 5. 

\begin{equation} \label{eqBarST}
𝜎_{xx} = E \left[\frac{d u_{combined}}{dx} - \frac{d u_{T}}{dx} \right]= E \left[\frac{d u_{combined}}{dx} -  𝛼 ΔT \right] 
\end{equation} 

Using the foregoing equation for bar-2, for example, we shall get the same formula as given in Eq.(\ref{eqBarST1}).   

### Example: Stress in an over-constrained bar loaded mechanically and thermally 

Consider a bar with length 1.0 subjected to $q_1=1.0$ over $0<x<0.5$ and $q_2=0.0$ over $0.5<x<1.0$. Let $A=1.0$ and $E=1.0$. Compute the reaction forces, the displacement and stress distributed over the entire bar. The bar may be loaded mechanically and thermally.

**Solution**: 

We first write a code function to calculate the displacement and stress at $x$ in an arbitrary 1D bar with length of $l_1+l_2$, subjected to 2 distributed body-force $q_1$ and $q_2$, respectively, over $l_1$ and $l_2$, for any given force at the right-end. This code considers also the thermal expansion, which is set as zero as default. 

```python
def twobars_bf(x, A, E, l1, l2, q1, q2, fR, aT=0, barId=1):
    '''Calculate the displacement and stress at x in a 1D bar subjected 
    to 2 distributed body-force q1 & q2 over l1 and l2.
    fR: force applied at the right-end of the bar
    A and E: area and Youngs modulus of the bar
    aT: 𝛼*ΔT (themal expansion coefficient * change in temperature)
    barId: if 1 (default), calculate the results over l1
           if 2, calculate the results over l2
    Return: displacement u, and stress s
    '''
    if barId == 1: 
        u = ((fR+q2*l2)*x+l1*q1*x-q1*x**2/2)/(A*E) + aT*x
        s = (fR+l2*q2 + q1*l1-q1*x)/A
    else: 
        u1R = l1*((fR+q2*l2) + l1*q1/2)/(A*E) + aT*l1
        u = u1R + (fR*x+l2*q2*x-q2*x**2/2)/(A*E) + aT*x
        s = (fR + q2*l2-q2*x)/A
    return u, s 
```

Set the values for the variables involved in this problem, assuming there is not temperature change. 

```python
q1 = 1.0; q2 = 0.0
l1 = 0.5; l2 = 0.5
E  = 1.0; A  = 1.0
𝛼  = 1.0e-4                            # thermal expansion coefficient
dT = 0.0                                          # temperature change
𝛼T = 𝛼*dT 
RT = A*E*𝛼T                     # Reaction force by temperature change

# Compute the reaction force at the right-end of the bar
fR = -(q1*l1**2 + q2*l2**2+2*q2*l1*l2)/(2*l1 + 2*l2)-RT    # fR=-0.125
print(f"Reation force at the right-end = {fR}")

dx = 0.002
x1  = np.arange(0.0, l1+dx, dx)       # coordinates in l1
x2  = np.arange(0.0, l2+dx, dx)       # coordinates in l2

# Compute the displacement and stress: 
u1, s1 = twobars_bf(x1, A, E, l1, l2, q1, q2, fR, aT=𝛼T, barId=1)
u2, s2 = twobars_bf(x2, A, E, l1, l2, q1, q2, fR, aT=𝛼T, barId=2)

x  = np.concatenate((x1, l1+x2))   # put the results in a single array
ux = np.concatenate((u1, u2))
sx = np.concatenate((s1, s2))

ux_max = np.max(abs(ux))             # find the absolute maximum value
sx_max = np.max(abs(sx))
print(f"The absolute maximum displacement: {ux_max}") 
print(f"The absolute maximum stress:       {sx_max}") 

step = max(int(len(x)/5), 1)
print(f"Location x:\n {x[::step]}")           # print out some results
print(f"Displacement ux:\n {ux[::step]}")
print(f"Stress 𝜎:\n {sx[::step]}") 
print(f"Normal force N:\n {A*sx[::step]}")
```

We now plot the distribution of the solutions. In order to plot both displacement and stress together in the same figure, we divide the solutions by its corresponding absolute maximum value. Therefore, the unit of the value on these curves are their corresponding maximum values.  

```python
# Plot the distribution curves: 

x_data = [x, x]                    # put x coordinate values to a list
y_data = [ux/max(ux_max,1.e-9), sx/max(sx_max,1.e-9)]  # put to a list
labels=['Displacement', 'Stress']                   # define the label
gr.plot_fs(x_data, y_data, labels, 'x',     'u(x),  $𝜎$(x)',   'temp')
                                  # x-label, y-label,        file-name
```

Note in the forgoing examples (and some example in the future), we used unit values for the parameters involved in the problem. Users may use the actual values for a given practical problem. Any unit system can be used, but all the units for the same problem must be consistent.  

The code given above can be used for many other problems. For example, if we set $l_1=1.0$ and $l_2=0$ and keep others unchanged, we obtain the reaction forces, displacement and stress for a fixed-fixed bar subjected to a uniform body force. Readers may play with it. 

It is a good exercise for readers to write a code to compute similar case, considering a fixed-fixed bar with a point force in the interior of the bar, as shown in Fig. \ref{image:BarfIfR}.

### Example: Thermal stress in an over-constrained bar 

Consider a bar with length 1.0 subjected to a temperature change of $ΔT=1.0$.  Let $A=1.0$ and $E=1.0$. Compute the reaction forces, the displacement and stress distributed over the entire bar.  

**Solution**: 

We use the same code function twobars_bf() written earlier, but with the following setting: 

```python
q1 = 0.0; q2 = 0     # no body force, consder only the thermal effects
l1 = 1.0; l2 = 0.0
E  = 1.0; A  = 1.0
𝛼  = 1.0; dT = 1.0                    # with temperature change dT: ΔT
aT = 𝛼*dT                               # thermal expansion per length
RT = A*E*aT                     # Reaction force by temperature change

# Compute the reaction force at the right-end of the bar
fR = -(q1*l1**2 + q2*l2**2+2*q2*l1*l2)/(2*l1 + 2*l2) - RT  
print(f"Reation force at the right-end = {fR}")

dx = 0.02
x1  = np.arange(0.0, l1+dx, dx)       # coordinates in l1
x2  = np.arange(0.0, l2+dx, dx)       # coordinates in l2

# Compute the displacement and stress: 
u1, s1 = twobars_bf(x1, A, E, l1, l2, q1, q2, fR, aT, barId=1)
u2, s2 = twobars_bf(x2, A, E, l1, l2, q1, q2, fR, aT, barId=2)

x  = np.concatenate((x1, l1+x2))   # put the results in a single array
ux = np.concatenate((u1, u2))
sx = np.concatenate((s1, s2))

ux_max = np.max(abs(ux))             # find the absolute maximum value
sx_max = np.max(abs(sx))
print(f"The absolute maximum displacement: {ux_max}") 
print(f"The absolute maximum stress:       {sx_max}") 

step = max(int(len(x)/5), 1)
print(f"Location x:\n {x[::step]}")           # print out some results
print(f"Displacement ux:\n {ux[::step]}")
print(f"Stress 𝜎:\n {sx[::step]}") 
```

```python
# Plot the distribution curves: 

x_data = [x, x]                    # put x coordinate values to a list
y_data = [ux/max(ux_max,1.e-9), sx/max(sx_max,1.e-9)]  # put to a list
labels=['Displacement', 'Stress']                   # define the label
gr.plot_fs(x_data, y_data, labels, 'x',     'u(x),  $𝜎$(x)',   'temp')
                                  # x-label, y-label,        file-name
```
